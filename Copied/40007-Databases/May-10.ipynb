1. Armstrong's Axioms
   These are fundamental rules used to derive new facts from existing ones.
   - REFLEXIVITY: If you know someone's `(CID, Name)`, you obviously know their
     `CID`.
      * Logic: $CID, Name -> CID$
   - AUGMENTATION: If `CID` determines `Email`, then `CID` plus a `T-shirt Size`
     still determines `Email`.
      * Logic: a -> b ==> ac -> bc.
   - TRANSITIVITY: If `CID` -> `Dept`, and `Dept` -> `Building`, then 
     `CID` -> `Building`.
      * Logic: The chain link. 
   - UNION/DECOMPOSITION: If `CID` tells me both `Name` and `Email`, I can write
     them together as one rule or split them into two separate rules.

2. Attribute Closure ($A^+$)
   This is a high-yield... The CLOSURE of an attribute is the list of "everything
   I can figure out" starting only with that attribuute.

   HOW TO SOLVE IN AN EXAM:
   1. Start your set with the attribute itself (e.g., {CID})
   2. Look at your list of FDs. If the LHS of any FD is already in your set, add 
      the RHS to your set.
   3. Repeat until your set stops growing. 

   SQL EQUIVALENT:
      There isn't a direct SQL command for this, but conceptually, it's like a 
      recursive query. If you have a `CID`, what is the maximum number of columns
      yoy can fill in a row without looking at other data?

3. Superkeys vs. Minimal Keys vs. Prime Atributes
   Now, let's look at how we use these FDs (X -> Y) to find the "keys" of our
   table. 

   Imagine a table: `ImperialStudent(CID, Email, Fingerprint, Name, Course)`

   A. Superkey (The Overachiever)
      A superkey is a set of columns (X) that determines EVERY OTHER COLUMN in 
      the table.

      If you know the Superkey, you can find the exact row.
      - Is `CID` a superkey? Yes. `CID` -> EVERYTHING. 
      - Is `Fingerprint` a superkey? Yes. `Fingerprint` -> Everything
      - Is `(CID, Name, Course)` a superkey? Yes! If you know all three of those
        things, youd efinitely know who the student is. 

      The problem with Superkeys is they can be bloated. ... is way too much 
      information just to look someone up. 


   ---
   B. Minimal Key (Candidate Key) (The Perfect Fit)

   A Minimal Key is a Superkey that has BEEN PUT ON A DIET. 

   It is the absolute minimum number of columns needed to determine everything. 
   If you take even one column away, it breaks. 

   - Is `(CID, Name)` a Minimal key? No. Why? Because ...
   - ... `CID` IS A MINIMAL KEY... because you can't make it any smaller. 

   Note: A tabl;e can have multiple Minimal Keys (like CID and Fingerprint). When
   you design a database, you pick one to be the PRIMARY KEY. 


---
   C. Prime Attribute (The VIPs)
      A Prime Attribute is simply any single column that is part of ANY Minimal
      Key. 
      In our table, the Minimal Keys are `{CID}` and `{Fingerprint}`
      - `CID` is a Prime Attribute
      - `Fingerprint` is a Prime Attribnute
      - `Name` is NON-PRIME (it's not part of any minimal key).

1. Pseudotransitivity: The "Substitution" Rule
   - THE RULE: If $X -> Y$ and $WY -> Z$m then $WX -> Z$
   - THE CONCEPT: Think of this as algebraic substitution.
      - need W and Y to make bread... WY -> Z
      - ... you know that X can magically be turned into yeast Y... that is 
        X -> Y
      - Therefore, if you have W and X, you can still make Bread Z... because
        the flour substitutes for the yeast. 


2. The Candidate Key Algorithm: The "L-Side Cheat Code"
   ... important algorithm for normalisation questions. If asked to "Find the
   Candidate Key(s)," do not guess. Follow these exact steps:
   1. The "Never on the Right" Check: Look at all your FDs. Find the attribute
      that NEVER appear on RHS of any arrow. Because no other attribute can
      determine them, they MUST be part of every single candidate key. 
   2. THE CLOSURE CHECK: Take those mandatory attributes and calculate their
      closure (^{+}). If their closure contains every attribute in the table, you
      are done. That is your one and only Candidate Key.
   3. The Mix-and-Match?: If the mandatroy attribute don't cover the whole table,
      start pairing the, with the remaining attributes one by one. Check the
      closures of these new combinations until you find the minimal sets that
      unlock the whole table.

1. Attribute Closure (X^+) vs. FD Closure (S^+)
   - Attribute Closure (X^+): "If I start with attribute $X$, what is the total
     list of attributes I can figure out?" Think of it like connecting flights...
   - FD Closure (S^+): This is the universe of all possible rules you could
     mathematically prove using Armstrong's Axioms... Exam rarely asks you to
     compute this because it's massive... Instead, they ask you for the CANONICAL
     COVER. 
   
2. CANONICAL COVER ($S_C$)
   The Canonical Cover (also called Minimal Cover) is the absolute bare minimum
   set of rules needed to describe your database perfectly.

   You are doing three things:
   1. Breaking complex rules into simple ones. 
   2. Trimming dead weight from the left side of the arrow.
   3. Deleting unneccessary shortcut rules.

   


   Let's demystify Normalisation. The easiest way to remember 1NF, 2NF, and 3NF
   is the famous database oath:

   "Every non-key attribute must privide a fact about the KEY (1NF), the WHOLE
   KEY (2NF), and NOTHING BUT THE KEY (3NF)."

   ... Imagine a amssive, messy spreadsheet:

   `R(StudentID, ModuleID, StudentName, ModuleName, Lecturer, LecturerRoom)`

   Assume the Primary Key is the combination of `(StudentId, ModuleID)`.


---
1. First normal Form (1NF): "The Key"
   - THE RULE: Data must be atomic (one value per cepp). No lists, no arrays, no
     repeating groups. Every row must be uniquely identifiable by a key.
   - THE VIOLATION: If Alice is taking 3 modules, and you put `[Math, CS, Physics]`
     all inside a single `ModuleName` cell.
   - The Fix: Split them into separate rows so each row has exactly one module.


---
2. Second Normal Form (2NF): "The Whole Key"
   - The Rule: You are in 1NF, AND there are no PARTIAL DEPENDENCIES. Every 
     non-key attribute must depend on the entire primary key, not just part of 
     it. 
   - THE VIOLATION: Look at `StudentName`. Do you need to know the `ModuleID` to
     figure out the `StudentName`? No! `StudentID -> StudentName`. It only
     depends on part of the key. 
   - THE FIX: Break it out.
      * Table 1: `Student(StudentID, StudentName)`
      * Table 2: `Enrollment(StudentID, ModuleID, ModuleName, Lecturer, LecturerRoom)`
