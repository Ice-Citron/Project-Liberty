Part 2: Recoverability (Handling The Crashes)
   Serialisability assumes everything succeeds. Recoverability assumes things
   will fail (abort) and asks: "Can we clean up the mess without corrupting the
   database?"

   The entire concept of recoverability revolves around one specific action: 
   READING UNCOMMITTED data (Dirty Reads). If T_2 reads a value written by T_1
   before T_1 has committe, T_2 is now chained to T_1's fate.

   ... hierarchy of safety, from weakest to strongest:
   
   1. RECOVERABLE (`RC`): "Don't lock in your answer until I lock in mine."
      - The Rule: If T_2 reads from T_1, T_2 must delay its commit until after
        T_1 commits.
      - Example: w_1[X], r_2[X], w_2[Y], c_1, c_2
      - WHY IT WORKS: T_2 read uncommitted data, which is risky. But because T_2
        waited to see c_1 before issuing its own c_2, it's safe. If T_1 had aborted (a_1), the database would force T_2... abort... before T_2
        would commit... 

   2. AVOIDS CASCADING ABORTS (`ACA`): "I won't even look at your data until
      you're done."
      - The Problem with RC: In RC, if T_1 aborts, the database has to hunt
        down T_2 (and anyone who reads from T_2) and abort them all. This is a 
        "cascading abort" and it wastes a massive amount of CPU time.
      - THE RULE (ACA): Transactions are strictly forbiden from reading 
        uncommitted data. You can only read data from a transaction that has
        already committed.
      - Example: w_1[x], c_1, r_2[x], w_2[Y]
      - ... no domino effect.

   3. STRICT (`ST`): "Don't touch my data until I am done."
      - The Problem with ACA: ACA prevents DR but it doesn't prevent
        DW (blindly overwriting uncommitted data). If T_1 writes X, and T_2 
        immediately overwrites X, and then T_1 aborts,... what value should 
        the database restore `X` to? It gets messy.
      - The Rule (Strict): You cannot read OR write an object until   

   ... final major hurdle of concurrency control: LOCKING,

   If serialisation is about looking at history to see if a crash happened,
   TWO-PHASE LOCKING (2PL) is the traffic light system put in place to prevent
   the crash from happening in the first place.

---
PART 1: TWO-PHASE LOCKING (2PL)
   Imagine a shared kitchen where chefs (Transactions) need ingredients (Data).
   - READ LOCK (Shared): Like reading a recipe card on the wall. Multiple chefs
     can look at it at the same time.
   - WRITE LOCK (Exclusive): Like taking the only chopping board. Nobody else
     can use it, and nobody else can even look at it until you are done. 

THE "TWO-PHASE" RULE:
   To prevent chaos, the database enforces a strict rule... cannot
   gather new ingredients if you have started putting ingredients back.
   - ... A transaction can acquire locks but cannot release any.
   - Once a transaction releases its first lock, it cannot acquire any new locks.
   
   Why does this work? It prevents a transaction from "sneaking" back into the
   data pool after it has already let other transactions see its immediate work,
   which mathematically guarantees CONFLICT SERIALISABILITY.


---
Part 2: DEADLOCKS and the WAIT-FOR GRAPH (WFG)
   The biggest flaw of 2PL is that it causes DEADLOCKS.
   - The Scenario: Chef 1 grabs the flour (Write Lock). Chef 2 grabs the sugar
     (Write Lock). Now, Chef 1 needs sugar, and Chef 2 needs flour.
   - The Problem: Because both are in the "Growing Phase," neither is allowed to 
     put their ingredient down. ... will stare at each other till end of time...
   - THE SOLUTION: The database draws a WAITS-FOR GRAPH (WFG) in real-time.
      - If T1 waiting for T2, draw an arrow: T1 -> T2.
      - ... T2 -> T1
      - CYCLE DETECTED! The database fixes this by acting as a bouncer: it picks
        one chef, aborts their transaction, forcefully takes their ingredients,
        and lets the other chef finish.

(Exam Note: Do not confuse the WFG with Serialisation Graph... SG looks at the 
past to check for bad serialisation. The WFG looks at the present to check for
frozen locks.)

3. Aggregation, GROUP BY, and HAVIN (The Nightclub Analogy)

      out. (e.g., `HAVING SUM(drinks_bought) > 10`). This filters GROUPS, not
      rows.
