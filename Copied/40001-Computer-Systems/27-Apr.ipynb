    A Karnaugh map (K-map) is a visual tool used to simplify Boolean algebra
    expressions without relying on complex mathematical rules. It translates a
    standard truth table into a specific grid format where you can visually 
    group adjacent outputs to find the smallest, most efficient circuit design. 

How the Grid Works
    The map is arranged to that adjacent rows and columns differ by only one
    variable value at a time.
    - Instead of standard binary counting (00, 01, 10, 11), the axes use GRAY 
      CODE: 00, 01, 11, 10.
    - This specific ordering ensures that any movement up, down, left, or right
      represents a change in exactly on variable, making visual simplification
      possible. 


The Rules of Grouping (Circling)
    To minimise an expression, you draw circles (or rectangles) around the 1s
    on the grid. Every valid circle you draw becomes a single term in your final
    , simplified Boolean equation.
    - RULE of 1s: Circles can only cover 1s (and "Don't Cares"), never 0s.
    - Power of 2 Size: The extend of the circle in both directions must be a 
      power of 2. You can group 1, 2, 4, 8 or 16 cells together. But you can
      never group 3, 5, or 6 cells.
    - BIGGER IS BETTER: The larger the circle, the fewer variables will be in 
      that resulting term leading to a smaller circuit.
    - COMPLETE COVERAGE: You must cover all the 1s in the map, and a single 1 can
      be covered by multiple overlapping circles if it helps those circles larger.
    - WRAPPING: The edges of the map wrap around. The top row is adjacent to the
      bottom row, and the far-left column is adjacent to the far right-column. 

Using "Don't Cares" (X)
    In some circuit designs, certain input combinations will never physically
    occur, so you do not care what the output is. These are marked as an "X" on
    the map. You can selectively treat an X as a 1 if it helps you draw a larger
    circle around existing 1s, or treat it as a 0 if it doesn't help. ...

1. The Magic of Feedback
    To make a circuit "remember" something, you have to take its output wire and
    loop back into its own input. This is called FEEDBACK. Because the gate is
    now feeding itself, the circuit can hold onto a state (1 or 0) even after
    the original input that caused it has turned off.

Propagation Delay
   Every gate takes a small time (t_d) to compute its output. During this delay,
   the output holds its old value. This is modeled by the `switch and delay`
   model.

Feedback
   When a gate's output is connected back to its own input (directly or via
   other gates), the circuit has feedback. Feedback can create memory (latches)
   or oscillation. 

Stable vs Unstable States
    A state is STABLE if the circuit's outputs don't change after one gate delay
    (NOW outpus = NEXT outputs). Unstable states transition to other states
    until a stable state is reached -- or oscillation ocucurs. 

Oscillation
    If two unstable states point to each other, the circuit flips between them
    forever. Example: NAND with A=1 and feedback -- R flips 0->1->0->1... This 
    is how ring oscillators work.

The NOR/NAND Latch (SR-Latch)
    Two cross-coupled NOR (or NAND) gates create a bistable circuit -- it has
    TWO stable states for the same input. This is the basis of memory: the
    circuit `remembers` which stable state it fell into. 

Static Hazard / Glitch
    When an input chaanges, different signal paths through a circuit have 
    different delays. The output may briefly glitch to the wrong value before
    settling. This is a static hazard -- the output should stay constant but
    briefly blips. 


1. What does "Latch" mean?
    ... a latch implies a STABLE, SELF-SUSTAINING STATE.

    Think of a mechanical door latch. Once it clicks into place, you can let go
    of the handle, and the door stays shut. In digital logic, once you pulse the
    `Set` wire to a `1`, the internal feedback loops "locks" the output to `1`.
    Even after the `Set` wire goes back to `0`, the circuit sustains that output
    indefinitely until someone explicitly pulses the `Reset` wire. That ability
    to hold a stable state without continuous input is what gives it "memory."


2. Is Q ...
    Q is the official memory state. If Q=1, we say "the latch is storing a 1."
       - If Q = 0, we say "the latch is storing a 0."
    Q' (Q-inverse) is just a convenient by-product. Because of how the internal
    `NOR` or `NAND` gates are cross-coupled, the circuit naturally produces the
    exact opposite of Q on the other side. Hardware engineers almost always need 
    the inverted version of a signal for the next stage of their circuit, so the
    latch simply provides both for free. But when evaluatng what data is 
    actually saved, you only look at $Q$.

    A few keywords to remember:

    - COMBINATORIAL: output is a combination of ONLY INPUTS

      need a transition table or state digram, it needs some notion of time)
    - ASYNCHRONOUS: output will have correct value after an UNDEFINED amount of
