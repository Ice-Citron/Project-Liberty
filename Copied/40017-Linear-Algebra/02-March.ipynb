   THE INTUITION: You want to find the special vectors (eigenvectors) that don't
   get knocked off their span by the matrix $A$; they only get scaled by a 
   factor $\lambda$. To find these, we need the matrix $(A - \lambda I)$ to 
   squish space down so much that its volume becomes zero.

   THE COURSE TECHNIQUE: 
      1. We rewrite the eigenvector equation as $(A - \lambda I)x = 0$.
      2. For a non-zero vector to exist in the kernel, the determinant must be
         zero: $det(A - \lambda I) = 0$

-- The kernel (or null space) of a linear transformation $T : V \rightarrow W$
   is the zero vector in the codomain $W (T(v) = 0)$. It represents all inputs
   "killed" by a matrix, serving as a subspace of the input space that measures
   how far a map is from being injective. 

   - DEFINITION: For a matrix $A$, the kernel is the solution set to the
     homogeneous equation $Ax = 0$.
   - PROPERTIES: The kernel is always a linear subspace of the domain.
   - "TRIVIAL" KERNEL: If the only vector that maps to zero is the zero vector
     itself ($v = 0$), the kernel is trivial, indicating the linear 
     transformation is injective (one-to-one).
   - CALCULATION: It is found by solving $Ax = 0$ using row reduction to find
     the span of vectors (basis) corresponding to free variables.
   - RELATION TO NULLITY: The dimension of the kernel is called the nullity of 
     the linear transformation.

-- A pivot is the first non-zero entry in a row of a matrix when it is in 
   row-echelon form (REF) or reduced row-echelon form (RREF). Pivots are used in
   Gaussian elimination to create zeros below (or above) them, identifying the
   existence and uniqueness of solutions to linear systems.

   KEY DETAILS ABOUT PIVOTS:
   - DEFINITION: The leftmost, non-zero entry in a row during row reduction.
   - LOCATION: In REF, each pivot sits to the right of the pivot in the row 
     above it.
   - FUNCTION: Used to eliminate other entries in their respective columns, 
     simplifying the matrix to solve systems, calculate determinants, or find 
     the basis of a space.
   - PIVOT COLUMNS: Columns containing a pivot are pivot columns; columns
     without them correspond to free variabels.
   - EXISTENCE: If the rightmost column of an augmented matrix has a pivot, the
     system is inconsistent (no solution).
