1. The Matrix $B$: "The Standardiser"
   When you have a matrix $B$whose columns are basis vectors (v1, v2, ...), B
   is a machine that TAKES BASIS B COORIDNATES AND TURNS THEM INTO STANDARD
   COORDINATES.
      [x]_B = $[x]_B = \begin{bmatrix} 2 \\ 3 \end{bmatrix}$ <-- take 2 of B(e_1) and 3 of B(e_2)

   AM - is simply how many times a specific eigenvalue appears as a root in your
   charactersitic equation. When you calculate $det(A - \lambda I) = 0$, you get
   a polynomial. If that polynomial factors into, ... then the eigenvalue 
   lambda = 5 has an AM of 2 (because it is squared), and lambda = 2 has an AM
   of 1. Think of AM as the "theoretical count" of hwo many times that eigenvalue
   exists on paper. 

   GM is about the actual, physical geometry of the matrix transformation. It is
   the number of linearly independent eigenvectors you can actually find for that
   specific eigenvalue. Mathematiacally, it is the dimension of the eigenspace, which
   equals the number of "free variables" you get when you plug the eigenvalue back in
   and solve (A - lambda I) x = 0. GM tells you how many distinct directions in
   space are being purely scaled by that eigenvalue.
