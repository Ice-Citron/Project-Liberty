## 2D Coordinate Intuition

   A VECTOR SUBSPACE in 2D is strictly a line that passes exactly through the
   origin (0, 0). Because it passes through the origin, you can scale any vector

   somewhere else in the room, like the ceiling.

## General Formal Mathematical Explanation
   For an exam, you must define these using their formal algebraic properties in
   $\mathbf{R}^n$.

### Vector Subspace:
   A subset $U \subseteq R^n$ is a vector subspace if and only if it satisfies
   three strict closure rules:

   1. Zero Vector: 0 \in U
   2. Additivity: For all u, v \in U, the sum u + v \in U.
   3. Scalar Multipllication: For all u \in U and scalars c \in R, the product
      cu \in U. 
         Exam context: The set of solutions c \in \mathbf{R}, the product 
         cu \in U. Exam context: The set of solutions to Ax = 0 (the kernel)
         always forms a vector subspace.


### Affine Set (Affine Space):
   An affine set $L \subseteq R^n$ is formally defined as a vector subspace U
   that has been translated by a fixed position vector $p$.

   $$L = p + U = {p + u | u \in U}$$

      Exam context: The set of general solutions to $Ax = b$ forms an affine set.
      The vector $p$ is your particular solution (the "shift"), and U is the
      kernel of A (the "direction" or "shape").

   An eigenspace, denoted $E_\lambda$, is the collection of all eigenvectors that
   share the exact same eigenvalue $\lambda$, together with the zero vector.
   Mathematically, it is defined as the kernel (or nullspace) of the matrix 
   (A - \lambda I). Becuase it is defined as a kernel, it automatically forms a
   proper vector subspace that passes through the origin. Geometrically, if $A$
   is a spatial transformation, $E_\lambda$ is the specific line, plane, or hyper-plane
   where $A$ acts purely as a scalar multiplier, scaling every vector within
   that space strictly by a factor of $\lambda$ without altering their span. 

   For exam mechanics, you compute the eigensapce immediately after finding your
   eigenvalues from the characterisitc polynomial. For a given eigenvalue $\lambda_i$
   , you set up the homogeneous system $(A - \lambda_i I) x= 0$ and reduce the
   augmented matrix to RREF. The parametric solution set to this system gives
   you the eigenspace. The vectors attached to your free parameters ($s, t,$ etc.)
   from the basis of $E_{lambda_i}$.

   The most critical exam concept tied to eigenspaces is MULTIPLICITY. The
   dimension of the eigenspace (the number of free vairables or basis vectors
   you found) is called the GEOMETRIC MULTIPLICITY. 


-- A VECTOR SUBSPACE is essentially a smaller "mini-universe" living inside a 
   larger vector space. For a subset of vectors to qualify as a subspace, it
   must satisfy three strict rules: it must contain the ZERO VECTOR (the origin)
   , it must be CLOSED UNDER ADDITION (adding any two vectors in the set stays
   in the set), and it must be CLOSED UNDER SCALAR MULTIPLICATION (stretching
   any vector in the set stays in the set). If these rules hold, you can do 
   linear algebra inside  that subset without ever "falling out of it".

-- A hyperplane is a generalisation of a plane, representing an (n-1)-dimensional
   subspace (a "flat" surface) within an n-dimensional ambient space. While a
   "plane" specifically refers to a 2D surface in 3D space, a hyperplane can be
   a line in 2D or a 3D volume in 4D, always reducing the dimension by exactly
   one.

If the geometric multiplicity is strictly less than the algebraic multiplicity for any eigenvalue, there is a "shortage" of independent eigenvectors, the matrix is called defective, and it cannot be diagonalised. If $\text{GM} = \text{AM}$ for all eigenvalues, you simply take all the basis vectors from all your eigenspaces, place them as columns in a matrix $B$, and you have successfully diagonalised the matrix to $B^{-1} A B = D$.
