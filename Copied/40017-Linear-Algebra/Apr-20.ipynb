   If the rank of a matrix A \in R^{m*n} is less than the number of columns
   (r < n), the formal way to say this is that the matrix is RANK-DEFICIENT or has
   RESTRICTED RANK. 

   In the context of the linear mapping $\phi(x) = Ax$, we say the mapping is
   NOT INJECTIVE (not one-to-one). Because the columns are LINEARLY DEPENDENT,
   there exists a non-trivial KERNEL (or nullspace). Mathematically, this
   means:

   $$nullity(A) > 0$$

### Relationship to the Determinant
   ... ONLY FOR SQUARE MATRICES
   - If $A$ is a square ($n \times n$): A rank less than $n$ (rank-deficient)
     is EQUIVALENT to saying $det(A) = 0$. This means the matrix is SINGULAR
     or non-invertible.
   - If $A$ is rectnagular ($m \times n$): The determinant is NOT DEFINED. In this
     case, you cannot use the determinant to check for "dimensional collapse";
     you must rely on Gaussian elimintation to find the rank.

   In the context of "dimensional collapse" we discussed... if $A \in R^{3x5}$,
   you have more variables (5), than equations (3). This structure guarantees
   at least 5 - 3 = 2 free varaibles (assuming the equations aren't contradictory),

   1. The "Squish": When you apply the matrix $A$ to the entire space, A acts
      like a transofrmation. If the determinant is 0 (or the matrix is rank-deficient)
      , it literally squashes space down--from 3D to a 2D plane...
   2. THE KERNEL (Ax = 0): The kernel is the specific "line" or "plane" in the#
      original input space that gets squashed directly into the ORIGIN (zero
      vector).
   3. THE BASIS: A "basis" for the kernel is just the set of "unit" directions
      that define that squashed zone. If your kenel is a line, you only need 1 
      basis vector to describe its direction. If it's a plane, you need 2.
