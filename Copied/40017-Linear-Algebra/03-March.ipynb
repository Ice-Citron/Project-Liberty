   This final segment is the "payoff" of the entire eigendecomposition process.

   Here is the exact breakdown of how this final answer is constructed, and the

THE GOLDEN RULE: ORDER IS EVERYTHING
   The matrix $D$ and basis $B$ are mathematically locked together. EVERY COLUMN
   IN $D$ MUST PERFECTLY MATCH THE CORRESPONDING COLUMN IN $B$. If you put an
   eigenvalue in column 1 of $D$, its specific eigenvector must be column 1 of
   $B$. If you mix them up, the equation $A = BDB^-1$ completely breaks. 


---
STEP 1: CONSTRUCTING THE DIAGONAL MATRIX ($D$)
   The diagonal matrix $D$ is nothing but your eigenvalues listed down the main
   diagonal (top-left to bottom-right). Every other number in the matrix is 0.

   In our earlier math, we found:
   - $\lambda = 0$ (Algebraic multiplicity of 2, so we list it twice)
   - $\lambda = 3$ (Algebraic multiplicity of 1, so we list it once)

   Let's place them in order: $0, 0, 3$.

   $$D = \begin{bmatrix} \mathbf{0} & 0 & 0 \\ 0 & \mathbf{0} & 0 \\ 0 & 0 & \mathbf{3} \end{bmatrix}$$

   (Note: You could also write it as $3, 0, 0$. The math doesn't care which one
   goes first, as long as you match it in Step 2!)


---
STEP 2: CONSTRUCTING THE BASIS ($B$)
   Now we take the eigenvectors we found for each of those eigenvalues and line
   them up in the exact same order.

   - COLUMN 1: Because our dfirst diagonal entry in $D$ is 0, we must put our
     first $\lambda = 0$ eigenvector here. $\implies \begin{bmatrix} -1 \\ 1 \\ 0 \end{bmatrix}$   
   - COLUMN 2: Because our second diagonal entry in $D$ is also 0, we put our
     second $\lambda = 0$ eigenvector here. $\implies \begin{bmatrix} -1 \\ 0 \\ 1 \end{bmatrix}$
   - COLUMN 3: Because our third diagonal entry in $D$ is 3, we put our
     $\lambda = 3$ eigenvector here. $\implies \begin{bmatrix} 1 \\ 1 \\ 1 \end{bmatrix}$
