   "Multiplicity" simply means "how many times a specific eigenvalue occurs."
   However, in linear algebra, an eigenvalue can occur in two different ways:
   algebraically (as a root of an equation) and geometrically (as an independent
   direction in space).

   ALGEBRAIC MULTIPLCIITY (AM): This is found purely by looking at your
   characteristic polynomial $\phi(\lambda)$. Once you factor it, the AM is simply
   the power (exponent) of the root. For example, if you factor your polynomial and get
   $\phi(\lambda) = (5 - \lambda)^2 (3 - \lambda)^1$, then the eigenvalue
   $\lambda = 5$ has an algebraic multiplicity of 2, and $\lambda = 3$ has an 
   algebraic multiplicity of 1.

   GEOMETRIC MULTIPLICTY (GM): This is the physical dimension of the eigenspace
   ($E_{\lambda}$). In an exam, after you plug a specific $\lambda$ back into 
   $(A - I \lambda)$ and row reduce it, the GM is exactly the NUMBER OF FREE
   VARIABLES (which equals the number of basis vectors) you find for that system.

## Paragraph 1: What does `AM = 2` actually mean? (The Shear)
   Algebraic Multiplicity (AM) is just the math; it is how many times an eigenvalue 
   shows up as a root in your polynomial. If AM = 2 for lambda = 5, the math is 
   "trying" to scale space by a factor of 5 in two separate directions. Geometric
   Multiplicity (GM) is the physics; it is how many independent directions
   actually survive the transformation without being knocked off their path.
   If $GM = 2$, space behaves perfectly and scales cleanly in both directions. 
   If $GM = 1$, the transformation has a SHEAR. A shear skews space diagonally,
   which permanently knocks one of those expected eigenvectors off its span. The
   eigenvector is "lost," and you are left with only one surviving direction 
   ($GM = 1$), even though the math expected two ($AM = 2$).


## Paragraph 2: Is GM = 2 a plane?
   Yes,... For a specific eigenvalue $\lambda$, if you find two linearly
   independent eigenvectors, they form the basis of a 2D plane. Because linear
   transformation keeps grid lines parallel and evenly spaced, every single
   vector that lies flat on the plane is also an eigenvector that gets scaled
   by $\lambda$. That entire plane if your eigenspace $E_{\lambda}$. If you had
   $GM = 3$ in a 3D room, it means the entire 3D room (a full cube of space) 
   scales uniformly in all directions, like resizing a 3D model in software.


## Paragraph 3: How does $(A - \lambda I)$ create the kernel?
   By definition, an eigenvector $x$ only scales: $Ax = \lambda x$. When we 
   calculate (A - \lambda I) = 0, we are artificially shifting the entire
   transformation. We are subtracting exactly enough from the matrix so that the
   specific scaling factor $\lambda$ drops to zero. Consequently, the eigenspace
   (which normally just stretches) is suddenly completely squashed into the
   origin. Therefore, the eigensapce becomes the KERNEL (nullspace) of this new
   shifted matrix (A - \lambda I).


## Paragraph 4: Rank, Nullity, and Dimensional Collapse
   This ties directly to your question about max rank. By the Rank-Nullity
   Theorem, the dimension of the input space (say, 3D) equals the Rank (the
   dimensions kept) plus the Nullity (the dimensions squashed). Since the Nullity
   of $(A - \lambda I)$ is your Geometric Multiplicity, the math is simple: 
   3 - Rank = GM. If (A - \lambda I) squashes space into a flat 2D plane (Rank
   2), the kernel is a 2D line (GM = 1). If it squished space into a 1D line
   (Rank 1), the kernel is a 2D plane (GM = 2).


## Paragraph 5: The Change of Basis (Determinant)
   Diagnoalisation is changing your camera angle. The standard x, y, z axes
   can make transformation look incredibly messy. But if you have enough 
   independent eigenvectors (meaning GM = AM everywhere, giving you 3 total in
   3D space), you can use them as your new x, y, z axes. This is the matrix 
   $B$. When you view the transofrmation from this "eigen-perspective" 
   $(B^-1 A B)$, there are no shears and no rotations--every axis just scales
   perfectly straight. It becomes a clean, diagonal matrix $D$. If a shear
   ruined one of your eigenvectors (GM < AM), you literally don't have enough
   axes to build a 3D coordinate system... so the matrix is defective and cannot
   be diagonalised.

   Computing $A^k$ by brute-force matrix multiplication is a computational 
   nightmare, scaling terribly with every power. The elegant mathematical trick
   to bypass this relies entirely on the diagonalisation we just performed.
   If we can decompose our matrix into $A = B D B^-1$ (where $B$ contains our
   eigenvectors and $D$ contains our eigenvalues), we unlock a massive 
   algebraic shortcut. 

   The magic happens through telescoping cancellation. When we multiply $A$
   by itself, the internal basis-change matrices annhilate each other.
   Condider $A^-3$:

   $$A^3 = (BDB^{-1})(BDB^{-1})(BDB^{-1})$$

   Because matrix multiplication is assosciative, the adjacent $B^-1 B$ pairs
   collapse into the identity matrix $I$. The entire expression implodes,
   leaving only the outer $B$ matrices and a core of $D$'s multiplied together:

   $$A^k = B D^k B^{-1}$$

   Computing $D^k$ is trivial: because all off-diagonal entries are zero, you
   simply raise each individual diagonal eigenvalue to the power of $k$.

   The characteristic equation in linear algebra is the polynomial equation
   $det(A - \lambda I) = 0$, where $A$ is a square matrix, $\lambda$ is the
   eigenvalue, and $I$ is the identity matrix. It is used to determine the 
   eigenvalues of a matrix, which are essetial for understanding linear
   transformations, stability analysis, and PCA.

   Cayley-Hamilton theorem states a beautiful, almost magical fact in linear
   algebra: every square matrix satisfies its own characteristic equation. If
   you find the characteristic polynomial $\phi(\lambda)$ of a matrix, you can 
   simply swap every scalar $\lambda$ for the matrix $A$, append the identity
   matrix $I$ to any constants, and the entire expression will equal the zero
   matrix. It is the ultimate algebraic shortcut for reducing high matrix powers
   and finding inverses without relying on Gaussian elimination or heavy matrix
   multiplication.

   $\lambda$. The constant 10 becomes $10I$, and the scalar zero becomes the zero
   matrix 0. The theorem guarantees that $A^2 - 3A - 10I = 0$. This single
   master equation is the core key for rest of the problem, establishing a direct

   substitution, any infinitely high power of $A$ can be systematically squashed
   down into a simple, flat linear combination of just A and $I$.
