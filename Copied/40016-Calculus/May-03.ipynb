1. What is a Metric Space?
   A metric space is simply a set of items (which we call "points") paired with
   a  sepcific rule (a function) for measuring the "distance" between any two
   of those items. This distance function is called a METRIC, usually denoted as
   $d(x, y)$.

   For a rule to legally be called a metric, it cannot be arbitrary; it must 
   perfectly obey three specific, logical laws. 


2. The Three Metric Axioms
   For any points x, y, and z in your set, the distance function d must satisfy:

   1. Identity (Definiteness): $d(x, y) = 0 <=> x = y$
      (The distance is exactly zero iff they are the exact same objec. Also,
      distance can never be negative).
   2. Symmetry:; d(x, y) = d(y, x)
      (The distance from A to B is exactly the same as the distance from B to A)
   3. The Triangle Inequality: $d(x, z) <= d(x, y) + d(y, z)$
      (The direct path from A to C is always the shortest. Taking a detour to
      B first will either be exactly the same distance, or longer).

3. The Intuition: "Long Common Prefix = Close"
   In this specific problem, our set is all possible infinite bit strings 
   (sequences of 1s and 0s). The metric is defined as:
   - $d(b,b') = 0$ if they are identical.
   - Otherwise, d(b, b') = 1/2^n, where $n$ is the exact index of the FIRST BIT
     where they disagree.

   Think of it like looking at a file system or an IP address. The longer two
   sequences perfectly match from the very beginning, the smaller the fraction
   $1/2^n$ becomes, meaning the "distance" between them shrinks toward zero.

   Taylor's Remainder Theorem ... "If I chop off my infinite series early to 
   save time, exactly how wrong am I?"

   ...


THE CONCEPT: The Polynomial, The Remainder, and The Ghost Point
   When you approximate a function $f(x)$ using a Taylor ... polynomial of 
   degree $n$, you are splitting the function into two distinct pieces:

   $$f(x) = P_n(x) + R_n(x)$$

   - $P_n(x)$ is your finite polynomial approximation (the part you calculate)
   - $R_n(x)$ is the exact error (the infinite tail of garbage you chopped off).

   The magic of the theorem is you do not need to calculate the infinite tail to
   know how big the error is. ... Lagrange proved that the remainder can be 
   squashed into a single highly predictable formula (the Lagrange Error Bound):

   $$R_n(x) = \frac{f^{(n+1)}(c)}{(n+1)!} (x - a)^{n+1}$$

   Look closely at ... It is literally just the very next term in the Taylor 
   series (the (n+1)-th term), but with one major difference: instead of 
   evaluating the derivative at your center a, you evalaute it at some mysterious
   "ghost point" $c$.

   THE CATCH: We have absolutely no idea what $c$ is. The theorem only 
   guarantees that $c$ is trapped somewhere stricly between your center $a$ and
   your evaluation point $x$.

   THE SOLUTION: Because we do not know $c$, we simply assume the WCS. We find
   the maximum possible value that the derivative $f^{(n+1)}(c)$ could possibly
   be on that interval, and we plug that maximum value in. This gives us a 
   guaranteed upper bound on our error.
