1. What is an Affine Map ($Y = mX + c$)?
   In linear algebra, a strictly "linear" map must pass through the origin
   ($Y = mX$). An AFFINE MAP is simply a linear map that has been shifted or
   translated $(+c)$.

   For 2D data, finding an affine best fit just means finding the equation of a
   straight line, $y = mx + c$, that doesn't necessarily have to cross through
   $(0, 0)$.


2. The Sum of Squares Error (SSE)
   Imagine a scatterpolot of data points $(x_i, y_i)$. You draw a candidate line
   $y = mx + c$ through them.
      - THE RESIDUAL (Error): The vertical distance from a real data point to 
        your line is $y_i - (mx_i + c)$.
      - WHY DO WE SQUARE IT? If we just added up the raw vertical distances, a
        point 5 units above the line (+5) and a point 5 units below the line
        (-5) would cancel out to 0 error, making a terrible line look perfect. 
        Squaring forces all errors to be posiitive. Furthermore, squaring
        heavily penalises massive outliers.
      - THE OBJECTIVE FUNCTION: The total error $E$ depends entirely on your 
        choice of slope $m$ and intercept $c$.

        $$E(m, c) = \sum_{i=1}^{n} (y_i - mx_i - c)^2$$

        This creates a 3D "bowl" shape. Your goal is to find the bottom of the
        bowl.


3. THE CLEAN EXAM METHOD (For Small Data Sets)
   You will not need to memorise the statistical covariance formulas for this
   exam... will just see calculus only instead.

   1. WRITE OUT THE SPECIFIC ERROR FUNCTION: Write out the $(y_i - mx_i - c)^2$
      bracket for every single point and add them together. Do not expand the
      brackets yet!
   2. Compute $\frac{\partial E}{\partial m}$: Differentiate the whole chain 
      using the chain rule.
   3. Compute $\frac{\partial E}{\partial c}$: Differentiate the whole chain
      using the chain rule.
   4. SET TO ZERO AND SOLVE: Set both partial derivatives to 0. This will give 
      you two simultaneous linear equations (called the "Normal Equations").
      Solve them for $m$ and $c$.

1. THE CONCEPT: Uniform vs. Ordinary Continuity
   ORDINARY CONTINUITY (at a point): For a specfici point $x$, you can find a 
   tiny margin of error ($\delta$) that keeps the output wihin your target 
   ($\epsilon$). If you move to a steeper part of the curve, you have to throw
   away your old $\delta$ and calculate a new, smaller one.

   UNIFORM CONTINUITY (on an interval): "One $\delta$ to rule them all." You are
   given an $\epsilon$ target, and you can find a single, global $\delta$ that
   works for EVERY SINGLE POINT in the entire interval simultaneously. The
   $\delta$ depends only on $\epsilon$, never on $x$.


2. THE MENTAL SKETCH (The Sliding Box)
   Imagine a rigid rectangular box of width 2-delta and height 2-delta.
   - UNIFORMLY CONTINUOUS: You can slide this box freely along the entire graph,
     and the curve will always exist through the left and right sides of the box
     , never blowing through the top or bottom.
   - NOT UNIFORMLY CONTINUOUS (e.g., 1/x near 0): As you slide the box towards
     x = 0, the curve becomes infinitely steep. Eventually, it shoots straight
     out the top/bottom of your rigid box. To contain it, you'd be forced to
     shrink $\delta$ infinitely, which breaks the rule of having "one global
     $\delta$".

1. IDENTIY THE GOAL:

   $\sqrt{u+v} \le \sqrt{u} + \sqrt{v}$


2. THE PROBLEM:
   You cannot easily manipulate, split, or simplify $\sqrt{u + v}$ directy to
   prove it is smaller than the other side.


3. THE TRICK (Squaring Both Sides):
   Because both sides of our goal inequality are positive numbers, if we can 
   prove that the SQUARE of the right side is greater than or equal to the
   SQUARE of the left side, then the original inequality must also hold true.
