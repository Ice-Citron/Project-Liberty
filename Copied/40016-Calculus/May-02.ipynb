The Setup: The "PCD" Checklist
   Before you do any calculus, you take your discrete sequence $a_n$ and swap
   the $n$ for an $x$ to make it a continuous function $f(x)$.

   To legally use the test, you must briefly state that for x >= 1 (or 
   whatever your starting point is), f(x) is:

   1. Positive: All outputs are above the x-axis.
   2. Continuous: No breaks, holes or vertical asymptotes in your interval.
   3. Decreasing: The curve is constantly heading downwards. (If it's not
      obvious, you might need to take the first derivative f'(x) and show it is
      < 0).


### The Execution: The Translation   
   If it passes the PCD checklist, you completely abandon the summation symbol
   $\sum$ ... calculate the improper integral instead:

   $$\int_{1}^{\infty} f(x) \, dx$$

   (... Evaluate this by setting it up as a proper limit: 
   $\lim_{t \to \infty} \int_{1}^{t} f(x) \, dx$)


The Verdict: They Share the Same Fate
   The test simply states that the series and the integral are "linked in 
   destiny":
   - If the integral evaluates to a finite numbger (Converges) ==> The series
     CONVERGES.
   - If the integral blows up to infinity (Diverges) ==> The series DIVERGES.

   Warning: If the integral converges to a number (like 5), it does NOT mean 
   the series adds up to 5. It just proves the series adds up to some finite
   number.
