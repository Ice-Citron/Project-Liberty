### 1 & 2. Continuity 

THE CONCEPT:
   At its core, a function is continuous at x = a if you can draw the graph
   through x = a without lifting your pen. 

THE FORMAL EXAM CHECKLIST:
   For f(x) to be continuous at x = a, three things must be true:
   
   1. f(a) exists: There must be an actual, defined solid dot at x = a.
   2. $\lim_{x \to a} f(x)$ exists: As you slide along the curve from the left
      and from the right, your fingers must meet at the exact same height.
   3. THE LIMIT = the function value: $\lim_{x \to a} f(x) = f(a)$. The place
      your fingers meet must be exactly where the solid dot is.

THE SEQUENCE CRITERION (crucial...):
   You can completely bypass epsilon-deltas using sequences. $f(x)$ is 
   continuous at x = a if and only if, for EVERY sequence (x_n) that converges
   to $a$, the output sequence f(x_n) converges to f(a).

   To prove a function is DISCONTINUOUS: Find just ONE sequence x_n -> a
   where f(x_n) does not go to f(a).


---
3. DISCONTINUITY EXAMPLES

- Removable Discontinuity: Imagine $f(x) = \frac{x^2}{x}$ for $x \neq 0$, and $f(0) = 5$
  . As x -> 0, the limit is 0. But the actual function value is 5. Since 0 != 5,
  it fails step 3 of the checklist. ...
- OSCILATORY DISCONTINUITY: f(x) = sin(1/x). As x->0, the function oscillates 
  infinitely fast between -1 and 1. The limit does not exist at all, so it fails
  step 2. No matter what value you assign to f(0), it cannot be continuous.

YOUR FIRST QUESTION: Can we choose any sequence?
   Yes, absolutely!

   To prove that a function is DISCONTINUOUS at a, you only need to find ONE
   sequence x_n -> a that breaks the rule. 

   You suggested using x_n = 1/n^2. Let's see if that works...
      1. As n -> \infty, x_n = 1/n^2 -> 0. (Perfect, the inputs converge to 0).
      2. Becuase n^2 is always positive, x_n > 0 for all n.
      3. Therefore, sgn(x_n) = 1 for every n.
      4. The sequence of outputs is 1, 1, 1, ... which converges to 1. 
      5. But the actual function value is sgn(0) = 0.
      6. Since 1 != 0, the function is discontinuous at 0.

   Your sequence ... works just as perfectly as x_n = 1/n. In the exam, just
   pick the simplest sequence you can think of that does the job.

YOUR SECOND QUESTION: Why does n \to \infty instead of x -> 0?
   This is the core difference between the "Standard Limit" and the "Sequence
   Criterion".

THE STANDARD LIMIT: $\lim_{x \to 0} f(x)$
   Here, x is a continuous variable sliding smoothly along the x-axis towards
   0.

THE SEQUENCE CRITERION: $\lim_{x \to 0} f(x_n)$   
   Hee, we have swapped out the smooth sliding variable x for a discrete
   sequence of jumping points: x_1, x_2, x_3, ...

   Think about how a sequence is defined. It is a lisst of numbers indexed by
   $n \in \mathbf{n}$
