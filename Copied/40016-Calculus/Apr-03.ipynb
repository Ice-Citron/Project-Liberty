### 2. Direct Expansion vs. Substitution
- DIRECT EXPANSION: You only do this if the question explicitly forces you to
  "find the series from the definition of a Maclaurin series" or if the function
  is so bizarre that it doesn't resemble any of the standard five. You must
  take repeated derivatives f'(x), f''(x), f'''(x), evaluate them at the center,
  and plug them into the formula.

- SUBSTITUTION/MANIPULATION: You do this 95% of the time. If the exam asks for
  $e^{-x^2}$, you do not differentiate that nightmare. You take the standard
  e^x series and substitutes (-x^2) everywhere you see an $x$. If the exam asks
  for x^2 sin(x), you take the standard sin x series and multiply every term by
  x^2.


### 3. Maclaurin vs. Taylor about $x = a$  
- TAYLOR SERIES: A polynomial approximation centered at any specific point
  $x = a$.
     Formula: ...
- MACLAURIN SERIES: This is literally just a Taylor Series where the center is 
  locked exactly at $x = 0$ (so $a = 0$). All the standard expansions listed
  abvoe are Maclaurin series.      

   The RADIUS OF CONVERGENCE ($R$) is exactly what it sounds like: it is the
   "safe distance" you can travel away from the center of your power series
   before the series blows up to infinity.

   When you reate a Taylor or Maclaurin series, you are bguilding an infinite
   polynomial that perfectly matches your function at a specific center point
   (like x = 0). As you plug in x-values further and further away from that 
   center, the approximation gets strained. The Radius of Convergence tells you
   exactly where that strian breaks the math entirely.


The Number-Line Sketch (Draw this in your head or on paper)
   Draw a horizontal number line.
   Mark the center of your series (let's call it $a$). 
   For Maclaurin series, $a = 0$.
   Walk a distance of $R$ to the right to mark $(a + R)$.
   Walk a distance of $R$ to the left to mark $(a - R)$.   


THE RULE OF THE ZONE:
   - Inside the zone $|x - a| < R$: The series absolutely converges. It is safe
     to use.
   - Outsidde the zone $|x - a| > R$: The series diverges. It is useless garbage
   - EXACTLY ON THE BORDERS $x = a +- R$: Danger zone. The standard tests break
     down, and you must check these two exact points manually using basic tests
     like the Alternating Series test or p-series test.


THE TECHNIQUE TO FIND $R$
   You don't usually memorise a separate formula for $R$. Instead, you use the
   Ratio Test (or Root Test) on the entire series term, including the $x$.

   Step 1: Set up the Ratio Test: 
      $\lim_{n \to \infty} \left| \frac{\text{Term}_{n+1}}{\text{Term}_n} \right| < 1$
   Step 2: All the $n$'s will cancel or limit out, leaving you with an 
      expression involving just $|x|$ (or $|x - a|$) and some constant.
   Step 3: Rearrange the inequality to look exactly like this:
      $$|x - a| < \text{Some Number}$$
   That "some number" is your Radius, $R$.

### Stage 1: Concepts and Intuition
   To conquer these ... you need to firmly separate the "Radius" from the 
   "Interval" in your mind.


1. Radius of Convergence ($R$) vs. Interval of Convergence
   - The Radius ($R$): This is a single number. It is the "length of the leash"
     from the center of your series (x = a). It tells you how far you can 
     safely walk left or right before the math blows up.
   - The Interval: This is the actual, physical stretch of the x-axis where the
     series works. It goes from (a - R) to (a + R). But importantly, the 
     interval forces you to answer: is the exact boundary line included?


2. The Formula for the Radius
   You don't usually need a separate formula for R You jsut forcefully apply
   the RATIO TEST (or Root Test) to the entire series term isnide absolute
   value bars: $\lim_{n \to \infty} \left| \frac{u_{n+1}}{u_n} \right| < 1$.

   When you simplify this, all the $n$'s will vanish, leaving you with something
   like:

   $$|x - a| * (Some Number) < 1$$

   Divide by that number, and you get |x - a| < R. Whatever sits on the right
   side of that inequality is your radius!

1. The Supremum (`sup`)
   - WHAT IT IS: The absolute highest ceiling for a set or sequences.
   - HOW IT WORKS: It looks at every single term from the beginning of time to
     infinity and finds the absolute highest peak.
   - THE PROBLEM: It is heavily influenced by the early "noisy" terms. If a 
     sequence starts at 100, but then immediately drops to 0 and stays there 
     forever, the `sup` is 100. It doesn't tell you anything about the long-term
     behavior.


2. The Limit (`lim`)       
   - WHAT IT IS: The single exact destination a sequence settles on as 
     $n \to \infty$
   - HOW IT WORKS: It completely ignores the early terms and demands that the
     sequence eventually flattens out and zeroes in on one specific number.
   - THE PROBLEM: It is fragile. If a sequence oscilates forever (like 1, -1, 1, -1...)
     , it never settles. Because it bounces, the standard limit DOES NOT EXIST.


3. The Limit Supremum (`lim sup`)
   - WHAT IT IS: The higest ceiling of the infinite tail of the sequence. It is
     the "ultimate upper bound."
   - HOW IT WORKS: Imagine chopping off the first $N$ terms of your sequence
     and finding the `sup` (ceiling) of whatever is left. Then, chop off another 
     term and find the new ceiling. As you chop off more and more of the front,
     the ceiling might drop, but it will eventually settle down to a final 
     resting height. That resting height is the `lim sup`.           
   - THE SUPERPOWER: Unlike a normal limit, THE `lim sup` ALWAYS EXISTS, even if
     the sequence oscilates forever. For 1, -1, 1, 01..., the sequence doesn't
     have a limit, but its ultimate ceiling (lim sup) is safely 1. 


### Why Cauchy-Hadamard uses `lim sup`
   In power series, the coefficients $(c_n)$ can be messy. For example, in the
   series x + x3 + x5 + ..., the even powers are missing, meaning the
   coefficients are 1, 0, 1, 0, 1, ... 

   If you try to take the standard root test limit ($\lim \sqrt[n]{c_n}$), it
   breaks because it keeps bouncing between 1 and 0. By using the $lim sup$,
   Cauchy-Hadamard ignores the bouncing and safely grabs the "worst-case 
   scenario" (the 1s) to guarantee you find the correct radius of convergence... 






### Stage 1: Concepts and Intuition
   ... CAUCHY-HADAMARD formulas, ... given a massive shortcut.

   Normally, to find the radius of convergence, you take the entire series term
   (let's call it $u_n$), plug it into the general Ratio Test... and do the
   algebra to get the $x$ isolated on one side. 

   CAUCHY-HADAMARD implies that you can skip the algebra. It provdes the
   dedicated formulas that calculate $R$ directly by completely ignoring the
   $x$'s and only looking at the COEFFICIENTS ($c_n$).

   For a standard power series $\sum c_n (x-a)^n$, the c_n is all the "math 
   junk" attached to the $x$.


1. The Ratio Version (The Shortcut)

   $$R = \lim_{n \to \infty} \left| \frac{c_n}{c_{n+1}} \right|$$

   THE INTUITION: Notice that this formula is UPSIDE DOWN compared to the normal
   Ratio Test! Why?

   If you do the normal test: $\lim \left| \frac{c_{n+1} x^{n+1}}{c_n x^n} \right| < 1$.

   The x's cancel to leave a single $x$ on the top. To isolate the $x$, you have
   to divide both sides by the c fraction, flipping it over. Cauchy-Hadamard
   just memorises the flipped result for you.


2. The Root Version (The Formal Definition)

   $$R = \frac{1}{\limsup_{n \to \infty} \sqrt[n]{|c_n|}}$$

   THE INTUITION: Same logic. The normal root test gives you 
   $\sqrt[n]{|c_n x^n|} = |x|\sqrt[n]{|c_n|} < 1$. To get $|x|$ by itself, you
   divide by the root term, sending it to the denominator.


WHAT THIS IMPLIES FOR YOUR EXAM:
   You can completely drop the $x$'s from your working out. Identify $c_n$,
   decide if factorials/exponentials make the Ratio version better or if
   heavy n-th powers make the Root version better, and jsut plug into the
   formula. It spits out $R$ instantly.

## Stage 1: Concepts and Intuition of Power-Series Applications
   Once you have the five standard Maclaurin series memorisedd (e^x, sin x, 
   cos x, 1/(1-x) and ln(1 + x)), the exam shifts from testing your calculus to
   testing your algebra.

   The core intuition is this: POWER SERIES TURN TERRIFYING TRANSCENDENTAL
   FUNCTIONS INTO SIMPLE POLYNOMIALS. Polynomials are the easiest objects in 
   math. You can add them, subtract them, multiply them, and shove them inside 
   each other using standard algebra rules.

   ... intuition behind the five common manipulation patterns you will need:


### 1. Substitution (The "Inner Block" Trick)
   THE CONCEPT: If you know the blueprint for e^u, you know it for e^anything.
   THE INTUITION: Never use differentiation from first principles for composite
   functions like e^{-x^2} or sin(2x). It takes too long and the product/chain 
   rules get messy. Instead, take the standard series and literally swap every
   x for the new "inner block."

   - Example: To find $\cos(\sqrt{x})$, take the $\cos u = 1 - \frac{u^2}{2!} + \frac{u^4}{4!} \dots$ 
     series and substitute $u = \sqrt{x}$. It instantly becomes $1 - \frac{x}{2} + \frac{x^2}{24} \dots$


### 2. Multiplying Two Series (Polynomial FOIL with a Speed Limit)
   THE CONCEPT: Treat infinite series exactly like giant brackets of polynomials
   THE INTUITION: If you need the series for e^x sin(x), you just write their 
   two series side-by-side in brackets and expand.
      - THE CRUCIAL TRICK: TRUNCATION. Exams will usually only ask for "the 
        first three non-zero terms" or "up to the x^3 term." You must 
        aggressively ignore higher powers while you multilply. If you are 
        multiplying an x^2 term by an x^3 term, it makes an x^5 term. Don't
        even write it down... saves massive amounts of time.


### 3. Dividing by a Simple Expression (The Geometric Rewrite)
   
THE CONCEPT:  Polynomial long division is a nightmare; avoid it by turning
   division intop multiplication.

THE INTUITION: If the exam asks for $\frac{\ln(1+x)}{1+2x}$, do not try to 
   divide the series. Instead, pull it apart into $\ln(1+x) \cdot \left[ \frac{1}{1 - (-2x)} \right]$.

   Now, you can see the standard geometric exapnsion $\frac{1}{1-r}$ on that 
   second bracket (with r = -2x), and then multiply the two resulting series
   together using Pattern 2. 


### 4. Matching Coefficients (The Identity Principle)

THE CONCEPT: If two polynomials are exactly equal for all values of $x$, their
   individual pieces must be identical.

THE INTUITION: Imagine you have an equation like $C_0 + C_1x + C_2x^2 = 5 - 2x + 4x^2$.
   Because the $x$ terms must match the $x$ terms, you instantly know $C_1$ = -2.
   This pattern is heavily used if the eaxm gives you a differential equation and 
   asks you to find the coefficients of its power series solution. You plug the
   generic series in, group the x's, x2's, etc. and force the left side to
   match the right side...


### 5. Evaluating a Sum (Reverse-Engineering)

THE CONCEPT: Recognising a Taylor series after a specific number has been
   plugged into it.

THE INTUITION: This is a classic "stretch" question. The exam will give you a
   weird infinite sum of pure numbers, like $\sum_{n=0}^{\infty} \frac{(-1)^n \pi^{2n}}{(2n)!}$
   , and ask for its exact value.

   You are looking at a "fossil" of a function. You have to recognise the 
   skeletaln structure (the alternating signs), the even powers, the (2n)!
   factorial). That is the skeleton of cos(x). The only difference is that
   $x$ has been replaced by $pi$. Therefore, the entire ugly sum evaluates 
   simply to $cos(\pi) = -1$.

### Stage 1: Concepts and Intuition

1. Rolle's Theorem
   - DEFINITION:
      If $f : [a, b] \to \mathbf{R}$ is continuous on the closed interval [a, b]
      , differentialble on the open interval (a, b), and f(a) = f(b), then 
      there exists some point $c \in (a, b)$ s.t. $f'(c) = 0$.


2. Mean Value Theorem (MVT)
   - DEFINITION:
      If $f : [a, b] \to \mathbf{R}$ is continuous on the closed interval [a, b]
      , differentialble on the open interval (a, b), then there exists some 
      point $c \in (a, b)$ s.t.:

      $$f'(c) = \frac{f(b) - f(a)}{b - a}$$


3. THE BIG COROLLARY: $f'(x) = 0$ implies $f$ is constant
   - THE INTUITION: If your speedometer reads exactly zero for your entire
     journey, your position never changed. You stayed in the exact same spot.
   - THE EXAM LOGIC: ... prove by using the MVT... pick two arbitrary points,
     apply MVT to bridge the gap between them, and use the fact that the 
     derivative is zero to prove their heights must be identical.

     
       


## Subtopic 1: Critical points and Hessian Classification

### Stage 1: Concepts and Intuition
   ...

The Workflow
   1. Find First Derivatives: Calculate $\frac{\partial f}{\partial x}$ (often
      written as $f_x$) and $\frac{\partial f}{\partial y}$ (or $f_y$).
   2. SOLVE FOR CRITICAL POINTS: Set f_x = 0 AND f_y = 0 simultaneously. This
      gives you a system of equations. Solve it to find the coordinate pairs
      (x0, y0) where the surface is flat.
   3. COMPUTE SECOND DERIVATIVES: Calculate all three second-order partials:   
      - $f_{xx}$ (differentiate $f_x$ with respect to $x$)
      - $f_{yy}$ (differentiate $f_y$ with respect to $y$)
      - $f_{xy}$ (differentiate $f_x$ with respect to $y$. Note: $f_{xy}$ always    
        equals $f_{yx}$ for these continuous functions.)
   4. FORM THE HESSIAN DETERMINANT: ... Let $a = f_{xx}$, $b = f_{xy}$, and 
      $c = f_{yy}$, all evaluated at your specific critical point $(x_0, y_0)$.
      Calculate the determinant: $D = ac - b^2$. (Or, 
      $D = f_{xx}f_{yy} - (f_{xy})^2$).


THE CLASSIFICATION RULES:
   - D > 0 and a > 0: Local MINIMUM (A bowl shape. The surface curves upwards
     in all directions)
   - D > 0 and a < 0: Local MAXIMUM (A hill shape...)
   - D < 0: SADDLE POINT... (A mountain pass. It curves up in one direction and
     down in another.)
   - D = 0: INCONCLUSIVE. The second derivative test fails. The surface might be
     perfectly flat like a plateau, or have a strange inflection.
