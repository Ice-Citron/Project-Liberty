### 1. Definitions: The Exam View
   When an exam asks for bounds, it is testing if you know the difference
   between a boundary and a member of the set. 

   * UPPER BOUND: Any number that is greater than or equal to every element in
     the set. (If 5 is an upper bound, so is 6, 1000, and 1 000 000).
   * SUPREMUM (Least Upper Bound): The smallest possible upper bound. It sits
     tightly against the top edge of the set.
   * MAXIMUM: A supremum that actually belongs to the set. If the set doesn't
     physically contains its supremum, it has no maximum.
   * LOWER BOUND: Any number less than or equal to every element in the set.
   * Infimum (Greatest Lower Bound): The largest possible lower bound. It sits
     tightly against the bottom edge.
   * MINIMUM: An infimum that actually belongs to the set.

   (Crucial Exam Rule: A supremum/infimum always exists for a bounded set of
   real numbers, but a max/min does NOT always exist.)

### 2. NUMBER-LINE for Key Cases

CASE A: Closed Interval $[1, 5]$
```
      1                       5
<-----[=======================]----->
```
   - Supremum: 5
   - Maximum: 5 (because the square bracket `]` means 5 is included).
   - Infimum: 1
   - Minimum: 1


---
CASE B: Open Interval $(1, 5)$
```
      1                       5
<-----(=======================)----->
```
- SUPREMUM: 5           <-- I assume no minimum because 4.999999 < 4.9999999 < 4.999...99999
- MAXIMUM: None. (The supremum exists, but because of the round bracket..., 5 is
  not in the set, so the maximum is not attained).
- INFIMUM: 1
- MINIMUM: None.  


---
CASE C: Discrete Increasing Set (e.g., $1 - \frac{1}{n}$ for $n \in \mathbb{N}^+$)
```
      0    0.5  0.66 0.75     1
<-----[-----|----|----| - - - )----->
```
   - SUPREMUM: 1 (The numbers get closer and closer to 1, but never reach it. 
     Supremum exists, but is not attained).
   - MAXIMUM: None.
   - Infimum: 0
   - Minimum: 0
