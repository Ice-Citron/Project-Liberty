### 1. Definitions: The Exam View
   When an exam asks for bounds, it is testing if you know the difference
   between a boundary and a member of the set. 

   * UPPER BOUND: Any number that is greater than or equal to every element in
     the set. (If 5 is an upper bound, so is 6, 1000, and 1 000 000).
   * SUPREMUM (Least Upper Bound): The smallest possible upper bound. It sits
     tightly against the top edge of the set.
   * MAXIMUM: A supremum that actually belongs to the set. If the set doesn't
     physically contains its supremum, it has no maximum.
   * LOWER BOUND: Any number less than or equal to every element in the set.
   * Infimum (Greatest Lower Bound): The largest possible lower bound. It sits
     tightly against the bottom edge.
   * MINIMUM: An infimum that actually belongs to the set.

   (Crucial Exam Rule: A supremum/infimum always exists for a bounded set of
   real numbers, but a max/min does NOT always exist.)

### The Math Behind Rational Sequence Limits   
   When you are dealing with a limit where $n \rightarrow \infty$, you are 
   looking at a battle of growth rates. In a polynomial (like $3n^2 + 2n$), as
   $n$ gets massively huge, the highest power of $n$ completely dominates the
   value of the expression. The smaller power and constants become practically
   invisible.

   For example, if ... barely changes the overall size.

   The mathematical goal is to isolate those dominnat terms to see exactly where
   the sequence is heading, and we do this using one core, indisputable fact of
   caclculus:

   $$lim_{n \rightarrow \infty} (\frac{1}{n^k} = 0)$$
   (for any $k > 0$).

   By dividing the entire numerator and denominator by the highest power of $n$
   found in the denominator, we force all the "weaker" terms to turn into 
   fractions with $n$ on the bottom. As we take the limit to infinity, all those
   weaker fractions instantly crush down to 0, leaving behind only the pure
   ratio of the dominant terms.

### 1 & 2. Main Sequence Types & Tiny Templates
   For exam, you don't always need an $\epsilon - N$ proof unless explicitly
   asked. Spotting the limit instantly relies on these four templates:

   - RATIONAL SEQUENCES (Polynomial / Poynomial)
      * The Pattern: $a_n = \frac{3n^2 + 2n}{6n^2 + 7n + 1}$
      * The Template: Divide every term in the numerator and the denominator
        by the highest power of $n$ found in the denominator. Terms with 
        $1/n$ or $1/n^2$ vanish to 0, leaving only the leading coeffcients.

   - ALTERNATING / OSCILLATING-OVER-GROWING SEQUENCES
      * The Pattern: $a_n = \frac{\sin(n)}{n}$ or $a_n = \frac{(-1)^n}{n^2}$
      * The Template (Sandwich Theorem): Bound the oscillating part between
        -1 and 1. For example, $-1 \leq sin(n) \leq 1$. Then divide the whole
        inequality by the growing denominator. Since the outer boundaries both
        shrink to 0, the sequence "sandwiched" to 0.

   - Root / Rationalisaion Sequence
      * The Pattern: $a_n = \sqrt{n + 1} - \sqrt{n}$
      * The Template: Multiply the top and bottom by the "conjugate" (e.g., 
        $\sqrt{n + 1} + \sqrt{n}$). The numerator becomes a simple interger 
        (difference of two squares), and the growing terms move to the 
        denominator, making the limit obvious.     
   
   - Sequences Tending to Infinity (Growth Hierarchices)
      * The Pattern: $a_n = \frac{n^n}{n!}$
      * The Template: Know your grwoth speed hierarchy. As $n -> \infty$,
        $n^n > n! > c^n > n^c > ln(n)$. If the top grows faster, it diverges to
        $\infty$. If the bottom grows faster, it converges to 0.

-- The Sandwich Theorem (or Squeeze Theoem) states that if three functions
   satisfy $g(x) \leq f(x) \leq h(x)$ near a point $c$, and the limits of the 
   bounding functions are equal, $lim_{x->c} g(x) = lim_{x->c} h(x) = L$, then
   the limit of the middle function is also $L$. It is a fundamental tool in
   calculus for finding the limits by "sequeezing" an unknown function between
   two known ones, such as calculating $lim_{x->0} x^2 sin(1/x) = 0$.

-- Rational sequence limit refers to the value that a sequence of rational
   numbers (fractions) approaches as the number of terms ($n$) tends to infinity
   . If the terms of a sequence ($a_n$) get arbitrarily close to a real number
   $L$ as $n$ grows, the sequence converges to limit $L$, which can be rational
   or irrational. 

KEY ASPECTS OF RATIONAL SEQUENCE LIMITS
   - DEFINITION: A sequence $(a_n)$ has a limit $L$ if, for any $\epsilon > 0$ 
     (really small +ve number), there exists an integer $N$ s.t. 
     $|a_N - L| < \epsilon, \forall n > N$.
   - DENSITY OF RATIONALS: Every real number, including irrational numbers like
     $\pi$ or $\sqrt{2}$, can be expressed as a limit of a Cauchy Sequence of
     rational numbers.
   - RATIONAL FUNCTIONS OF $n$: If a sequence is formed by the two ratio of two
     polynomials, $a_n = \frac{P(n)}{Q(n)}$, the limit depends on the highest
     power of $n$.
      * If `degree(P) = degree(Q)`, the limit is the ratio of the leading 
        coefficients.
      * If `degree(P) < degree(Q)`, the limit is 0.
      * If `degree(P) > degree(Q)`, the limit diverges to $\pm\infty$.
   - PROPERTIES: Limits of rational sequences follow standard rules: the limit
     of a sum is the sum of limits, and the limit of a product is the product of
     limits.

IMPORTANT DISTINCTIONS
   - A sequence of rational numbers does not need to have a rational limit, it 
     can converge to an irrational number.
   - While every Cauchy sequence of rationals converges, a general sequence of
     rationals can diverge if it does not approach a single value.

   Here is the breakdown of bounded sequences and the core theorem: "EVERY
   CONVERGENT SEQUENCE IS BOUNDED."

### 1. The Intuitive Explanation
   If a sequence converges to a limit $L$, it means that eventually, the 
   sequence "settles down" and gets trapped arbitrarily close to $L$.

   If all terms from a certain point onward are trapped in a tiny neighbourhood 
   around $L$, they can't shoot off to infinity. What about the terms before
   it settles down? Well, there are only a finite number of them. A finite list
   of numbers cannot be infinite. Therefore, the entire sequence is trapped
   with some highest and lowest value.


### 2. The \eps = 1 Trick (The Proof)
   Examiners love this proof because it is elegant and uses this limit
   definition creatively.

   THE GOAL: Prove that if $a_n -> L$, there is some massive number $M$ s.t.
   |a_n| \leq M for all $n$.

   THE TRICK:
      1. Since we know $a_n -> L$, the definition of a limit works for any 
         eps > 0.
      2. So, let's just pick a convenient number: LET eps = 1.
      3. By definition, there exists some cutoff point $n$ s.t. for all n > N,
         $|a_n - L| < 1$. 
      4. This means that for all $n > N$, the sequence is trapped:
            L - 1 < a_n < L + 1
      5. So, the "tail" of the sequence is bounded by $|L| + 1$.
      6. What about the "head" of the sequence (a_1, a_2, ..., a_N)? There are
         only $N$ of these terms. Because $N$ is finite, we can just look at 
         them and pick the largest one. 
      7. THE FINAL BOUND: Let $M = max(|a_1|, |a_2|, \dots, |a_N|, |L| + 1)$.
         Because $M$ covers the biggest outlier at the start and the entire
         trapped tail at the end, |a_n| <= M for all $n$. The sequence is bounded.

### 5. Connection to Supremum / Infimum
   Why do we care about bounded sequences? Because of the AXIOM OF COMPLETENESS.

   If a sequence is bounded, it means the set of all its terms has an upper 
   bound and a lower bound.

   The Completeness Axiom guarantees that:
   - If it has an upper bound, it MUST have a Supremum (a least upper bound).
   - If it has a lower bound, it MUST have an Infimum (a greatest lower bound).

   So, saying a sequence is bounded is mathematically identical to saying
   "This sequence has a finite Supemum and Infimum." (And if you have those, 
   you are perfecrtly set up to use the Sandwich Theorem!).

### EXERCISE 2: The Contrapositive

SCENARIO: $a_n = n sin^2(n)$, which grows unboundedly.
ANSWER: The sequence DIVERGES.
RIGOROUS JUSTIFICATION: Let $P$ be "the sequence converges" and $Q$ be "the
   suquence is bounded."

   The core theorem states $P \Rightarrow Q$ (Every convergent sequence is bounded).

   Logically, a conditional statement is strictily equivalent to its 
   contrapositive: $\neg Q \Rightarrow \neg P$ (If a sequence is not bounded, 
   then it is not convergent).

   Because the amplitude of $n sin^2(n)$ grows as $n \rightarrow \infty$, for
   any real number $M > 0$, there exists an $n$ s.t. $|a_n| > M$. Therefore, the
   sequence is unbounded ($\neg Q$). By the contrapositive of the thoerem, the
   sequence diverges ($\neg P$).

   Then the middle sequence is forced to the same limit:

      $$\lim_{n \to \infty} b_n = L$$
