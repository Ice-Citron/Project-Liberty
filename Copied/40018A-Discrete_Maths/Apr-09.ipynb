### Block 2: Relations, Matrices, and Digraphs
   At its core, a RELATION is just a rule that connects items from one set to 
   items in another (or the same) set. If an item $x$ is related to an item $y$,
   we write it as a coordinate pair $<x, y> \in R$, or sometimes $xRy$.

   When a relation connects a set to itself (a binary relation on $A$), we can
   visualise it as a DIGRAPH (dots and arrows) or a MATRIX (a grid of 1s and 0s)

   ... exam will test your ability to spot four key properties in these different
   formats.


### VISUALISATION: The 4 Key Properties
   Property // Plain English // In a Digraph // In a Matrix 

   REFLEXIVE 
   - Everything points to itself    
   - Every node has a self-loop.
   - The main diagonal is entirely 1s.

   SYMMETRIC
   - Every one-way street is actually two-way.
   - Every arrow has a matching reverse arrow.
   - The matrix is mirrored across the main diagonal ($M^T = M$)

   ANTISYMMETRIC
   - No two-way streets allowed (except self-loops).
   - If $A \to B$, then $B \to A$
   - $B \not\to A$.If $M_{i,j} = 1$, then $M_{j,i} = 0$ (unless $i=j$).

   TRANSITIVE
   - "Friend of a friend is my friend."
   - If you can walk from $A \to B \to C$, there is a direct shortcut $A \to B$
   - If $M_{i,j}=1$ and $M_{j,k}=1$, then $M_{i,k}$ must be $1$.

   



- rows    == source
- columns == destination

Yes, exactly! 

In a relation matrix, the **Row** is the source ("from") and the **Column** is the destination ("to").

So if you are looking at row $i$ and column $j$:
* A **$1$** means there is an arrow from $i \to j$, which means the ordered pair $(i, j)$ is in the relation.
* A **$0$** means there is no arrow from $i \to j$.

**Quick check:** If row 2, column 1 has a `1`, that means $(2, 1)$ is in the relation (arrow goes from 2 down to 1).

## Block 3: Equivalence Relations, Equivalenece Classes & Partitions
   The core idea of this block is that certain relations behave exactly like an
   "equals" sign. They allow us to treat different mathematical objects as if
   they are the same, provided they share a specific defining property.


### 1. Equivalence Relations
   A binary relation $R$ on a set $A$ is an EQUIVALENCE RELATION if and only if
   it satisfies three strict properties:
      - REFLEXIVE: $\forall x \in A, x R x$ (Every element is equivalent to 
        itself)   
      - SYMMETRIC: $\forall x, y \in A, x R y \Rightarrow y R x$ (If $x$ is 
        equivalent to $y$, then $y$ is equivalent to $x$).
      - TRANSITIVE: $\forall x, y, z \in A, (x R y \land y R z) \Rightarrow x R z$
        (If x is equivalent to y, and y to z, then x is equivalent to z).  

   If a relation fails even one of these properties, it is not an equivalence
   relation.


### 2. EQUIVALENCE CLASSES
   When an equivalence relation $R$ is applied to a set $A$, it groups the 
   elements into subsets. The subset of all elements related to a specific
   element $a$ is called the ERQUIVALENCE

   THERE ARE TWO DIFFERENT SETS IN PLAY, AND THEY LIVE AT DIFFERENT "LEVELS". 
   The relation `~` itself is a set of ordered pairs -- that's the `<a, b>`

   `~ = {⟨1,1⟩, ⟨1,3⟩, ⟨3,1⟩, ⟨3,3⟩, ⟨2,2⟩, ⟨2,4⟩, ⟨4,2⟩, ⟨4,4⟩}`. So `~ ⊆ A × A`

   not a set of pairs. Look at the definition: `[a] = {x ∈ A : x ~ a}`. The thing
   inside the braces ranges over `x ∈ A` -- so the members of `[a]` are elements
   of `A`, full stop. For our example...
