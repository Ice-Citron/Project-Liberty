1. THE TRANSLATION KEY
   Whenever you see $\exists \in \mathbf{N} (p \times m = n)$, your brain should
   immediately cross it out and write $p | n$.

   Now look at the det definition again: 
   $$V_n = {p \in P | p | n}$$

   PLAIN ENGLISH TRANSLATION: "$V_n$ is just the bag of unique prime factors 
   that make up the number $n$"

   And the relations:
   $$$n \sqsubseteq m \iff V_n \subseteq V_m$$$

   PLAIN ENGLISH TRANSLATION:
      "Number $n$ is `smaller` than number $m$ if every prime ingredient of $n$
      is also a prime ingredient of $m$."


---
2. THE "EDGE CASES" VISUALISATION (Your Cheat Code)
   To solve partial order questions instantly, you don't look at normal numbers
   like 6 or 10. You test the weirdest mathematical edge cases. Write these outs
   on your scratch paper immediately:
      - Case: $n = 1$ (The Empty Case)
        What primes divide 1? None.
        $V_1 = \emptyset$
      - Case: $n = 0$ (The infinite Case)
        What primes divide 0? All of them, because p * 0 = 0 works for every 
        rime.
      - Case: $n = a prime$, like 7 (The Singleton Case)
        What primes divide 7? Only 7.
      - Case: $n = a composite$, like 12 (The Normal Case)
        What primes divide 12? 2 and 3.
