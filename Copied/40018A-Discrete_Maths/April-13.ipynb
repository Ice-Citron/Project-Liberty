   The INCLUSION FUNCTION ... is a specific type of function used in set theory
   to formally describe the relationship between a subset and its parent set.

### MICRO-EXPLANATION
   When we have a subset $V \subset W$, every element that lives in V also lives
   in W. The inclusion $\iota : V \to W$ is the map that take an element $v \in V$
   and sends it to the EXACT SAME ELEMENT in $W$.

   While it looks identical to the IDENTITY FUNCTION ($id_V(v) = v$), they are
   technically different because their CODOMAINS differ:
      * IDENTITY FUNCTION: Maps $V \to V$   
      * INCLUSION FUNCTION: maps $V \to W$
