   The INCLUSION FUNCTION ... is a specific type of function used in set theory
   to formally describe the relationship between a subset and its parent set.

### MICRO-EXPLANATION
   When we have a subset $V \subset W$, every element that lives in V also lives
   in W. The inclusion $\iota : V \to W$ is the map that take an element $v \in V$
   and sends it to the EXACT SAME ELEMENT in $W$.

   While it looks identical to the IDENTITY FUNCTION ($id_V(v) = v$), they are
   technically different because their CODOMAINS differ:
      * IDENTITY FUNCTION: Maps $V \to V$   
      * INCLUSION FUNCTION: maps $V \to W$

      

---

   This proof is a classic "mapping" argument. Instead of proving $P$ is 
   uncountable directly using a messy diagonal argument, we create a BIJECTION
   (a perfect one-to-one correspondence) between our set $P$ and a set we
   already know is uncountable: the power set of the natural numbers, 
   $\wp(\mathbb{N})$.

---

1. THE SETUP: What is in the set $P$?
   A partial function $f : \mathbf{N} \to {1}$ is like a machine that takes a 
   natural number as input. For any number $n$, the machine has only two choices:

   - It outputs the number 1. 
   - It is UNDEFINED (crashes/gives no output).

   Because the output is always 1 if it exists, the only thing that makes one
   function $f$ different from another function $g$ is WHERE they are defined.


2. IDENTIFYING THE FUNCTIONS WITH SETS
   We define the "domain of definition" $D_f$ as the collection of all inputs 
   that actually give an output.

   - EXAMPLE: If $f$ is defined only on even numbers, $D_f = {0, 2, 4, ...}$
   - EXAMPLE: If ... prime numbers..., $D_g = {2, 3, 5, 7 ...}$

   Every possible subset of $\mathbf{N}$ corresponds to exactly one partial 
   function in $P$. If you give me a subset like ${1, 5, 10}$, I can build a 
   function that is defined only at 1, 5, and 10, and outputs "1" at each.


3. THE BIJECTION $\phi : P \to \wp(\mathbf{N})$
   We define a mapping $\phi$ that takes a function and "extracts" its domain
   $D_f$.
      1. Injective: If two functions have the same domain of definition, they
         must be the same function (since they both output 1 everywehre they are
         defined).
      2. SURJECTIVE: Every subset of $\mathbf{N}$ is the domain of some partial
         function in $P$.

   Since $\phi$ is a bijection, the sets $P$ and $\wp(\mathbf{N})$ have the 
   same CARDINALITY (size). In math notation, we say $P \approx \wp(N)$
