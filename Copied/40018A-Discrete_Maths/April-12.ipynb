## Cardinality: The Size of Infinity
   Cardinality is how we measure the "size" of a set. For finite sets, it's
   jsut the count of elements. For infinite sets, we say two sets have the SAME
   CARDINIALITY ($A \approx B$) if and only if there exists a BIJECTION between
   them.


1. The Cardinality Hierarchy
   Not all infinities are created equal. We categorise them into two main groups:

   - COUNTABLE ($\mathbf{N_0}$): Sets that are finite or can be put into a 
     1-to-1 correspondence with the natural numbers $\mathbf{N}$. You can "list"
     them as a0, a1, a2, ...
       - EXAMPLES: $\mathbf{N}$, Z (integers), Q (rationals), and even $\mathbf{N}^2$

   - UNCOUNTABLE: Sets that are "larger" than $\mathbf{N}$. There is no way to
     list them without missing elements. 
       - EXAMPLES: $\mathbf{R}$, $\wp(\mathbb{N})$ (the power set of naturals).


2. CANTOR'S DIAGONAL ARGUMENT
   How do we prove a set is uncountable? We use DIAGONALISATION. If you try to
   list every subset of $\mathbf{N}$, I can always construct a new set $W$ that
   is different from every set in your list by looking at the "diagonal" and
   flipping the membership.f

### Micro-explanation: The Formal Meaning of Countability
   ... To formally prove a set $A$ is COUNTABLE, you have a choice of three 
   equivalent definitions. The one you pick depends on whether you are trying
   to prove two sets are exactly the same size, or if you are just proving one
   set fits inside another.

   1. THE BIJECTION DEFINITION (Exact Match): A set $A$ is countably infinite if
      there exists a bijection $f : A \to \mathbf{N}$. (Every element in $A$
      gets exactly one unique tag from $\mathbf{N}$, and no tags are leftover).

   2. THE INJECTION DEFINITION (Fits Inside): A set $A$ is countable if there
      exists an injection $f: A \to \mathbf{N}$. (Every element in $A$ gets a 
      unique tag from $\mathbf{N}$, but you might have tags left over. This 
      covers both finite and countably infinite sets).

   3. THE SURJECTION DEFINITION (Can be covered): A non-empty set $A$ is countable
      if there exists a surjection $f : N \to A$. (You can "list" the elements
      of $A$ using $\mathbf{N}$, even if you repeat some elements)...
