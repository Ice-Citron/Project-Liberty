   do surjectivity andi injectivity not hold if a function isn't total?

---
   YES, generally speaking, if a function is not total (i.e. it's a partial
   function), surjectivity and injectivity do not hold in the strict sense
   applied to standard functions.

   While partial functions can be considered injective or surjective within
   their specific subdomain or image, they generally fail the standard 
   definitions because those definitions assume every eleement in the domain maps
   to something, an d every element in the codomain is potentially hit.
