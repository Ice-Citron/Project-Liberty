1. HIGH-LEVEL SUMMARY: THE "WHY" AND "WHAT"

The lecture focuses on an algorithmic method to solve systems of linear 
equations: GAUSSIAN ELIMINATION.

From an AI/ML perspectie, think of this as data processing or dimensionality
reduction. You have a messy system of constraints (equations), and you want to
transform them into a "canonical form" (Row Echelon Form) where the solution
becomes obvious, without changing the fundamental truth (the solution set) of
the system.


THE LECTURE COVERS:
    1. AUGMENTED MATRICES: Compressing the system $Ax = b$ into $[A|b]$.
    2. ROW OPERATIONS: The legal moves you can make to simplify the matrix.
    3. REF vs RREF: The target states of the matrix.
    4. INVERTING MATRICES: Using this same method to find $A^{-1}$

---


2. GEOMETRIC INTUITION (2D vs. 3D)

.... you know that a linear equation in 2D is a LINE, and in 3D it is a PLANE.


THE GOAL - Finding the intersection.
- UNIQUE SOLUTION: The single point $(x, y, z)$ where all planes intersect.
- INFINITE SOLUTIONS (FREE VARIABLES): The planes intersect along a line or lie
  on top of each other.
- NO SOLUTION: The planes form a triangular prism (parallel interactions) and 
  never all meet at once.
  
  
THE GEOMETRIC MEANING OF "ROW OPERATIONS":
When we perform row operations (add row 1 to row 2, scale row 2), we are 
CHANGING THE NORMAL VECTORS (ANGLES) of the individual planes, but we are 
LOCKING THE INTERSECTION POINT IN PLACE.

Imagine three sheets of paper intersecting at a single dot. Gaussian elimination
twists and rotate these sheests until:
1. SHEET 1 is flat on the floor: ($z$ = constant).
2. SHEET 2 is a wall aligned with the x-axis ($y + z$ = constant).
3. SHEET 3 is a wall aligned with the y-axis. ($x + y + z$ = constant).

This makes reading the coordinates of the intersection trivial.

3. Trace of the Lecture ExampleLet's trace the specific example used in the lecture, bridging the algebra with the geometry.The System:$$\left[\begin{array}{rrr|r}
2 & 1 & 1 & 3 \\
4 & -6 & 0 & -2 \\

\end{array}\right]$$Step 1: Elimination (Reaching Row Echelon Form)We want zeros below the main diagonal (a triangular matrix).Op 1: Swap $R1$ and $R2$. GEOMETRIC INTERPRETATION OF THE "ZERO ROW": The row of zeros $[0 \ 0 \ 0 \ | \ 0]$
means one of your planes was REDUNDANT. It didn't add a new constraint. Two of
your planes effectively constrained the solution to a line, and the third plane 
just happened to pass through that same line.



SETUP 2: THE SOLUTION (PARAMETRIC)

Because we have a zero row, we have a FREE VARIABLE.
- PIVOT VARIABLES: $x$ and $y$ (columns with the leading 1s).
- FREE VARIABLE: $z$ (column with no pivot).

The lecturer sets $z$ as the free variable. $$x - z = -2 \implies x = z - 2$$ $$y + 2z = 3 \implies y = -2z + 3$$The Solution Set (Vector Form):$$\begin{bmatrix} x \\ y \\ z \end{bmatrix} = z \begin{bmatrix} 1 \\ -2 \\ 1 \end{bmatrix} + \begin{bmatrix} -2 \\ 3 \\ 0 \end{bmatrix}$$


In AI/ML code, you might see this output as 
`output = lambda * direction_vector + bias_+vector`.

GEOMTRIC MEANING: The solution is not a point, it is a line in 3D space.

- $\begin{bmatrix} -2 \\ 3 \\ 0 \end{bmatrix}$ is a specific point on that line.
- $\begin{bmatrix} 1 \\ -2 \\ 1 \end{bmatrix}$ is the direction vector of the line.
- $z$ (or $\lambda$) is how far you walk along that line.

4. MATRIX INVERSION VIA GAUSSIAN ELIMINATION

The lecture concludes with an application: Finding $A^{-1}$.

THE METHOD: Set up an augmented matrix $[A | I]$. Use row operations until the
left side becomes $I$. Th right side will magically become $A^{-1}$.


$$[ A \ | \ I ] \xrightarrow{\text{Gauss-Jordan}} [ I \ | \ A^{-1} ]$$


GEOMETRIC INTUITION (TRANSFORMATION VIEW): Since you do AI, think of Matrix A as
a transformation (e.g., a shear or rotation) of space.

- $A$ moves the basis vectors $\hat{i}$ and $\hat{j}$ to new locations.
- The Identity matrix $I$ represents the basis vectors staying where they are.
- Row operations are "undoing" the transformation $A$ step-by-step to get back
  to $I$.
- By applying those same "undoing" steps to $I$ (on RHS), you are recording the 
  "recipe" to reverse $A$. That recipe is $A^{-1}$.


THE "SINGULAR" CASE: The lecturer notes: "If you are unable to reduce matrix 
down to... identity... A is not invertible."

- GEOMETRICALLY: This means transformation $A$ squished 3D space into 2D (a
  plane) or 1D (a line). You cannot "unsquish" a line back into 3D space because
  you've lost information. The determinant is 0.

1. ARE PIVOTS ALWAYS ON THE DIAGNOL?

NO. This is a common misconception.

Pivots are just the FIRST NON-ZERO NUMBER in a row.
- In a PERFECT scenario (square, invertible matrix), yes, they land on the 
  diagnol like a neat set of stairs.
- But in "messy" AI/ML data (rectangular matrices or dependent systems), pivots
  "slide" to the right.


EXAMPLE: A "WIDE" MATRIX (MORE VARIABLES THAN EQUATIONS) Imagine a system with
variables $x, y, z, w$.

$$\begin{bmatrix}
\mathbf{1} & 2 & 3 & 4 \\
0 & 0 & \mathbf{1} & 5 \\
0 & 0 & 0 & 0
\end{bmatrix}$$

- PIVOT 1 is at $(1, 1)$.
- PIVOT 2 is at $(2, 3)$. It skipped the diagnol spot $(2, 2)$!
- The variable $y$ (column 2) has NO PIVOT. Tht means $y$ is a FREE VARIABLE.


GEOMETRIC INTUITION: The pivot positions tells you which axes "lock" the 
solution, and the missing pivots tell you which deimsnions the solution "slides"
along (infinite solutions).

---

2. WON'T RREF BASICALLY BE THE SAME AS THE IDENTITY MATRIX ($I$)?

ONLY IF THE MATRIX IS "PERFECT" (SQUARE AND INVERTIBLE).


If you have a square matrix with a unique solution, yes, RREF becomes $I$:

$$\text{RREF} = \begin{bmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{bmatrix}$$

BUT FOR AI/ML, WE CARE ABNOUT THE "BROKEN" CASES:
If the matrix is singular (determinant is 0) or rectangular, RREF looks like 
Identity WITH EXTRA STUFF ATTACHED or ROWS OF ZEROS at the bottom.


EXAMPLE: THE "SINGULAR" CASE ($RREF \neq I$)

$$\begin{bmatrix}
1 & 0 & \mathbf{2} \\
0 & 1 & \mathbf{-1} \\
0 & 0 & 0
\end{bmatrix}$$


- This is in RREF (Leading 1s, zeros above/below them).
- It is NOT the identity matrix.
- The non-zero parts in the third column $(2, -1)$ are actually the "recipe" for
  your free variables (the Null space).

---

REF (ROW ECHELON FORM)
WHAT IT LOOKS LIKE: Zeros below diagnol. Pivots can be any number (e.g., 5).
USE CASE 1: SOLVING: FASTER. Computers use this. You just "back-substitute" to
  get the answer.
IMPERIAL EXAM TIP: Use this if the question asks for "RANK" or "SOLVE THE 
  SYSTEM."


RREF (Reduced ROW ECHEDLON FORM)
WHAT IT LOOKS LIKE: Zeros above and below diagnol. Pivots must be 1.
USE CASE 1: SOLVING: SLOWER. Humans use this to read the answer directly
  without math.
IMPERIAL EXAM TIP: CRITICAL: Use this is the question asks for the "BASIS OF THE
  NULL SPACE" or "KERNEL".



WHY RREF matters for EXAMS: In exam questions asking for the "GENERAL SOLUTION"
or "BASIS FOR THE KERNEL," the RREF matrix literally SHOWS YOU THE VECTORS you
need to write down.
  - REF: You have to do algebra to find the relation between $x, y, z$.
  - RREF: You can literally "read" the vector off the matrix.

    - Look at the "Singular" eaxmple above: The column is $\begin{bmatrix} 2 \\ -1 \\ 0 \end{bmatrix}$
      . The Null Space basis is just the negative of that:  $\begin{bmatrix} -2 \\ 1 \\ 1 \end{bmatrix}$.



IMPERIAL EXAM "GOTCHA": If a question asks: "For what values of $k$ does this 
system have no solution?"
  - You only need REF.
  - Stop as soon as you see a row like $[0 \ 0 \ 0 \ | \ 5]$. You don't need to
    go all the way to RREF to prove it's broken.

... exams ... love questions where you cannot blindly reduce to the Identity
matrix $I$ because of a hidden parameter (usually $k$, $a$, or $\lambda$).


QUESTION: Consider the system of linear equations:

$$\begin{aligned}
x + y + z &= 1 \\
x + 2y + 3z &= 2 \\
x + 2y + kz &= 4
\end{aligned}$$


1. For what value of $k$ does the system have NO SOLUTION?
2. For what value of $k$ does the system have INFINITELY MANY SOLUTIONS?
3. In the case of infinitely many solutions, find the GENERAL SOLUTIONS in 
   vector form.

---

THE SOLUTION (and the Trap)


STEP 1: GAUSSIAN ELIMINATION TO REF (NOT RREF YET!)
Form the augmented matrix and subtract Row 1 from the others to get zeros below
the first pivot.

$$\left[\begin{array}{rrr|r}
1 & 1 & 1 & 1 \\
1 & 2 & 3 & 2 \\
1 & 2 & k & 4
\end{array}\right]
\xrightarrow{R2-R1, R3-R1}
\left[\begin{array}{rrr|r}
1 & 1 & 1 & 1 \\
0 & 1 & 2 & 1 \\
0 & 1 & k-1 & 3
\end{array}\right]$$

Now, eliminate the entry below the second pivot (the 1 in Row 3) by doing 
$R3 - R2$.

$$\left[\begin{array}{rrr|r}
1 & 1 & 1 & 1 \\
0 & 1 & 2 & 1 \\
0 & 0 & k-3 & 2
\end{array}\right]$$



THE TRAP: TRYING TO FORCE IDENTITY

A student relying on a calculator or rote memorization tries to turn the diagnol
into 1s immediately.
- They try to divide R3 by $(k -3)$.
- FATAL ERROR: If $k = 3$, you just divided by zero. You must stop at REF to
  analyze the parameter.


STEP 2: ANALYZE THE PIVOT
Look at the last row: $0x + 0y + (k - 3)z = 2$.

- CASE A (NO SOLUTION): If $k = 3$, the equation becomes $0 = 2$. This is 
  impossible.
    - ANSWER TO Q1: $k = 3$.
- CASE B (UNIQUE SOLUTION): If $k /neq 3$, we can divide by $(k - 3)$ to get a 
  pivot of 1. The system is solvable and has a unique point solution.
- WAIT... WHERE IS THE "INFINITE SOLUTIONS" CASE? Imperial examiners often set
  questions where one of the standard cases DOESN'T EXIST to test your 
  confidence.     
    - For this specific system, there is NO VALUE OF K that gives infinitely
      many solutions.
    - Why? For infinite solutions, we would need a $0 = 0$ row. Here, we have
      $0 = 2$ (impossible) or $non-zero = 2$ (unique).
    - ANSWER to Q2: No such value exists.    

---

A "MODIFIED TRAP" (WHERE INFINTIE SOLUTIONS DO EXIST)

Let's change the last number in the original system from 4 to 2. Now the final
row becomes:

$$\left[\begin{array}{rrr|r}
0 & 0 & k-3 & 0
\end{array}\right]$$

Now, if $k = 3$, we get $0 = 0$. This is the INFINITE SOLUTIONS case.



FINDING THE GENERAL SOLUTION (THE RREF vs. REF decision):
Set $k = 3$. The matrix is:

$$\left[\begin{array}{rrr|r}
1 & 1 & 1 & 1 \\
0 & 1 & 2 & 1 \\
0 & 0 & 0 & 0
\end{array}\right]$$

You could go to full RREF (clearing the 1 above the second pivot), but you don't
have to. You can just back-substitute from here.
- VARIABLE $z$ has no pivot $\rightarrow$ FREE VAIRABLE ($\lambda$).
- Row 2: $y + 2\lambda = 1 \Longrightarrow y = 1 - 2\lambda$
- Row 1: $x + (1 - 2\lambda) + \lambda = 1 \Longrightarrow x = \lambda$ 



GENERAL SOLUTION:

$$\begin{bmatrix} x \\ y \\ z \end{bmatrix} = \begin{bmatrix} 0 \\ 1 \\ 0 \end{bmatrix} + \lambda \begin{bmatrix} 1 \\ -2 \\ 1 \end{bmatrix}$$

WHY RREF != IDENTITY HERE: Notice our final matrix had a row of zeros. The RREF
would look like this:

$$\begin{bmatrix}
1 & 0 & -1 & 0 \\
0 & 1 & 2 & 0 \\
0 & 0 & 0 & 0
\end{bmatrix}$$


It is clean, but it is NOT the Identity matrix.

THE CONFLICT
- THE "ROW PICTURE" (CLASSROOM/GAUSSIAN): You are intersecting planes. $0 = 2$
  means two planes are parallel and never meet.
- THE "COLUMN PICTURE" (3B1B/AI): You are transforming space. Matrix $A$ moves
  the basis vectors $\hat{i}$, $\hat{j}$, $\hat{k}$. The vector $x$ is the 
  "recipe" of how much of each column to use to reach the target $b$.

Let's stick strictly to your preferred TRANSFORMATION VIEW to explain what 
Gaussian Elimination is actually "detecting" geometrically.

---

1. VISUALISING: "NO SOLUTION" (0 = 2)

THE SETUP: You have a transformation $A$ (the matrix) and a target point $b$.       <--target relies on new basis vectors!!
You want to find an input vector $x$ that lands on $b$ after the transformation.

THE GEOMETRIC EVENT: When Gaussian Elimination gives you a row of zeroes on the
left (e.g. $[0 \ 0 \ 0]$), it means the transformation $A$ has SQUISHED space.

- It took full 3D space and flattened it into a 2D sheet (plane) or a 1D line.
- You have lost a dimension.


THE "0 = 2" MEANING:
The left side (0) says: "We are stuck on this flat 2D sheet." The right side (2)
says: "The target $b$ is floating 2 units ABOVE the sheet."

- CAN YOU REACH $b$? No. No matter how you combine the basis vectors (which are
  are all trapped on that sheet), you can never reach the point floating above
  it. 
- CONCLUSION: The target $b$ is OUTSIDE THE COLUMN SPACE (Span) of $A$.



2. VISUALISING "INFINITE SOLUTIONS" (0 = 0) 

THE GEOMETRIC EVENT: Same as above. The matrix $A$ has still SQUISHED 3D space
into a 2D sheet. You have still lost a dimension.

THE "0 = 0" MEANING: The left side (0) says: "We are stuck on this flat 2D 
sheet." The right side (0) says: "The target $b$ is also flattened onto this
sheet."
    - CAN YOU REACH $b$? Yes! It's right there on the sheet.
    - WHY INFINITE SOLUTIONS? Imagine squishing a 3D block of jelly flat onto a
      table.
        - Many different atoms that were originally "above" each other in the
          vertical line all get squished onto the EXACT SAME SPOT on the table.
        - If $b$ is that spot on the table, then the entire vertical line of
          original atoms maps to it.
        - That "vertical line" of original inputs is your line of solutions.

3. HOW GAUSSIAN ELIMINATION "SEES" THIS

You mentioned you haven't learned to visualize augmented matrices yet. Here is
the trick:

DON'T VISUALIZE THE AUGMENTED MATRIX AS ONE GIANT MATRIX. Visualise it as a 
MACHINE (Left Side) and a TRACKER (Right Side).

$$\left[\begin{array}{ccc|c}
\text{Transformation } A & & & \text{Target } b
\end{array}\right]$$


When you apply row operations, you are changing the basis vectors.
- STEP 1: You subtract rows to eliminate variables. Geometrically, you are 
  aligning the basis vectors with the axes.
- THE MOMENT OF TRUTH: You arrive at the bottom row.
    - If the Left Side becomes $[0 \ 0 \ 0]$, it means your basis vectors have
      collapsed. You tried to point $\hat{k}$ in a new direction, but it fell
      onto the $\hat{i} - \hat{j}$ plane. You have zero "height" left.
    - Now look at the Right Side (the target tracker).
    - If the tracker says 2 (height = 2), but your basis vectors have 0 height
      ... ERROR. You cannot build a height of 2 using bricks of height 0.



SUMMARY TABLE: THE ROSETTA STONE

- ALGEBRAIC RESULT: Unique Solution
- ROW PICTURE (PLANES): 3 planes meet at 1 dot.
- COLUMN/TRANSFORMATION PICTURE: The Basis Vectors span all of 3D space. You can
  reach any target $b$ exaclty once.

- ALGEBRAIC RESULT: $0 = 2$ (NO SOLUTION)
- ROW PICTURE (PLANES): Planes form a tent/prism (parallel).
- COLUMN/TRANSFORMATION PICTURE: Transformation squished space into a 2D sheet.
  Target $b$ is OFF THE SHEET.

- ALGEBRAIC RESULT: $0 = 0$ (INF. SOLUTIONS)
- ROW PICTURE (PLANES): Planes intersect along a line.
- COLUMN/TRANSFORMATION PICTURE: Transformation squished space into a 2D sheet.
  Target $b$ is ON THE SHEET. You lost info during the squish, so you can't tell 
  which original input $x$ was used (inverse doesn't exist).

Your intuition about TARGET $b$ is correct. Your intuition about VECTOR $x$ is
slightly backwards.

Here is the precise way to view $Ax = b$:



1. THE TARGET $b$ (The Destination)

... The vector $b$ (e.g.,  $\begin{bmatrix} 1 \\ 2 \\ 3 \end{bmatrix}$) is a
fixed point in space.
- Its coordinates are measured using the STANDARD BASIS ($\hat{i}, \hat{j}, \hat{k}$).
- It sits there, unmoving, waiting for you to reach it.

4. Trace: The "Free Variable" Example
Let's trace the specific example he did on the board to solidify the "Free 
Variable" method, which is critical for exams. 


THE RREF MATRIX:

$$\left[\begin{array}{rrrr|r}
1 & 2 & 0 & 1/14 & -1/2 \\
0 & 0 & 1 & 2/7 & 1 \\
0 & 0 & 0 & 0 & 0
\end{array}\right]$$


STEP 1: IDENTIFY PIVOTS & FREE VARIABLES

- PIVOTS: Columns 1 ($x_1$) and 3 ($x_3$) have the leading 1s. --> DEPENDENT VARIABLES
- NO PIVOTS: Columns 2 ($x_2$) and 4 ($x_4$) are empty. --> FREE VARIABLES



STEP 2: WRITE EQUATIONS (DEPENDENT = Function of Free)

$$x_1 = -\frac{1}{2} - 2x_2 - \frac{1}{14}x_4$$
$$x_3 = 1 - \frac{2}{7}x_4$$



STEP 3: VECTORISE (THE EXAM STANDARD) 
You need to split the solution into three parts: The CONSTANT part, the $x_2$
PART, and the $x_4$ PART.

$$\mathbf{x} = \begin{bmatrix} -1/2 \\ 0 \\ 1 \\ 0 \end{bmatrix} + x_2 \begin{bmatrix} -2 \\ 1 \\ 0 \\ 0 \end{bmatrix} + x_4 \begin{bmatrix} -1/14 \\ 0 \\ -2/7 \\ 1 \end{bmatrix}$$

- NOTE: See how for the free variables ($x_2$, $x_4$), we put a 1 in their own
  slot (row 2 for $x_2$, row 4 for $x_4$), and 0 in the other free variable's
  slot. This is the standard "recipe" to ensure independence.

THE BOTTOM LINE (THE EXAM ANSWER)
The time compelxity of GAUSSIAN ELIMINATION is $O(n^3)$ (Cubic Time).

If you want to be precise... the number of operations is approximately $\frac{2}{3}n^3$


-------|--<>
THE STEP-BY-STEP "RUN THROUGH"

To see why it is cubic, we have to count the operations (additions, 
multiplications, divisions) required to turn a square $n \times n$ matrix into
Row Echelon Form (zeros below the diagnol).

Imagine you are the computer processing a massive $n \times n$ matrix.


STEP 1: HANDLING THE FIRST COLUMN
- You need to put zeros everywhere below the first pivot (1, 1).
- There are roughly $n$ rows below it.
- For EACH of those rows, you have to modify EVERY entry in the row (roughly $n$
  columns) to subtract the scaled first row.
- COST: $n rows \times n columns \eq n^2$ operations.


STEP 2: HANDLING THE SECOND COLUMN
- Now you move to pivot (2, 2). You ignore the first row and first column (they
  are "frozen").
- You now have a sub-matrix of size $(n - 1) \times (n - 1)$

... 

STEP 3: HANDLING THE THIRD COLUMN
- ... $(n - 2) \times (n - 2)$...
- COST: $(n - 2)^2$ operations.

...

$$\text{Total work} \approx n^2 + (n - 1)^2 + (n - 2)^2 + ... + 1^2$$

^^^ seems to translate to $\frac{2}{3}n^3$ complexity...

BACK SUBSTITUTION (THE EASY PART)

Once the matrix is in Echelon form (the triangle of zeros is done), finding the
values of $x, y, z$ is cheap.
- You start at the bottom, solve for $z$ (1 op).
- Plug $z$ into the row above (2 ops).
- Plug $z, y$ into the row above (3 ops).
- ...
- TOTAL: Sum of $1 + 2 + ... + n \approx \frac{1}{2}n^2$.


Since $n^3$ dominates $n^2$, we ignore the back substitution part in the final
Big-O analysis.

- A matrix is SINGULAR if ITS DETERMINANT IS ZERO, meaning it's a square matrix
  that is NOT INVERTIBLE, has LINEARLY DEPENDENT ROWS/COLUMNS, and leads to
  non-unique (no or infinite) solutions for a system of linear equations 
  ($Ax = b$). In contrast, a NON-SINGULAR matrix has non-zero determinant, is 
  invertible, and yields a unique solution.

- A rectangular matrix is a grid of numbers or symbols (elements) arranged in 
  rows and columns where the number of rows ($m$) is NOT EQUAL to the number of
  columns ($n$), denoted as $m x n$ where $m \neq n$. This contrasts with a 
  square matrix, where rows and columns are equal, with the name "rectangular"
  stemming from its shape when elements are displayed.

- REF (Row Echelon Form) and RREF (Reduced Row Echelon Form) are matrix forms
  used in linear algebra, with RREF being a stricter version of REF; REF 
  requires zero rows at the bottom, pivots (first non-zero entries) 
  stair-stepping down and right, and zeros BELOW pivots, while RREF adds that
  pivots must be `1` and all entries above pivots must also be zero, making
  solutions direct, whereas REF needs back-substitution. RREF is unique for a
  given matrix, computationally costlier, and provides solutions instantly,
  whereas REF isn't unique...

- Back Substitution is the process of solving a system of linear equation after
  its augmented matrix has been transformed into Row Echelon Form (REF) or
  Reduced Row Echelon Form (RREF) by starting from the last equation and
  working upwards, substituting the found variable's value into the preceiding 
  equation to solve for the next variable, until all variables are determined,
  revealing the system's solution (unique, infinite or none).

        WHEN IS IT USED?
        - GAUSSIAN ELIMINATION: Back substitution is the second part of Gaussian
          Elimination, following the forward elimination that puts the matrix 
          into REF.
        - SYSTEM STRUCTURE: It translates the structured matrix form into 
          concrete numerical solutions, identifying if a system has one solution
          (unique), infinite solutions (free variables), or no solution.
