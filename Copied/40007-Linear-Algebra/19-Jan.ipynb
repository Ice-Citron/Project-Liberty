2. THE "WHAT": SYSTEMS OF LINEAR EQUATIONS

The core mathematical object of the course is the SYSTEM OF LINEAR EQUATIONS.

- THE NOTATION:
    - Instead of writing out long equations $(3x + 2y + ... = 5)$, we compact
      them into Matrix form: $Ax = b$.
    - $A$: The Coefficient Matrix (the numbers in front of your variables).
    - $x$: The Vector of variables (what you are trying to solve for).
    - $b$: The constant vector (the results).    


- HOMOGENEOUS vs. NON-HOMOGENEOUS:
    - HOMOGENEOUS: If $b = 0$ (all equations equal zero), the system always has
      at least one solution (the zero vector, where all $x$ are 0).
    - NON-HOMOGENOUS: If $b != 0$, the solution does not pass through the 
      origin.

- GEOMETRIC INTERPRETATION: 
    - 2D: Each equation is a line. The solution is where they intersect.
    - 3D: Each equation is a flat plane. The solution is where the planes 
      intersect.
    - HIGH DIMENSIONS: The solution is a "Hyperplane."

- TYPES OF SOLUTIONS:
    1. UNIQUE SOLUTION: The lines/planes cross at exactly one point.
    2. NO SOLUTION: The lines/planes are parallel and never touch (e.g., $x = 1$
       and $x = 2$).
    3. INFINITELY MANY SOLUTIONS: The lines/planes lie on top of each other.
