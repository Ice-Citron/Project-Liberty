WHAT IS A MUX?
   A MULTIPLEXER (`mux`) is just a switch with two inputs and one output. A
   control signal decides which input gets passed through. Think of it like a
   railway junction -- the signal lever determines which track the train comes
   from.

   ...

```bash
   - if C = 0 then P
   - if C = 1 then Q
```

   This tells you:
   - C = 0 --> the output comes from P (the top input)
   - C = 1 --> the output comes from Q (the bottom input)



Now Look at M_1 Specifically
   M_1 is the mux that decides WHAT ADDRESS GOES TO MEMORY. Looking at the 
   diagram, two things feed into M_1:
      - UX connects from the top/upper side --> this is the P input
      - PC connects from the bottom/lower side --> this is the Q input
   
   And the control wire is labelled C_1.

   So applying the legend: 


C_1 value // Which input selected // Meaning
C_1 = 0 // P = UX(IR_6-15)  // Memory address comes from the instruction's address field
C_1 = 1 // Q = PC           // Memory address comes from the program counter

   WHY THIS MATTERS FOR THE LOAD INSTRUCTION
      CYCLE 1 -- we're fetching the instruction itself. The instruction lives
      at whatever address PC is pointing to. So we need the memory address to
      be PC. That means we select the Q input --> C_1 = 1.

      CYCLE 2 -- we already have the instruction in IR. Now we want to read the
      data that the instruction refers to. The data address is encoded in bits
      6-15 of IR, extended by UX.

In computer architecture, the INSTRUCTION REGISTER (IR) is a specialised,
high-speed storage location within the CPU's control unit that temporarily
holds the machine code instruction currently being decoded or executed...

- UX == sign-extension or unit-extension... Its primary purpose is to take a
  smaller number (an immediate value) and expand it to a larger bit-width to
  match the rest of the architecture.

THE ROLE OF UX (Unit Extension)
   The UX unit is responsible for Sign Extension or Unit Extension.

---
In your 16-bit P1 Processor: The address segment is only 10 bits ($IR_{6-15}$). If you want to access address 321 (which is 0101000001 in binary), the UX unit pads it with 6 zeros to turn it into a full 16-bit address: 000000 0101000001.

In your 32-bit RISC-V Module: If you have a 12-bit immediate like 0x321, the hardware uses sign-extension to turn it into a 32-bit value: 0x00000321.
