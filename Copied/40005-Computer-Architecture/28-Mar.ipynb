   The INSTRUCTION REGISTER (IR) and INSTRUCTION MEMORY are two completely
   different things and the IR only appears once you move from a single-cycle 
   design to a MULTI-CYCLE DESIGN. Instruction Memory is the "library" where all 
   your code is stored (RAM), but the Instruction Register is a small "buffer"
   inside the CPU that holds exacly ONE instruction at a time while the CPU 
   works on it. In a multi-cycle datapath, it takes several clock cycles to
   finish one command; the IR is necessary to "freeze" the instruction bits so
   the Control Unit doesn't forget what it's doing while the Program Counter (PC)
   potenially moves on or the memory is used for other data.

   The reason you might not have seen it before is that in a SINGLE-CYCLE CPU,
   the IR isn't needed -- the instruction stays "live" on the memory wires for 
   the entire duratioj of the clock. In the diagram you shared, notice the 
   little triangle (clock symbol) on the IR. This means that during the FETCH
   stage, the 32 bits from memory are "latched" into this register, For the 
   next 3 to 5 cycles, the IR provides the `opcode`, `rs1`, `rs2`, and `imm`
   fields to the rest of the hardware, even if the meomory is later busy reading
   actual data (like in a `lw` instruction).

   Regarding the COMPARATOR and the "br" arrow, your intuition about the PC
   is close! That arrow is not an error; it is the BRANCH signal. The comparator
   looks at the register values (`rs1` and `rs2`) and determines if they are
   equal or if one is is less than the other. The "br" output is a 1-bit
   signal that goes back up to the CONTROL UNIO, If you are executing a `beq`
   (Branch if Equal) and the comparator says "Yes, they are equal," the Control
   Unit sees that "br" signal and then flips the `PCSrc` MUX to make the CPU
   jump to a new address. Without that "br" wire, the Control Unit would have
   no way of knowing if the `if` statement condition was actually met!
