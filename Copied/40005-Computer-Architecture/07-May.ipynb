3. Multiplication && Booth's Algorithm
   Standard multiplication in hardware is ... long-multiplication in binary...
   ... shift and repeat.

   BOOTH'S ALGORITHM exists to solve two problems: standard multiplication is
   slow when dealing with long strings of `1`s, and it breaks when dealing with
   negative two's complement numbers. 
   
   - THE MATH TRICK: Think about the binary number 0111_2... Normal 
     multiplication requires three separate addition... Booth's algo realises 
     that 7 is just 8 - 1... So only does one subtraction at start of string
     of 1s and one addition at the end...
   - THE RULES: You scan the multiplier from right to left, looking at the 
     current bit and the previous bit (you start by assuming a phantom `0` to
     the right of the LSB).
