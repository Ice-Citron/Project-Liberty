1. THE BRIDGE: ARGUMENTS vs. FORMULAS.
The lecture introduces a way to convert a "validity check" into a simple formula
calculation.


THE CONCEPT: CORRESPONDING IMPLICATION
- THE ARGUMENT ($\models$): You are given a list of premises 
  ($\phi_1, \dots, \phi_2$) and asked if they force a conclusion ($\psi$) to be
  True.
   - Notation: $\phi_1, \dots, \phi_n \models \psi$
- THE FORMULA ($\rightarrow$): We can glue all the premises together with 
  AND ($\wedge$) and point them at the conclusion with an IMPLICATION ($\rightarrow$)   
   - Notation: $(\phi_1 \wedge \dots \wedge \phi_n) \rightarrow \psi$

1. THE LINK BETWEEN ARGUMENTS AND FORMULAS (S10-11)

The lecture establishes a mathematical bridge between checking an ARGUMENT
(Premises $\therefore$ Conclusion) and checking a FORMULA. This is crucial 
because it allows us to use standard formula tools (like truth tables) to solve
arguments.

   - THE CONCEPT (DEFINITION 0.5): If you have an argument with premises 
     $\phi_1, \dots, \phi_n$ and conclusion $\psi$, you can convert it into a
     a single logical sentence called the CORRESPONDING IMPLICATION FORMULA.
      - SYNTAX: You connect all premises with AND ($\wedge$) and connect that
        whole block to the conclusion with IMPLIES ($\rightarrow$)
      - MATH FORM: $(\phi_1 \wedge \dots \wedge \phi_n) \rightarrow \psi$

VALID FORMULA

Definition 0.2 (Valid Formula)
   - A propositional formula is (logically) valid if it is true in every 
     situation.

Valid propositional formulas are often called TAUTOLOGIES. A tautology is just
another name for a valid propositional formula.     




SATISFIABLE FORMULA

Definition 0.3 (Satisfiable Formula)
   - A propositional formula is SATISFIABLE if it is true in at least one
     situation.             (situation being $v$ the dictionary...)... situation is evalauted by atomic evaluation
                            function $v$...

We typically say a formula is a CONTRADICTION if it is not satisfiable.
   - $\top$     satisfiable
   - $\bot$     contradiction
   - $p$        satisfiable
   - $p \wedge \neg p$      contradiction
   - $p \rightarrow p$      satisfiable




EQUIVALENT FORMULAS

Definition 0.4 (Equivalent formulas)
   Two propositional formulas $\phi, \psi$ are LOGICALLY EQUIVALENT if they are
   true in exactly the same situations. Roughly speaking: they mean the same.

Some people write $\phi \equiv \psi$ for this. But $\equiv$ is also used in 
other ways, so watch out.

  interpretation, often called a tautology in propositional logic. These 
  formulas represent necessary truths (e.g., $P \vee \neg P$) that cannot be 
  false regardless of the truth values of their components.
