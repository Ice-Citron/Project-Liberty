2. THE LATIN NAMES (Why are they special?)
These fancy names come from Latin. They are "special" because they are the 
standard "moves" in logic--like knowing how a Knight or a Bishop moves in chess.



MODUS PONENS
   - Translation: "The Method of Affirming" (from Latin ponere, "to place" or
     "affirm").
   - The Vibe: Forward motion. You have the trigger, so you get the result.
   - Why It Matters: It is the most basic "If-Then" logic used in code 
     (`if x: do x`)  

MODUS TOLLENS
   - Translation: "The method of Denying" (from Latin tollere, "to destroy" or
     "remove").
   - The Vibe: backward deduction. You see the result didn't happen, so you know
     the trigger didn't happen.
   - Why It Matters: This is how debugging works. "The database didn't update 
     (result false), so the `save()` function definition didn't run (trigger
     false)."

3. EXAMPLE WALKTHROUGH (Using Rain/Umbrella)

Let's go through the list from your slide using $\phi$ = RAIN and 
$\psi$ = UMBRELLA.


A. THE "DUH" ONE (IDENTITY)
   - FORMULA: $\phi \models \phi$   ("Rain. Therefore, Rain.")
   - VERDICT: VALID.
   - WHY: If it is true that it's raining, then... it's raining. Impossible to
     break.


B. THE "DISCARD" ONE (SIMPLIFICATION)
   - FORMULA: $\phi \wedge \psi \models \phi$ ("It is Raining AND I have an
     Umbrella. Therefore, it is Raining.")
   - VERDICT: VALID.
   - WHY: You started with more information than you needed. If both are true,
     then the first one is definitely true.        


C. THE "SPECULATOR" (Invalid Addition)
   - FORMULA: $\phi \models \phi \wedge \psi$ 
           ("It is Rainging. Therefore, it is Raining AND I have an Umbrella.")
   - VERDICT: INVALID. ($\not \models$)
