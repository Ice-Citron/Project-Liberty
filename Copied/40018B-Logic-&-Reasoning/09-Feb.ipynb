- In FOL, a TABLEAU (specifically, an analytic tableau or truth tree), is a 
  systematic, tree-structured method used to prove the validity of a formula or
  the satisfiability of a set of formulas. It operates by trying to find a 
  counterexample to a given formula; if no counterexample can be found (i.e., all
  branches of the tree close), the formula is proven valid.

- The domain of discourse (or universe of discorse) in FOL is the specific, non-
  empty set of objects, entities, or individuals over which the quantifiers
  ($\forall, \exists$) range and variables operate in a given interpretation.
  It defines the boundaries of what is being discussed, determining the truth 
  values of logical statements.


- Prenex Normal Form (PNF) is a formula representation in FOL where all 
  quantifiers ($\forall, \exists$) are moved to the front (prefix), followed by
  a quantifier-free formula (matrix). It takes the form 
  $Q_1x_1Q_2x_2 \dots Q_nx_n\phi$, where $Q_i$ are quantifiers and $\phi$ is the
  matrix. It is a standardised form, useful in automated theorem proving.
