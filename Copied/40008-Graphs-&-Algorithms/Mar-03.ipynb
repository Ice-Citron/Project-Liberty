THE CONCEPT: DFS + MEMOIZATION (Dynamic Programming)      
   Imagine you are in a city with one-way roads and no loops (a DAG), and you 
   want to know HOW MANY DIFFERENT WAYS you can drive from your current city 
   ($x$) to the destination city ($d$).

THE CORE LOGIC:
   If you know that city $x$ has three roads leading to cities $A, B,$ and $C$,
   the total number of paths from $x$ to the destination is simply:

   $(Paths from A to d) + (Paths from B to d) + (Paths from C to d)$

   To code this, we use DFS to travel as far down the road as possible until we
   hit the destination. As we backtrack (or "exit" a node), we pass the number
   of successful paths back up the chain.

THE TRICK (MEMOIZATION):
   If multiple cities connect to city $Y$, we don't want to recalculate the 
   paths from $Y$ to $d$ every single time we visit it. That would take forever!
   Instead, the first time we calculate $Y$, we save the answer in the `np`
   array and mark `entered[Y] = true` (this is a cache, a pattern you'll see
   all the time in Kotlin!). Next time we see $Y$, we just grab the saved 
   answer.

SOLVING THE PSEUDOCODE BLANKS
   
1. INSIDE THE `for y in adj[x]` LOOP
   You are standing at node $x$, looking at a neighor $y$.
   - FIRST: Have we visited $y$ yet? If not, we must explore it by calling
     `numpaths(y, d)`.
   - SECOND: Once $y$ is fully explored (either just now, or previously), we
     add its paths to our running total for $x$.

```
if not entered[y]:
    numpaths(y, d)
np[x] = np[x] + np[y]    
```


2. AT `# exiting x`
   We gave looked at all neighbors and summed up their paths. But there is one
   massive exception:
      WHAT IF $x$ IS THE DESTINATION NODE $d$?

   The problem explicitly states: "`np[d] = 1`, with the empty path being the
   only path fom $d$ to $d$." If we are standing on the destination, there is
   exactly 1 way to get there (do nothing). We need to force this base case.

```
if x == d:
    np[x] = 1
```


3. THE FINAL LOOP (The Trigger)
   You have this beautiful recursive function, but it needs to be triggered. If
   the graph is disconnected or has multiple starting points, just running
   `numpath` from one node won't visit everything. We must loop through every
   node in the graph and trigger the search if it hasn't been epxlored yet. 
   (This is the standard way to run DFS on a whole graph).

```
if not entered[x]:
    numpaths(x, d)
```


```
procedure numpaths(x, d)
   entered[x] = true
   for y in adj[x]:
      if not entered[y]:
         numpaths(y, d)
      np[x] = np[x] + np[y]

   # exiting x
   if x == d:
      np[x] = 1

for x in Nodes:
   entered[x] = false
   np[x] = 0

for x in Nodes:
   if not entered[x]:
      numpaths(x, d)   
```

   The absolute biggest number is the "boss" at the very top. Right now, our 
   tree from part (a) is a mess (e.g., the root is 3, but the child is 5).

   To fix this, the `BuildMaxHeap` algorithm works from the BOTTOM UP.
      1. We ignore the bottom row of "leaves" (7, 8, 2, 6) because they don't
         have children to be bigger than.
      2. We start at the lowest parents (working right to left) and ask: "Are
         you bigger than your kids?"
      3. If the answer is no, the parent SWAPS PLACES WITH ITS LARGEST CHILD.
      4. If a parent gets demoted downwards, we have to keep checking it against
         its new kids until it stops moving.

-- Memoization is a software optimisation technique that speeds up computer 
   programs by storing (caching) the result of expensive function calls. It 
   returns the cached result when the same input occur again, reducing 
   computation time by replacing recalculations with fast lookups. It is 
   commonly used in recursive functions and dynam ic programming to avoid
   redundant work.

   KEY ASPECTS OF MEMOIZATION
   - CACHING MECHANISM: Typically implemented using a hash table or object to
     map inputs to results.
   - SPACE-TIME TRADEOFF: It improves runtime performance by increasing memory
     usage.
   - APPLICABILITY: Best used on "pure functions" -- functions that always 
     return the same output for the same input.
   - COMMON USE CASES: Recurisve algorithms (e.g., Fibonacci numbers), heavy
     database queries, or API calls.
