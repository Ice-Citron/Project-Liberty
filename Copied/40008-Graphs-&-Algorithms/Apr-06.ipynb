### Block 1: Heaps and BuildMaxHeap
   Before we look at the algorithm, let's nail down exactly what a heap is and
   how it translates between a visual tree and a flat array.    


1. The Shape (Left-Complete Binary Tree)
   A heap structure is just a binary tree with a very strict rule: FILL IT 
   TOP-TO-BOTTOM, STRICTLY LEFT-TO-RIGHT. There are no gaps. You cannot start
   a new level until the current level is completely full, and you must fill the
   current level from the left edge.

   Because the shape is so predictable, we don't need pointers to store it. We
   just dump it into a 1-indexed array!
      * The root is at index `1`
      * The left child of index `i` is at `2i`
      * The right child of index `i` is at `2i + i`
      * The parent of index `i` is at `floor(i / 2)`


2. THE MAX-HEAP PROPERTY
   A MAX HEAP adds one rule to the shape: EVERY PARENT MUST BE GREATER THAN OR
   EQUAL TO BOTH OF ITS CHILDREN.

   * Consequently, the absolute largest element is guaranteed to be sitting at 
     the root (index 1)
   * Notice that there is no guaranteed relationship between siblings. The left
     child could be larger or smaller than the right childl it doesn't matter, 
     as long as the parent is bigger than both.


3. BuildMaxHeap (The Bottom-Up Fix)
   If you are handed a random array and told to make it Max Heap, you do NOT
   start at the root. You fix it from the bottom up!

   - THE MECHANISM: 
      1. Find the last node in the tree that actually has children. (This is 
         always at index `floor(n / 2)`).
      2. Call `fixMaxHeap` on it. If it is smaller than its children, swap it
         with the LARGEST CHILD. (If it swaps, follow it down and make sure it's
         bigger than its new children, too).
      3. Move backwards one index at a time (`i = floor(n/2) - 1`, then `-2`,
         etc.) all the way up to index `1`.
      4. By the time you fix the root, the entire tree is a perfect Max 
         Heap.
