   A SPANNING TREE of a connected graph is a subgraph that includes every node 
   from the original graph but has no cycles. Becuase it must be acyclic and 
   connect all $n$ nodes, a spanning tree always contains exactly $n - 1$ edges. 
   If you were to add even one more edge to a spanning tree, you would create a 
   cycle; if you were to remove one edge, the graph would become disconnected.
   Essentially, it is the "skeleton" of the graph that keeps all nodes 
   reachable from each other using the minimum number of connections possible.

   ... typically generate spanning trees using search algorithms like DFS or BFS
   . A DFS tree is formed by the edges used to discover new nodes during a 
   depth-first traversal, ... resulting in a "deep" or thin tree structure.
   Conversely, a BFS tree ... typically results in a "shallow" or wide tree that
   shows the shortest path from the starting node to all others in an unweighted
   graph. Note that a single graph can have many spanning trees unless the graph 
   itslef is already a tree.

### Prim && Kruskal 
   Both find a MST -- a tree connecting all nodes with minimum total edge weight.

   PRIM grows a single tree from a chosen starting node. At each step, add the 
   CHEAPEST EDGE that cross the boundary between the tree and the remaining 
   nodes. Think: "grow the tree by the cheapest extensiion."

   KRUSKAL sorts all edges by weight and processes them in order. Add each edge
   if it DOESN'T CREATE A CYCLE (i.e., its endpoints are in different connected
   components). Think "add the globally cheapest safe edge."

   UNION-FIND implements Kruskal's cycle check. Each component has a leader. 
   `find(x)` returns x's leader. `union(x, y)` merges two components. With 
   WEIGHTED UNION (smaller tree attaches to larger), tree deph is O(log n). The
   exam often asks you to draw the final parent-pointer tree.
