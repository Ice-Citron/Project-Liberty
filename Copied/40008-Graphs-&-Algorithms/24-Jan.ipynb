2. PYTHON VISUALISATION (DFS vs BFS Trees)

This code recreates the "Diamond/Grid" graph from your slides and visualises how
DFS and BFS produce completely different "spanning trees" from the same graph.

````python
import networkx as nx
import matplotlib.pyplot as plt

def visualize_traversals():
    # 1. Create the Graph from Lecture Slides (Approximate structure)
    G = nx.Graph()
    # Edges based on the visual in Slide 71/72
    edges = [
        (1, 2), (1, 8),
        (2, 3), (2, 4),
        (8, 4), (8, 7),
        (3, 5),
        (4, 5), (4, 6),
        (7, 6)
    ]
    G.add_edges_from(edges)
    
    # Fixed layout to match the slide (Diamond/Grid shape)
    pos = {
        1: (0, 1),
        2: (1, 2), 8: (1, 0),
        3: (2, 3), 4: (2, 1), 7: (2, -1),
        5: (3, 2), 6: (3, 0)
    }

    fig, axes = plt.subplots(1, 3, figsize=(18, 6))

    # --- Plot 1: The Original Graph ---
    nx.draw(G, pos, ax=axes[0], with_labels=True, node_color='lightgray', 
            node_size=700, font_weight='bold', edge_color='gray', width=2)
    axes[0].set_title("Original Graph", fontsize=14)

    # --- Plot 2: DFS Tree (Depth First) ---
    # NetworkX has a helper to generate the DFS tree from a source
    dfs_tree = nx.dfs_tree(G, source=1)
    
    nx.draw(G, pos, ax=axes[1], node_color='lightgray', node_size=500, alpha=0.3, edge_color='gray')
    nx.draw(dfs_tree, pos, ax=axes[1], with_labels=True, node_color='salmon', 
            node_size=700, font_weight='bold', edge_color='red', width=3, arrows=True)
    axes[1].set_title("DFS Traversal (Recursive/Stack)\nNotice: Long, deep paths", fontsize=14)

    # --- Plot 3: BFS Tree (Breadth First) ---
    bfs_tree = nx.bfs_tree(G, source=1)
    
    nx.draw(G, pos, ax=axes[2], node_color='lightgray', node_size=500, alpha=0.3, edge_color='gray')
    nx.draw(bfs_tree, pos, ax=axes[2], with_labels=True, node_color='lightblue', 
            node_size=700, font_weight='bold', edge_color='blue', width=3, arrows=True)
    axes[2].set_title("BFS Traversal (Queue)\nNotice: Layers based on distance", fontsize=14)

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    visualize_traversals()
````

Let's look at the mechanics strictly.

To answer your specific questions about "markings": You are correct that the
Stack/Queue is just a "Todo list" (TDL). It does not store history.

Therefore, we MUST have a second, separate lsit (usually a boolean array or set)
called `visited`.
- ToDo List (`queue`/`stack`): Stores nodes we need to look at IN THE FUTURE.
- Visited List (`visited`): Stores nodes we have ALREADY SEEN so we don't add 
  them to the ToDo list again.

Here is exactly how a node moves through the system, step-by-step, using the 
graph from slides... Node 1 is connected to 2 and 8.


<<<<<>>>>>
CASE 1: BREADTH-FIRST SEARCH (BFS) - The Queue

GOAL: Process layers evenly.
SETUP: `Queue[1]`, `Visited = {1}`


The Lifecycle of Node 2:
    1. Start (Discovery):
        - We are currently processing Node 1.
        - We look at its neighbours: 2 and 8.
        - CHECK: Is 2 in the `Visited` list? NO.
    2. ENQUEUE (The "PUSH"):
        - ACTION A: We mark `Visited[2] = True`. (This stops anyone else from
          adding it).
        - ACTION B: We add 2 to the BACK of the `Queue`.
        - State: `Queue = [8, 2]` (assuming 8 was added first).
    3. In Between (The Wait):
        - Node 2 sits in the queue doing nothing. It waits while we finishes 
          processing Node 1, then Node 8.
    4. Dequeue (The "Pop"):
        - Node 2 reaches the FRONT of the line.
        - ACTION: We remove 2 from the `Queue`.
        - State: `Queue = []`.
    5. PROCESSING (THE "AFTER"):
        - Now that 2 is active, we look at its neighbours (3, 4).
        - We repeat step 1 for them.
    6. END:
        - Node 2 is never added to the queue again because `Visited[2]` is 
          strictly `True`.



<<<<<<>>>>>>                            
CASE 2: DEPTH-FIRST SEARCH (DFS) - THE STACK

GOAL: Dive deep immediately.
SETUP: `Stack = [1]`, `Visited = {1}`


THE LIFECYCLE OF NODE 2:
1. Start (Discovery):
    - We are processing Node 1.
    - We look at its neighbors: 8 and 2.
2. Push:
    - Check: Is 2 in `Visited`? No.
    - Action A: Mark `Visited[2] = True`.
    - Action B: Add 2 to the TOP of the `Stack`.
    - State: `Stack = [8, 2]` (2 is on top).
3. In Between (The Wait):
    - THERE IS NO WAIT. Because 2 is on top, it is next.
4. Pop:
    - Action: We remove 2 from the `Stack` IMMEDIATELY in the very next step.
5. Processing (The "After"):
    - Node 2 is now the active node.
    - We look at its neighbours (3, 4).
    - We push 3 and 4 onto the stack on top of 8.
    - State: `Stack = [8, 4, 3]` (8 is buried at the bottom).
6. End:
    - Node 8 (which we saw way back in step 1) is still stuck at the bottom of
      the stack. It will only be processed after we finish 2, and 3, and 5... 
      and backtrack all the way.

SUMMARY OF DIFFERENCES


When is it marked VISITED?
- BFS (Queue): The moment we SEE it (before adding to list).
- DFS (Stack): The moment we SEE it (before adding to list).


When is it PUSHED?
- BFS (Queue): Immediately after marking.
- DFS (Stack): Immediately after marking.


Where does it GO?
- BFS (Queue): To the BACK of the line.
- DFS (Stack): To the TOP of the pile.


When is it POPPED?
- BFS (Queue): After everyone ahead of it is done.
- DFS (Stack): IMMEDIATELY (it cuts the line).


What happens after POP?
- BFS (Queue): We add its neighbours to the BACK.
- DFS (Stack): We add its neighbours to the TOP.

- You have an array `boolean visited[n]`.
- When you write code, you literally type:

if not visited[neighbour]:
