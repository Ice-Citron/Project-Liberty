The core goal of both algorithms is to visit EVERY NODE in a connected graph 
exactly once. The difference is the order in which they visit.


A. Depth-First Search (DFS)
    - The Philosophy: "GO as deep as possible, then backtrack."
    - The Metaphor: Solving a maze. You keep walking down a path until you hit a
      dead end, then you walk back to the last junction and try the other path.
    - Data Structur: STACK (or RECURSION, which uses the system call stack).
    - LECTURE EXAMPLE:
        - Start at 1. Go to neighbour 2.
        - From 2, immediately go deeper to 3.
        - From 3, go to... dead end? Backtrack to 2.
        - From 2, go down to 4.
        - Result: Long, winding paths. The resulting "Spanning Tree" is deep and
          narrow.

    - DATA STRUCTURE: QUEUE (FIFO - First In, First Out).

PART 2.6: CONNECTIVITY & CYCLE DETECTION

We can use DFS (Depth-First Search) not just to visit nodes, but to diagnose the
structure of the graph.

1. CONNECTIVITY CHECK
    - THE LOGIC: If you run DFS starting at Node $X$, you will visit exactly the
      "Connected Component" containing $\mathbf{x}$.
    - THE ALGORITHM:
        1. Mark all nodes as unvisited.
        2. Run DFS (StartNode).
        3. Check: Did the size of the "Visited" list equal the total number of
           nodes $N$?
        4. RESULT: If $len(visited) == N$, the graph is Connected. If not, it is
           Disconnected.


2. CYCLE DETECTION (THE "BACK EDGE" RULE)
   Finding a cycle is distinct from just knowing one exists (using arc counts).
   We want to find the specific loop.

   - THE LOGIC: As you traverse down the DFS tree, you are moving from
     $\text{Parent} \rightarrow \text{Child}$.
   - THE TRIGGER: If you encounter a neighbour $Y$ that has ALREADY BEEN VISITED
     $\text{AND} Y$ is NOT YOUR DIRECT PARENT, you have found a "Back Edge."
   - CONCLUSION: This edge connects you "back" to an ancestor, closing a loop.
     Therefore, a cycle exists.


3. ANCESTOR LOGIC (MATH SYNTAX)     
   The lecturer introduces a theorem about the structure of DFS trees:
    - THEOREM: If there is an arc between $X$ and $Y$ in the graph, then in the
      DFS tree, one must be the ancestor of the other.
    - IMPLICATION: You never have "Cross Edges" between different branches of 
      the DFS tree in an undirected graph. You only have Tree Edges (down) and
      Back Edges (up).

PART 2.7: TOPOLOGICAL SORTING

This is the standard algorithm for "Dependency Resolution".


1. THE PROBLEM: DIRECTED ACYCLIC GRAPHS (DAGs)
    - Scenario: You have a list of tasks. Some tasks must be done before others
      (e.g., "Put on socks" $\rightarrow$ "Put on shoes").
    - Goal: Find a TOTAL ORDERING (a linear list) s.t. if Task A must be done 
      before Task B, A appears earlier in the list than B.  
    - CONSTRAINT: The graph must be ACYCLIC (No cycles). if Cycle 
      $A \rightarrow B \rightarrow A$ exists, you can never start.


2. THE LOGIC: "EXIT ORDER IS REVERSE TOPOLOGICAL SORT"
   This is the specific algorithm your lecturer highlighted.
    - IDEA: You cannot add a task to the "Done" list until all its dependencies
      (children) are processed.
    - ALGORITHM: 
        1. Run DFS
        2. Record the moment you EXIT a node (backtrack from it).
        3. This means "I have finished everything required by this node."
        4. Add the node to a list.
    - RESULT: The list of "Exit Times" gives you the REVERSE Topological Sort.
      Flip the list to get the answer.          


3. MATH SYNTAX (PARTIAL vs. TOTAL ORDER)
    - PARTIAL ORDER ($\preceq$): The raw graph. We know $A \preceq B$, but we 
      don't know if $A$ comes before $C$ or vice versa (they are independent).
    - LINEARISATION: Topological sorting "flattens" this 2D web of dependencies
      into a 1D line (Total order) without breaking any rules.  

PYTHON LAB: CYCLE DETECTION & TOP SORT

This script implements both concepts.
    1. CYCLE DETECTION: Highlights the "Back Edge" that creates a loop.
    2. TOPOLOGICAL SORT: Resolves dependencies for a "Getting Dressed" graph.

```````````````python
import networkx as nx
import matplotlib.pyplot as plt

def visualize_algorithms():
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    # --- 1. CYCLE DETECTION (Undirected) ---
    # Create a graph with a loop: 0-1-2-0
    G_cycle = nx.Graph()
    G_cycle.add_edges_from([(0, 1), (1, 2), (2, 0), (2, 3)])
    
    # We use DFS to find the cycle
    try:
        cycle_nodes = nx.find_cycle(G_cycle, orientation='ignore')
        print(f"Cycle Found: {cycle_nodes}")
        title_cycle = "Cycle Detected via Back-Edge logic\n(Red Edge closes the loop)"
        edge_cols = ['red' if edge in cycle_nodes or (edge[1], edge[0]) in cycle_nodes else 'black' for edge in G_cycle.edges()]
    except:
        title_cycle = "No Cycle Found"
        edge_cols = 'black'

    pos_cycle = nx.spring_layout(G_cycle, seed=42)
    nx.draw(G_cycle, pos_cycle, ax=axes[0], with_labels=True, node_color='lightblue', edge_color=edge_cols, width=2, node_size=1000)
    axes[0].set_title(title_cycle)

    # --- 2. TOPOLOGICAL SORT (Directed Acyclic Graph) ---
    # Scenario: Getting Dressed
    G_dag = nx.DiGraph()
    # Dependencies: Socks -> Shoes, Underwear -> Pants -> Shoes
    edges = [
        ("Socks", "Shoes"),
        ("Underwear", "Pants"),
        ("Pants", "Shoes"),
        ("Shirt", "Tie"),
        ("Tie", "Jacket")
    ]
    G_dag.add_edges_from(edges)

    # Algorithm: DFS Post-Order (Reverse of Top Sort)
    if nx.is_directed_acyclic_graph(G_dag):
        # NetworkX has a built-in function, but the lecturer's logic is:
        # list(dfs_postorder_nodes(G)) -> gives exit order (Reverse Top Sort)
        top_sort = list(nx.topological_sort(G_dag))
        print(f"\nTopological Sort Order:\n{top_sort}")
        
        # Visualize Layers (approximate)
        # Manually setting layer positions for clarity
        pos_dag = {
            "Underwear": (0, 2), "Socks": (1, 2), "Shirt": (2, 2),
            "Pants": (0, 1), "Tie": (2, 1),
            "Shoes": (0.5, 0), "Jacket": (2, 0)
        }
        
        nx.draw(G_dag, pos_dag, ax=axes[1], with_labels=True, node_color='lightgreen', node_size=2000, arrowsize=20)
        axes[1].set_title(f"Topological Sort (Dependency Resolution)\nOrder: {top_sort}")
    else:
        axes[1].set_title("Graph has a cycle! Cannot Top-Sort.")

    plt.tight_layout()
    plt.show()

visualize_algorithms()
```````````````

PART 2.6: CONNECTIVITY & CYCLE DETECTION (APPLICATIONS OF DFS)

We move beyond simply traversing a graph to analysing its structural properties
using Depth First Search (DFS).


1. CONNECTIVITY CHECK
    - PROBLEM: Is the graph $G = (N, A)$ connected?
    - ALGORITHM:
        1. Select an arbitrary start node $s \in N$.
        2. Run DFS starting from $s$ to generate a set of visited nodes $V$.
        3. CONDITION: The graph is connected if and only if the number of 
           visited nodes equal the total number of nodes.
        - SYNTAX: $G$ is connected $\Longleftrightarrow |V| = |N|$.
    - DISCONNECTIVITY: If $|V| < |N|$, the nodes in $V$ form the CONNECTED
      COMPONENT containing $s$. To find other components, you must restart DFS
      from an unvisited node.



2. CYCLE DETECTION (THE "BACK EDGE" RULE)
   A graph with $N$ nodes and $|A| \geq N$ must contain a cycle. However, to 
   find the cycle, we use the structure of the DFS execution.

   - THE DFS TREE THEOREM: When DFS is run on a connected graph $G$, it 
     implicitly builds a "DFS Spanning Tree" $T$.
        - THEOREM: For any arc $(x, y) \in A$ in the original graph:
            - Either $(x, y) \in A$ in the original graph:
            - OR $x$ is an ANCESTOR of $y$ (or vice versa).
        - IMPLICATION: In an undirected graph, there are no "cross-edges" 
          between different branches of the tree. Every edge either goes down
          the tree or strictly back up.

   - THE DETECTION ALGORITHM: While performing DFS at node $x$, if we encounter
     a neighbour $y$:
        1. If $y$ is currently being visited (it is on the recursion stack), 
           then $y$ is an ancestor of $z$.
        2. The edge $(x, y)$ connects a node to its ancestor. This is called a
           BACK EDGE.
        3. RESULT: A Back Edge completes a cycle.
        -  SYNTAX: If $y \in \text{Visited}$ AND $y \new parent(x)$, then Cycle
           Found.

PART 2.7: TOPOLOGICAL SORTING

This algorithm applies to Direct Acyclic Graphs (DAGs). It is the mathematical
foundation for scheduling and dependency resolution.

1. THE PROBLEM DEFINITION
    - INPUT: A Directed Acyclic Graph $G = (N, A)$ representing tasks and 
      dependencies.
    - GOAL: Find a linear ordering (sequence) of nodes $x_1, x_2, \dots, x_n$
      s.t. all dependencies are respected.
    - CONDITION: For every arc $(u, v) \in A$ (meaning $u$ must happen before
      $v$), node $u$ must appear BEFORE node $v$ in the sequence.
    - MATHEMATICAL SYNTAX: In the sorted list, if index $j > i$ (meaning $x_j$
      comes later than $x_i$), then there must be NO PATH from $x_j$ to $x_i$
      in the graph.


2. PARTIAL vs. TOTAL ORDERS
    - PARTIAL ORDER ($\preceq$): The DAG represents a partial order. We know 
      $u \preceq v$ if there is a path from $u$ to $v$. However, for two 
      independent nodes $a$ and $b$, neither $a \preceq b$ nor $b \preceq a$
      holds.
    - LINEARIZATION: Topological sorting is the process of exenting this 
      PARTIAL ORDER into a TOTAL ORDER (a straight line) compataible with the
      original constraints.


3. THE "REVERSE EXIT ORDER" ALGORITHM
   The lecturer emphasizes an efficient $O(N)$ algorithm based on DFS.

   - CONCEPT: You can only say you have "finished" a task when you have finished
     all the tasks that depend on it (in a reverse dependency graph) or, more
     intuitively, you add a node to the front of the list ony when you are done #
     exploring all its children.

   - THE ALGORITHM:
     1. Run DFS on the graph.
     2. Record the time/event when you EXIT a node (backtrack from it).
     3. This implies all descendants of that node have been fully processed.
     4. RESULT: The Topological SORT IS THE lift of nodes in REVERSE ORDER OF
        EXIT.
     - Example: If you exit "Shoes" then "Socks", the reverse list is 
       "Socks, Shoes".


    - ALGORITHM SYNTAX (PSEUDOCODE FOR EXAM):
```
L = Empty List
For each node u in G:
    If u is unvisited:
        DFS(u)

Function DFS(u):
    Mark u as Visited
    For each neighbour v of u:
        If v is unvisited: DFS(v)
    Mark u as Exited
    Prepend u to L                  <-- (Add to front of list upon Exit)        
```

   - THE TRAP: Students often draw the original graph.
   - THE FIX: You must only draw the edges you TRAVERSED.

Here is the breakdown of Partial Order vs. Total Order, specifically using the
"Getting Dressed" example, which maps perfectly to Graph Theory.



---.
1. THE CORE DIFFERENCE (In Plain English)

- PARTIAL ORDER (THE "WEB"): Some things have to happen before others, but not
  everything is relarted.
    - Example: You msut put on SOCKS before SHOES. You must put on UNDERWEAR
      before PANTS.
    - The "Partial" Part: Does the rulebook say you must put on your SOCKS 
      before your UNDERWEAR? NO. They are unrelated. You can do them in any 
      order you want.
    - Graph: This is a DIRECTED ACYCLIC GRAPH (DAG). It branches out.

- TOTAL ORDER (THE "LINE"): Every single item has a strict place in the sequence
  . No ambiguity allowed.
    - Example: A timeline of years. 1990 comes before 1991. 1991 comes before 
      1992.
    - The "Total" Part: Pick ANY two years. You can always say which one came 
      first. There are no "unrelated" years.
    - Graph: This is a LINE (A straight path visiting every node).



<<<<<<<<<>>>>>>>>>      
2. HOW THIS RELATES TO GRAPH ALGORITHMS

The algorithm TOPOLOGCAL SORT is litrally a machine that converts a PARTIAL
ORDER into a TOTAL ORDER.
    - Input (Graph): "Here is a pile of clothes and some rules (Partial Order)."
        - Rule 1: Socks --> Shoes
        - Rule 2: Pants --> Shoes
    - THE PROBLEM: A CPU (or a human) can only do one thing at a time. We need
      a single linear list of instructions.
    - OUTPUT (TOPOLOGICAL SORT): "Here is a valid Total Order to execute:"
        1. Socks
        2. Patns
        3. Shoes
          - Note: The algorithm made a choice. It decided to do Socks before 
            Pants. The original Partial Order didn't care, but the Total Order
            must decide.



<<<<<<<<<>>>>>>>>>            
3. MATHEMATICAL SYNTAX (THE "REFRESHER")
If you see these symbols in your slides, here is what they mean in this context:

A. PARTIAL ORDER ($\preceq$) A relation is a Partial Order if it satisfies three
   rules:

   1. REFLEXIVE: $x \preceq x$ (Task X is done at the same time as itself).
   2. ANTISYMMETRIC: If $x \preceq y$ AND $y \preceq x$, then $x$ and $y$ are 
      the same thing. (You can't have a cycle where X is before Y and Y is
      before X).
   3. TRANSITIVE: If $x \preceq y$ and $y \preceq z$, then $x \preceq z$.
      (If Socks before Shoes, and Shoes before Leaving, then Socks before
      Leaving).


B. TOTAL ORDER
A Total Order is a Partial Order plus ONE EXTRA RULE:
   4. COMPARABILITY: For ANY distinct $x, y$, either 
      $x \preceq y \text{OR} y \preceq x$. *This forbids "independent" tasks. In
      a Total Order, you can never say "It doesn't matter".




SUMMARY
- PARTIAL ORDER: The "Logic" (The graph with branches). "Socks --> Shoes".
- TOTAL ORDER: The "Schedule" (The straight line). "Socks --> Pants --> Shirt --> Shoes..."
- Topological Sort: The converter that turns the Logic into a Schedule.









```````````````python
import networkx as nx
import matplotlib.pyplot as plt

def visualize_orders():
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    # --- 1. PARTIAL ORDER (The DAG) ---
    # Rules:
    # - Socks before Shoes
    # - Pants before Shoes
    # - Shirt before Belt
    # - Belt before Jacket
    # Note: Socks and Shirt are NOT related. (Incomparable)
    G_partial = nx.DiGraph()
    edges = [
        ("Socks", "Shoes"),
        ("Pants", "Shoes"),
        ("Shirt", "Belt"),
        ("Belt", "Jacket"),
        ("Pants", "Belt") # Pants also before Belt
    ]
    G_partial.add_edges_from(edges)

    # Draw Partial Order (Tree-like structure)
    pos_partial = nx.spring_layout(G_partial, seed=42)
    nx.draw(G_partial, pos_partial, ax=axes[0], with_labels=True, 
            node_color='lightblue', node_size=2000, arrowsize=20, font_weight='bold')
    axes[0].set_title("Partial Order (The Rules)\nNotice: Socks & Shirt are unrelated")

    # --- 2. TOTAL ORDER (Topological Sort) ---
    # We flatten the graph into a single line.
    # The algorithm makes arbitrary choices for unrelated items.
    
    # Get the list
    total_order_list = list(nx.topological_sort(G_partial))
    
    # Create a new Linear Graph to visualize the Total Order
    G_total = nx.DiGraph()
    for i in range(len(total_order_list) - 1):
        u = total_order_list[i]
        v = total_order_list[i+1]
        G_total.add_edge(u, v)

    # Draw simple straight line
    pos_total = {node: (i, 0) for i, node in enumerate(total_order_list)}
    
    nx.draw(G_total, pos_total, ax=axes[1], with_labels=True, 
            node_color='lightgreen', node_size=2000, arrowsize=20, font_weight='bold')
    axes[1].set_title(f"Total Order (One Valid Schedule)\n{total_order_list}")

    plt.tight_layout()
    plt.show()

visualize_orders()
```````````````

Let's look at the mechanics strictly.

To answer your specific questions about "markings": You are correct that the
Stack/Queue is just a "Todo list" (TDL). It does not store history.

Therefore, we MUST have a second, separate lsit (usually a boolean array or set)
called `visited`.
- ToDo List (`queue`/`stack`): Stores nodes we need to look at IN THE FUTURE.
- Visited List (`visited`): Stores nodes we have ALREADY SEEN so we don't add 
  them to the ToDo list again.

Here is exactly how a node moves through the system, step-by-step, using the 
graph from slides... Node 1 is connected to 2 and 8.


<<<<<>>>>>
CASE 1: BREADTH-FIRST SEARCH (BFS) - The Queue

GOAL: Process layers evenly.
SETUP: `Queue[1]`, `Visited = {1}`


The Lifecycle of Node 2:
    1. Start (Discovery):
        - We are currently processing Node 1.
        - We look at its neighbours: 2 and 8.
        - CHECK: Is 2 in the `Visited` list? NO.
    2. ENQUEUE (The "PUSH"):
        - ACTION A: We mark `Visited[2] = True`. (This stops anyone else from
          adding it).
        - ACTION B: We add 2 to the BACK of the `Queue`.
        - State: `Queue = [8, 2]` (assuming 8 was added first).
    3. In Between (The Wait):
        - Node 2 sits in the queue doing nothing. It waits while we finishes 
          processing Node 1, then Node 8.
    4. Dequeue (The "Pop"):
        - Node 2 reaches the FRONT of the line.
        - ACTION: We remove 2 from the `Queue`.
        - State: `Queue = []`.
    5. PROCESSING (THE "AFTER"):
        - Now that 2 is active, we look at its neighbours (3, 4).
        - We repeat step 1 for them.
    6. END:
        - Node 2 is never added to the queue again because `Visited[2]` is 
          strictly `True`.



<<<<<<>>>>>>                            
CASE 2: DEPTH-FIRST SEARCH (DFS) - THE STACK

GOAL: Dive deep immediately.
SETUP: `Stack = [1]`, `Visited = {1}`


THE LIFECYCLE OF NODE 2:
1. Start (Discovery):
    - We are processing Node 1.
    - We look at its neighbors: 8 and 2.
2. Push:
    - Check: Is 2 in `Visited`? No.
    - Action A: Mark `Visited[2] = True`.
    - Action B: Add 2 to the TOP of the `Stack`.
    - State: `Stack = [8, 2]` (2 is on top).
3. In Between (The Wait):
    - THERE IS NO WAIT. Because 2 is on top, it is next.
4. Pop:
    - Action: We remove 2 from the `Stack` IMMEDIATELY in the very next step.
5. Processing (The "After"):
    - Node 2 is now the active node.
    - We look at its neighbours (3, 4).
    - We push 3 and 4 onto the stack on top of 8.
    - State: `Stack = [8, 4, 3]` (8 is buried at the bottom).
6. End:
    - Node 8 (which we saw way back in step 1) is still stuck at the bottom of
      the stack. It will only be processed after we finish 2, and 3, and 5... 
      and backtrack all the way.

SUMMARY OF DIFFERENCES


When is it marked VISITED?
- BFS (Queue): The moment we SEE it (before adding to list).
- DFS (Stack): The moment we SEE it (before adding to list).


When is it PUSHED?
- BFS (Queue): Immediately after marking.
- DFS (Stack): Immediately after marking.


Where does it GO?
- BFS (Queue): To the BACK of the line.
- DFS (Stack): To the TOP of the pile.


When is it POPPED?
- BFS (Queue): After everyone ahead of it is done.
- DFS (Stack): IMMEDIATELY (it cuts the line).


What happens after POP?
- BFS (Queue): We add its neighbours to the BACK.
- DFS (Stack): We add its neighbours to the TOP.
