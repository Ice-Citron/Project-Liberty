1. THE CONCEPTS: DFS vs. BFS

The core goal of both algorithms is to visit EVERY NODE in a connected graph 
exactly once. The difference is the order in which they visit.


A. Depth-First Search (DFS)
    - The Philosophy: "GO as deep as possible, then backtrack."
    - The Metaphor: Solving a maze. You keep walking down a path until you hit a
      dead end, then you walk back to the last junction and try the other path.
    - Data Structur: STACK (or RECURSION, which uses the system call stack).
    - LECTURE EXAMPLE:
        - Start at 1. Go to neighbour 2.
        - From 2, immediately go deeper to 3.
        - From 3, go to... dead end? Backtrack to 2.
        - From 2, go down to 4.
        - Result: Long, winding paths. The resulting "Spanning Tree" is deep and
          narrow.


B. BREADTH-FIRST SEARCH (BFS)
    - THE PHILOSOPHY: "Visit neighbours first, then neighbours of neighbours."
    - THE METAPHOR: A ripple in a pond or a wildfire. It expands in layers 
      (Distance 1, then Distance 2, etc.)
    - DATA STRUCTURE: QUEUE (FIFO - First In, First Out).
    - LECTURE EXA

1. WHY DFS IS A STACK (LIFO: Last In, First Out)

THE LOGIC: "I just found a new path! I'm going to explore it IMMEDIATELY and
ignore the other options for now."

Imagine you are at Node 1. You see Node 2 and node 8.
    1. You put Node 8 on your TDL.
    2. You put Node 2 on your TDL.
    3. THE STACK RULE: You must take the TOP item (the one you just put there).
    4. So you pick NODE 2 immediately. (You ignore node 8).
    5. At Node 2, you see Node 3. You put Node 3 on the stack.
    6. THE STACK RULE: You pick NODE 3 immediately.

    RESULT: You keep going deeper (1 -> 2 -> 3 ...) because you are always 
            chasing the newest discovery. The "old" options (like Node 8) sit
            at the bottom of the pile waiting for you to backtrack.
                - RECURSION: When you write `dfs(x)` inside a function, the 
                  computer literally uses a "Call Stack" in memory. It pauses
                  `dfs(1)` to run `dfs(2)`. It pauses `dfs(2)` to run `dfs(3)`.
                  It's a stack!
