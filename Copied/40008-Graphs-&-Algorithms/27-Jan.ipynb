This mark scheme is written in a very condensed, mathematical shorthand. It is
confusing because it skips the "real world" logic of what a complement actually
is.

Here is the breakdown of the logic in plain English, using a "Pizza" analogy.



1. THE CORE CONCEPT: THE "WHOLE PIZZA" ($K_n$)

Imagine the COMPLETE GRAPH ($K_n$)--a graph where every single node is connected
to every other node--as a WHOLE PIZZA. 
    - GRAPH $G$ represents the slices you eat.
    - COMPLEMENT $\overline{G}$ represents the slices left in the box.

Together, $G$ and $\overline{G}$ must add up to the Complete Graph. Every 
possible edge exists either in $G$ or $\overline{G}$, but never in both.



2. THE CONDITION: "SELF-CMPLEMENTARY"

The problem states that $G$ is isomorphic to $\overline{G}$. In our analogy, 
this means THE SLICES YOU ATE ($G$) MUST BE EXACTLY EQUAL TO THE SLIDES LEFT IN
THE BOX ($\overline{G}$).

For this to be true, you must have eaten EXACTLY 50% of the pizza.
    - Number of edges in $G$ = Number of edges in $\overline{G}$
    - Therefore, the TOTAL NUMBER OF EDGES in the complete graph ($K_n$) must be
      an EVEN NUMBER (so it can be divided by 2).



3. THE TEST: SIX NODES ($n = 6$)

Let's see if a 6-node graph can be split exactly in half.

STEP 1: CALCULATE THE "WHOLE PIZZA" (TOTAL POSSIBLE EDGES)
For 6 nodes, how many possible connections are there? We use the formula for 
edges in a complete graph $K_n$:

$$\text{Total Edges} = \frac{n(n-1)}{2}$$

Substitute $n = 6$:

$$\text{Total Edges} = \frac{6 \times 5}{2} = 15$$

So, a complete graph with 6 nodes ($K_6$) has 15 EDGES.



STEP 2: TRY TO SPLIT IT IN HALF
If $G$ is isomorphic to $\overline{G}$, they must share those 15 edges equally.

$$\text{Edges in G} = \frac{15}{2} = 7.5$$

THE CONTRADICTION: You cannot have 7.5 EDGES in a graph. You can't have half an
edge. Because 15 is an odd number, you cannot split the edges perfectly into two
equal sets. Therefore, a 6-node graph can never be isomorphic to its complement.



<<<<<<<<<<>>>>>>>>>>
COMPARISON: WHY IT WORKS FOR $n = 4$ OR $n = 5$

TRY 4 NODES ($n = 4$):
    - Total edges in $K_4 = \frac{4 \times 3}{2} = 6$
    - Can you split 6 in half? YES. Each graph gets 3 edges.
    - Result: POSSIBLE (e.g., the Path graph $P_4$).

TRY 5 NODES ($n = 5$):
    - Total edges in $K_5 = \frac{5 \times 4}{2} = 10$
    - Can you split 10 in half? YES. Each graph gets 5 edges.
    - Result: Possible (e.g., the Cycle graph $C_5$).



THe "GENERAL RULE" (The bit in brackets)

The mark scheme mentions $n$ must be congruent to 0 or 1 mod 4. This is just a
shotcut derived from the logic above.

For $\frac{n(n-1)}{2}$ to be an even number (so it can be split in half), the
numerator $n(n-1)$ must be divisible by 4.
    - If $n = 6$: $6 \times 5 = 30$ (Not divisible by 4). Fail.
    - If $n = 4$: $4 \times 3 = 12$ (Divisible by 4). Pass.
    - If $n = 5$: $5 \times 4 = 20$ (Divisible by 4). pass.


SUMMARY OF THE MARK SCHEME LOGIC:
    1. PREMISE: If $G \cong \overline{G}$, then $|E(G)| = |E(\overline{G})|$.
    2. CONSTRAINT: $|E(G)| + |E(\overline{G})| = |E(K_6)|$
    3. CALCULATION: $|E(K_6)| = 15$
    4. CONFLICT: $2 \times |E(G)| = 15$. This has no integer solution.
    5. CONCLUSION: Contradiction found. Impossible.

2. A bipartite graph (notes p12) is a graph $G$ where nodes($G$) is the union of
   two disjoint sets $A, B$, and there are no arcs connecting two members of $A$
   or two members of $B$. A graph is bipartite iff it is 2-colourable.

   Suppose that $G$ is bipartite (split into sets $A$ and $B$) and connected. 
   Define the distance between two nodes to be the length of the shortest path
   joining them. What can you say about the distance between two nodes in $A$?
   What about the distance between a node in $A$ and one in $B$? Explain your
   answers.



<<<<<<<<<>>>>>>>>>

(I) Let the graph have $n$ nodes. Then the distance between any two points is
    $\leq n - 1$ (since if it were $\geq n$ one would be taking a path with a
    repeated node, and there would clearly be a shorter path without repeats).

(II)Furthermore the distance between a node in $A$ and one in $B$ is odd, and 
    distance between two nodes of $A$ is even. This is because any path between
    two nodes must swap to and fro between $A$ and $B$, since nodes in $A$ are
    not directly connected, and the same for $B$.
    Let $P_n$ be the graph with $n$ nodes connected in a row--e.g., $P_5$. 
    Then $P_n$ is bipartite and can be used to show that the distance between 
    nodes can attain all possible values subject to conditions (I) and (II).

(III)(Thanks to Alexandros Bouganis) By combining the ideas of (I) and (II), one
    can see that the distance between any two nodes in $A$ is 
    $\leq 2 \times min(|B|, |A|-1)$. This represents an improvement on (I) in
    the case where the split is "unbalanced", with A and B not of roughly equal
    size. Similarly, the distance between a node in $A$ and one in $B$ is
    $\leq 2 \times min(|A|, |B|) - 1$.

THE CORE CONCEPT: THE "RIVER CROSSING"
A BIPARTITE GRAPH splits all nodes into two islands (Set A and Set B).

    - THE CONSEQUENCE: To move from a node in $A$, you MUST cross to $B$. To

Every pat looks like this:

PART 1: DISTANCE BETWEEN TWO NODES IN A (Parity Argument)

The question asks: "What can you say about the distance between two nodes in $A$
?"

THE LOGIC: Imagine you start at your house on ISLAND A.
1. STEP 1: You cross the bridge. You are now on ISLAND B. (Distance = 1, Odd)
2. STEP 2: You cross back. You are now on ISLAND B. (Distance = 2, Even)
3. STEP 3: You cross again. You are on ISLAND B. (Distance = 3, Odd)

THE ANSWER: To start at $A$ and end at $A$, you must make a return trip. 
Therefore, the distance (number of edges) MUST BE EVEN.

    MARK SCHEME QUOTE: "The distance between two nodes of a is even."


<<<<<<<<<<<<<<<<>>>>>>>>>>>>>>>>
PART 2: DISTANCE BETWEEN A AND B

The question asks: "What about the distance between a node in $A$ and one in 
$B$?"

THE LOGIC: Using the same logic above, if you start on Island $A$ and want to
finish on Island $B$, yu must end on an "outward" trip.
    - 1 step  $(\rightarrow B)$
    - 3 steps $(\rightarrow B \rightarrow A \rightarrow B)$
    - 5 steps...

THE ANSWER: Te distance MUST BE ODD.

    MARK SCHEME QUOTE: "The distance between a node in A and one in B is odd."




<<<<<<<<<<<<<<<<>>>>>>>>>>>>>>>>    
PART 3: THE "MAX LIMIT" $(n - 1)$
The mark scheme adds a general rule for any graph, not just bipartite ones.

THE LOGIC: If you have $n$ nodes, the longest "shortest path" possible is 
$n - 1$.
    - WHY? If a path has $n$ edges, it touches $n + 1$ nodes.
    - By the Pigeonhole Principle, if you touch $n + 1$ nodes in a graph that 
      only has $n$ nodes, you must have visited a node twice.
    - Visiting a node twice means you made a LOOP.
    - If you made a loop, it's not the shortest path (you could've just skipped
      the loop!)

    MARK SCHEME QUOTE: "The distance between any two points is $\leq n - 1$."  




<<<<<<<<<<<<<<<<>>>>>>>>>>>>>>>>
PART 4: THE "ADVANCED" BOUND (Alexandros' Note)

The mark scheme includes a sophisticated "extra" point at the bottom (labelled
III). This is useful for "A*" understanding.

It says the distance is limited by TWICE THE SIZE OF THE SMALLER SET.

WHY? Imagine Island $A$ has 100 nodes, but Island $B$ has only 1 NODE.
    - Start at $A_1 \rightarrow B_1 \rightarrow A_2 \rightarrow B_1 \dots$
    - Wait! You just visited $B_1$ twice. You made a loop!
    - Therefore, if Island $B$ is tiny, you can't have a long path without
      repeating nodes in $B$.

THE FORMULA: Distance between two $A$ nodes $\leq 2 \times |B|$. (You can only
bounce off the nodes in $B$ as many times as there are nodes in $B$.)

    MARK SCHEME QUOTE: "The distance between any two nodes in $A$ is $\leq 2$.
    $min(|B|, |A| - 1)$"

    - A to B: Always ODD (Ping-Pong away).
    - MAX DISTANCE: Always $< n$ (No loops allowed).

3 (a) Assume that $G$ has a 2-colouring col. Let $n_1, \dots, n_k, n_1$ be a
      cycle with $k$ odd. Then $col(n_1) = col(n_k)$, since colours must 
      alternate along the cycle. This is a contradiction since $n_k$ is adjacent
      to $n_1$.

      NB The argument works even if $k = 1$ (i.e. the cycle is a loop). In fact
      a node with a loop cannot be coloured no matter how many colours we allow.


  (b) Assume that $G$ has no odd length cycles. Using the idea of 2(II), define
      the distance between two nodes to be even (resp. odd) if they are joined
      by a path of even (resp. odd) length.

      This is well defined; suppose not. Then we would have two nodes $n$, $n'$
      joined by a path $\pi_1$ of even length and another path $\pi_2$ of odd
      length. Suppose further that the sum of the lengths of the two path is as
      small as possible (this amounts to showing that there can be no such 
      example $\pi_1$, $pi_2$ by strong induction).

      First note that neither $pi_1$ nor $pi_2$ can repeat a node; otherwise 
      there would be a cycle, which must be of even length, and which can 
      therefore be removed to get a shorter example.

      Furthermore $pi_1$, $pi_2$ cannot share any node apart from $n, n'$:
      otherwise we would have a shared node $n'' \neq n, n'$. In that case
      either the two paths from $n$ to $n''$ or the to paths from $n''$ to $n'$
      would have different parities. So we would have a smaller example than
      $pi_1$, $pi_2$.

      Then we can combine the paths $pi_1$, $pi_2$ to get an odd length cycle,
      which is impossible by assumption (thanks to Razvan Rosie).


      Next we define the colouring: Take any node $n$ as a "start" node, and
      set $col(n) = 1$. Set $col(n') = 1$ for any node $n'$ which is an even
      distance from $n$. If the graph is not connected then there will be
      uncoloured nodes. Take any uncoloured node as a new "start" node and 
      repeat the procedure. We must now check that this is a 2-colouring.
      Suppose that $n$ and $n'$ are adjacent. Then what $n$ was coloured $n'$
      must also have been coloured, and must have been given the opposite colour
      , since its distance from the start node is clearly of opposite parity.

This is a classic question. In Graph Theory terms, this question is asking you
to prove the "Fundamental Theorem of Bipartite Graphs".

It establishes the "Golden Rule" of 2-colouring: A GRAPH IS 2-COLOURABLE IF AND
ONLY IF it has no odd cycles.

Here is the run-through of the logic, decoding the dense mark scheme for you.



<<<<<<<<<<<<<<<<<<<>>>>>>>>>>>>>>>>>>>
PART (a): THE EASY DIRECTION

"If you can 2-colour it, you can't have an odd cycle."

Think of this as the "Traffic Light" proof.
    - Imagine you are walking around a cycle in the graph.
    - The rules of 2-colouring say every step MUST change colour.
        - Step 1: Red -> Blue
        - Step 2: Blue -> Red
        - Step 3: Red -> Blue
        - ...
    - Notice the pattern?
        - Odd steps land on BLUE.
        - Even steps land on RED.

    - THE CONTRADICTION: If you have an ODD CYCLE (say, length 5), you take 5
      steps to get back to the start.
        - Start Node: RED.
        - After 5 steps: You are back at the Start Node.
        - According to the "Odd Step" rule, you must be on BLUE.
        - IMPOSSIBLE! A node cannot be Red and Blue at the same time.

    - MARK SCHEME SUMMARY:
        - "If $k$ is odd, $col(n_1)$" and $col(n_k)$ must be different colours. 
          But they are neighbours (closing the loop), so they must be different
          colours. Contradiction"

PART (B): THE HARD DIRECTION (THE CONSTRUCTION)

"If there are no odd cycles, prove you CAN 2-colour it."


This relies heavily on QUESTION 2 (the Ping-Pong distance logic we just 
discussed). The mark scheme uses a proof by contradiction, but here is the
CONSTRUCTIVE LOGIC which is much easier to understand intuitively.



STEP 1: THE "ALGORITHM"
How do you actually colour this graph? You use the "Distance from Start" method.
    1. Pick a random node, call it the ROOT ($n$). Colour it COLOUR-1.
    2. Measure the distance ($d$) from the Root to every other node.
    3. THE RULE:
        - If distance $d$ is EVEN --> Colour 1.
        - If distance $d$ is ODD --> Colour 2.

    (This is exactly the Ping-Pong logic: Even bounces land home, Odd bounces
    land away).



STEP 2: THE "DANGER" (IS IT WELL-DEFINED?)
The mark scheme spends most of its time worrying about one specific problem:
"What if a node has TWO different paths to it--one Even and one Odd?"

If Node $X$ is 4 steps away (Even) but also 7 steps away (Odd) via a different
route, our colouring rule breaks. We wouldn't know whether to paint it COLOUR-1
or COLOUR-2.



STEP 3: THE "TRAP" (WHY THE DANGER IS IMPOSSIBLE)
The mark scheme proves that this "Danger Scenario" CANNOT HAPPEN because we 
assumed there are no odd cycles.

HERE IS THE VISUAL PROOF:
    1. Imagine Node $X$ has an EVEN PATH ($\pi_1$) from Root.
    2. Imagine Node $X$ also has an ODD PATH ($\pi_2$) from ROOT.
    3. COMBINE THEM! Walk from Root to $X$ (Even) and then walk backwards from
       $X$ to Root (Odd).
    4. Total Length = Even + Odd = Odd Number.
    5. You just created a closed loop (a cycle) of ODD LENGTH.
    6. CONTRADICTION! The question stated "Assume $G$ has no odd length cycles."

Therefore, it is impossible for a node to have "mixed parity". Every node is
strictly "Even-distance" or "Odd-distance".



DECODING THE MARK SCHEME'S NOTATION
The mark scheme gets very technical with $pi_1$ and $pi_2$. Here is what they
are saying:

   - "SUM OF LENGTHS IS AS SMALL AS POSSIBLE (STRONG INDUCTION)..."
   - TRANSLATION: They are being rigorous to avoid "messy" cycles (like 
     figure-8s). They are saying "Let's look at the smallest, tightest loop that
     causes a problem."

   - "$\pi_1, \pi_2$ CANNOT SHARE ANY NODE APART FROM $n, n'$"
   - TRANSLATION: If the two paths touched in the middle, we would just look at
     the smaller loop formed by that touch-point. (Again, simplifying the 
     problem to a simple ring).

   - "SUPPOSE THAT $n$ AND $N'$ ARE ADJACENT..."  
   - TRANSLATION: This is the final check. We coloured everything based on #
     distance. Now we check: Do any neighbours have the same colour?
   - If $n$ and $n'$ are neighbours, their distances from the start must differ
     by exactly 1 (Parity flips).
   - Therefore, their colours must be different.
   - THE COLOURING WORKS.


SUMMARY FOR YOUR REVISION
To answer this question in an exam:

   1. DEFINE THE COLOURING: "Colour nodes based on the parity of their shortest
      path from a start node $n$."
   2. VERIFY IT WORKS: "If neighbours $u$, $v$ had the same colour, they would
      both be Even distance or both Odd distance from start."#
   3. THE KNOCKOUT PUNCH: "This implies a closed walk of length (Even + Even + 1)
      or (Odd + Odd + 1), which is ODD. Since there are no odd cycles, this is
      impossible."

- In Graph Theory, the expression $G \cong \overline{G}$ signifies that a graph
  $G$ is SELF-COMPLEMENTARY.
  - $G$: Represents a simple graph.
  - $\overline{G}$: Represents the complement graph of $G$ (a graph with the
    same nodes, but where an edge exists if and only if it does not exist in $G$
    )
  - $\cong$: Means "is isomorphic to," indicating that $G$ and $\overline{G}$
    have the same structure/shape, even if their vertex labels are different.



  KEY DETAILS ABOUT SELF-COMPLEMENTARY GRAPHS ($G \cong \overline{G}$)
  - STRUCTURE: If $G \cong \overline{G}$, the graph is isomorphic to its own
    complement.
  - VERTEX COUNT: The number of vertices $n$ in a self-complementary graph must
    be $n \equiv$ 0 or 1 (mod 4) (i.e., $n$ must be $4k$ or $4k + 1$).
  - EDGE COUNT: The number of edges in $G$ is exactly half of the total edges in
    a complete graph with $n$ vertices:

    $|E(G)| = \frac{1}{2} {n \choose 2} = \frac{n(n - 1)}{4}$
  - EXAMPLES: The path graph $P_4$ (4 vertices in a line) and the cycle graph 
    $C_5$ (a pentagon) are classic examples of self-complementary graphs.   



- A graph is BIPARTITE if and only if it is 2-colourable.
