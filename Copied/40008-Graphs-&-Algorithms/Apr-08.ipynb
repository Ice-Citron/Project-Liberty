-- In graph theory, the word "PROPER" refers to the golden rule of colouring:
   it means that NO TWO ADJACENT NODES ARE ALLOWED TO SHARE THE SAME COLOUR. If
   there is a line (an edge) connecting Node A and Node B, they must be painted
   differently... if not ... entire colouring is considered "improper"
