## What a reduction is saying
   If you show

   $$A \leq B$$

   you are saying
   
   1. take any instance `x` of a problem `A`
   2. transform it in polynomial time into `f(x)`, an instance of problem `B`
   3. in such a way that

   $$A(x) \Longleftrightarrow B(f(x))$$

   So solving `B` on `f(x)` would let you solve `A` on `x`.


## Why that proves hardness
   If `A` is already known hard, and $A \leq B$, then:
      * if I had a fast algorithm for `B`
      * I could use it to solve `A` fast:
         1. compute `f(x)`
         2. run the fast `B` algorithm on `f(x)`
         3. use the equivalence to answer `A(x)`

   So `B` cannot be easier than `A`
