   ...


WHY IS THIS NECESSARY?
   In a formal proof, you cannot just say "everyone knows a King is worth 10."
   You have to point to a specific rule.

   Look at the `cardValue` implementation again.
      - ACES: If played `Lo`, they are 1.
      - FACE CARDS (J, Q, K): They are exactly 10.
      - NUMBER CARDS (2-10): They range from 2 to 10.

   Because 10 is the absolute maximum value any single card can contribute when
   Aces are Low, this lemma gives you an UPPER BOUND.
