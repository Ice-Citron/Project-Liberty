## 1. The Loop Variant ($V$) -- "The Countdown Timer"

### The Intuition:
   How do you prove mathematically that a loop will eventually stop? You find an
   integer formula that acts like a countdown timer.
   
   - Rule 1: It must strictly decrease by at least 1 every single iteration.
   - Rule 2: It must always be >= 0 as long as the loop condition is True.
   - Visualisation: If your loop says `while (i < n)`, your countdown timer is
     simply the distance between them: $V = n - i$. As $i$ grows, the distance
     shrinks. When it hits 0, the loop dies.


---
## 2. The Loop Invariant ($I$) -- "The Progress Bar"

### The Intuition:
   A loop is chaotic; variables are constantly mutating. The Invariant is a 
   mathematical anchor. It is a statement that is true before the loop starts,
   true during every single iteration, and safety true after the loop
    terminates.

   - VISUALISATION: Think of a progress bar filling up an array.
     An invariant usually says:
      1. THE BOUNDS: 0 <= i <= a.size (where is my progress bar?)
      2. THE DONE PILE: "Everything from index 0 to $i$ is perfectly sorted."
      3. THE UNTOUCHED PILE: "Everything from index $i$ to the end is exactly
         how it was in $a_pre$."


---
3. THE MID-CONDITION ($M$) immediately after a loop

### The Golden Rule:
   When a loop finishes, you know exactly two things for absolute certainty.
   1. The Invariant ($I$) survived.
   2. The loop condition is now mathematically False ($\neg cond$). Therefore,
      the mid-condition right after a loop is always: $M = I ^ \neg cond$.

### Question: 2025 - perfect_shuffle
(Note: `r` in the POST condition refers to the returned array, which in the code
is `res`). `PRiffle(a, k)` is just a helper function that tells us the correct
shuffled value for index `k`.

```Kotlin
fun perfect_shuffle(a: Array<String>): Array<String>
// PRE: true
// POST: r[..] ~ a[..] ≈ a[..]pre ^ \forall k \in [0..a.size). [r[k] = PRiffle(a, k)]
{
   val res = Array<String>(a.size){""}
   val mid = a.size / 2
   var i = 0

   // INV: ??? (I)
   // VAR: ??? (V)
   while (i < mid) {
      res[2*i] = a[i]
      res[2*i + 1] = a[i + mid]
      i++
   }
   // MID: ??? (M)

   if (a.size % 2 == 1) {
      res[2*i] = a[2*i]
   }
   return res
}
```

### Why this works:
   - THE VARIANT: Starts at `mid`, decreases by 1 each time `i++` runs, and hits
     0 exactly when the loop breaks.
   - THE PROGRESS BAR (in the Invariant): The loop sets `res` at index `2*i` and
     `2*i + 1`. This means after `i` iterations, you have successfully filled a
     solid block in `res` from index `0` up to `2*i - 1`. The `[0..2*i)` half-open
     interval captures this perfectly.
   - THE MID-CONDITION: Notice how the mid-condition proves the loop filled
     `res` up to `2*mid`. If `a.size` was an even number (like 52), 2 * 26 = 52,
     meaning the whole array is shuffled! If it was an odd number (like 51), it
     proves we filled up to index 50, which perfectly sets up the final 
     `if (a.size % 2 == 1)` check outside the loop to handle the last
     leftover card.

1. PROOF OBLIGATIONS (The Core Concept)
   THE CHEATSHEET RULE: To prove {P} `code` {Q}, you must show that P (modified
   by the code) --> Q.

### THE INTUITION:
   A HOARE TRIPLE -- writeen as {P} code {Q} -- is a legal contract.
   - P (Pre-condition): The state of the universe before the code runs.
   - Q (Post-condition): The promised state of the universe after the code runs.

   A PROOF OBLIGATION is the actual mathematical homework you have to do to
   prove the code didn't break the contract. 


### How it works: 
   You take the "Before" universe ($P$), you mathematically apply the exact 
   action the code took, and you set up an implication (-->). If yu can successfuly
   deduce Q on the right side of the arrow, you have fulfilled your obligation.

   - Example...
   - The obligation: true ^ x - 5 --> x > 0 (This is trivially true, so the code is
     proven correct).

   In the context of program reasoning and proof obligations, $Mod(S)$ (short
   for MODIFIED) refers to the set of program variables whose values may be
   altered during the executuion of a statement or block $S$. It is a critical
   component used to define the frame of a computation, ensuring that any

   the modification set is essential for maintaining logical consistency, as it 
   allows the verifier to safely assume that properties concerning "untouched"
   variables in the precondition will stay hold in the postcondition. 

   Formally, $Mod(S)$ is defined syntactically by examining the structure of the
   program code. For a simple assignment statement such as $x := e$, the 
   modification set is simply ${x}$. In more complex structures like a sequence
   of statements S_1; S_2, the set is the ...

   The primariy utility of $Mod(S)$ lies in generating FRAME AXIOMS. Without a
   formal way to track modifications, a proof would technically need to 
   explicitly state for every single variable in the system that it did
   not change ... By using the modification set, we can invoke a general rule:

## 1. What is the "bar"s ($\bar{x}$ and v-bar)
   The bar just means "a list of".

   - x-bar (The parameters): The placeholder variables defined in the function 
     signature. (e.g., `fun add(x1: Int, x2: Int)`)
   - v-bar (The Arguments): The actual real variables you are passing into the
     function from your code. (e.g., `add(a, b)`)

## 2. What is the double bracket? `S [x |-> v] [v_pre |-> v_old]`
   This is SEQUENTIAL FIND AND REPLACE. You run the first substitution, get a 
   new string, and then immediately run the second substitution on that new
   string. 

   Let's look at why we need two steps to use a Black Box function@:
   - The function promises a Post-condition $S$. But $S$ is written using the 
     function's internal parameters (x-bar).

     "initial state" of the parameter using `_pre` (e.g., $a_pre$). But in your
     timeline, the code just ran, so the "initial state" before the line executed
     is called `_old` (e.g., )... So you find every `_pre` and replace it with 
     `_old`.

   You have now successfully translated the Black Box's generic contract into the

### 1. What to look out for: The Loop Invariant (I)
   A strong loop invariant always consists of TWO OR THREE DIFFERENT CLAUSES
   joined by an logical AND (^). When you look at a loop, mechanically extract
   these three pieces:

A. THE COUNTER BOUNDS
   - WHERE TO LOOK: Look at the variable initialisation right before the loop
     (e.g., `var i = 0`) and the `while` condition (e.g., `while (i < N)`).
   - WHAT TO WRITE: Define the absolute minimum and maximum values the counter
     can hold.
   - THE TRAP: The upper bound MUST include the exit value. If the loop is 
     `while (i < N)`, the invariant bound is 0 <= i <= N. It is NOT i < N. Why?
     Because the invariant must still be true the exact millisecond the loop
     terminates (when i = N).

B. THE "PROCESSED" STATE (The Work Done)
   - WHERE TO LOOK: Look inside the loop body. What is the code doing to the 
     current index $i$?
   - WHAT TO WRITE: Write a universal quantifier ($\forall$) that states the
     intended post-condition holds for all elements from the start up to the
     current counter. 
   - STANDARD FORMAT: $\forall j \in [0..i)$ [Result at j is correct].
   - EXAMPLE: If the loop compute sequares, look at `res[i] = a[i] * a[i]`. The
     processed clause is: $\forall i \in [0..in) . [res[j] = a[j]^2]$

C. THE "UNPROCESSED" STATE (The Frame)
   - WHERE TO LOOK: Look at the array or variables that are being modified.
     Do the elements after $i$ need to remain unotuched for the logic to work.
   - WHAT TO WRITE: A statement guaranteeing the remainder of the array hasn't
     been correupted yet.
   - STANDARD FORMAT: $\forall k \in [0..N) . [a[k] == a_pre[k]]$


PUTTING $I$ TOGETHER:  
   $I \equiv (Bounds) \wedge (Processed[0..i) \wedge (Unprocessed[i..N))$
