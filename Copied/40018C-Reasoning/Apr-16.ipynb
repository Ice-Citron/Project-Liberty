   ... we are transitioning away from the rigid, microscopic tress of ND (from
   logic) and moving into Semi-Structured Proofs.

   The goal of a semi-structured proof is to combine the rigor of formal logic
   with the readability of a human argument. It allows you to make the structure
   clear and check each step locally, while giving you the freedom to vary the
   "size" of your logical steps.

   ... four core structural concepts you need for the exam.


---
### 1. What is Given
   Before you write any logical steps, you must explicitly unpack your starting 
   materials.   
      - THE CONCEPT: You are setting up your workspace and giving a short,
        unique name to every single premise so you never have to rewrite the
        full formula again.
      - THE RULE: Write out and name each given formula.
      - THE SYNTAX: You create a `Given:` block at the top of the page. You
        list the premises and label them with numbers, e.g., (1), (2), (3).

### 2. What is to be Shown
   In ND, the target was just floating at the bottom of the page. In semi-structured
   proofs, you declare your exact target upfront.
   - THE CONCEPT: This acts as your compass. By writing it down explicitly, you
     define exactly what success look like.
   - THE RULE: Write out and name each formula to be shown.
   - THE SYNTAX: You create a `To show:` block. To prevent confusing your
     targets with your premises, targets are usually labeled with Greek letters,
     such as ($\alpha$).

### 3. What is taken Arbitrary
   This is the semi-structured equivalent of opening a Universal Introduction
   ($\forall I$) box.
   - THE CONCEPT: When your `To show` target involves proving a universal
     statement (e.g., "All green dragons are happy"), ou cannot prove it by
     testing every dragon. You must pick a completely random, featureless
     entity to represent all of them.
   - THE RULE:; You must explicitly declare what is taken arbitrary.
   - THE SYNTAX: You simply write "take an arbitrary [entity], [Name]" or
     "take arbitrary $x$". You no longer need to draw physical scoope boxes 
     around your work, but this declaration is strictly required for the
     examiner to accept your universal conclusion.

### 4. STRUCTURING A PROOF
   A semi-structured proof reads like a highly organised script. Here is the
   exact architectural skeelton you are expected to use in the exam:

   GIVEN:
   (1) [Formula 1]
   (2) [Formula 2]

   TO SHOW:
   ($\alpha$) [Target Formula]

   PROOF: 
   - ARBITRARY DECLARATIONS: Start by unpacking outer quantifiers. (e.g., "Take
     ~$x$ arbitrary").
   - ASSUMPTIONS (`asm`): If your target is an implication ($A \to B$), you
     write "(ass) A" and then explicitly state your sub-goal by writing 
     "need to show B". You can have multiple assumptions labeled (ass1), (ass2),
     etc. 
   - NUMBERED STEPS: Every new logical deduction gets a sequential number,
     e.g., (3), (4), (5).
   - RIGHT HAND JUSTIFICATIONS: This is the most critical rule. Every single
     numbered step must be justoified on the RHS. You justify each step in terms
     of earlier steps, assumptions, or lemmas using phrases like "follows from
     (1) and (ass)".
   - THE CONCLUSION: You end the proof by explicitly stating that the target
     has been reached: "Thus ($\alpha$) holds."

## Subtopic 1: Mathematical Induction (The Intuition)

### The Intuition
   Induction is the domino effect translated into mathematics. If you want to
   prove a property holds for an infinite number of cases (like all positive
   integers), you cannot test them one by one. Instead, you only need to prove
   two things:

   1. THE BASE CASE: You can push the very first domino over.
   2. THE INDUCTIVE STEP: If any ...

   Once those two facts are proven, the infinite chain reaction is guaranteed.

## Subtopic 2: The Standard Induction Proof Skeleton
   When the exam asks you to "Prove by induction on $n$", you must use a rigidly
   structured skeleton. If you miss the headings or forget to define your 
   predicate, examiners will aggressively ... before reading algebra.

### THE MANDATORY SKELETON:
   1. THE PREDICATE: Define what exactly $P(n)$ is.
   2. BASE CASE: Explicitly write `Base Case: n = [start value]`. Prove LHS = RHS.
   3. INDUCTIVE HYPOTHESIS: Explicitly write `Let k be arbitrary` and `Assume IH: P(k)`
   4. INDUCTIVE STEP: Explicitly write the target $P(k + 1)$. Then use a numbered
      chain of equalities to get there. You MUST cite the IH the exact moment you
      use it.
   5. CONCLUSION: A single wrap-up sentence.

## Subtopic 3: Algebraic Clean-up (The Engine)

### The Intuition
   The entire game of the Inductive Step is a jigsaw puzzle. You start with the
   expression for $k + 1$. Your only goal is to algebraically rip that 
   expression apart until you find the exact shape of $P(k)$ hiding inside it.

   Once you expose the $P(k)$ shape, you swap it out using your IH. The
   rest is just basic algebraic clean-up (expanding brackets, factoring)
   to make it match your $k + 1$ target.

## Subtopic 1: Strong Induction (The Safety Net)

### The Intuition
   Standard induction assumes you only need the immediately previous step
   (rung $k$) to prove the next step (rung $k + 1$). But what if your recursive
   algorithm splits a problem in half (like merge Sort) and calls itself on
   $k / 2$? The immediately preceding rung $k$ is completely useless.

   Strong induction gives you access to the entire staircase. You assume that 
   the property holds for ALL values up to $k$. When trying to prove $k + 1$,
   you can reach back and grab the facts for $k, k - 1, k / 2$, or any other
   smaller value you need.


### The Exam Skeleton (What Changes):
   The only thing that changes is the wording of your Inductive Hypothesis.
   - STANDARD IH: Let $k \geq 0$ be arbitrary. ASSUME IH: $P(k)$
   - STRONG IH: Let $k \geq 0$ be arbitrary. ASSUME IH: $P(m)$ HOLDS FOR ALL
     $m$ s.t. $0 \leq m \leq k$. (When you use the IH in your proof, you must
     explicitly state which value of $m$ you are using. e.g., "by IH with
     $m = k - 2$".)

## Subtopic 2: Recursive Function Equalities (The Code-to-Math Bridge)

### The Intuition
   You cannot do algebra on raw Kotlin or Haskell syntax. Before you can prove
   anything about a program, you must "compile" the code into strict 
   mathematical axioms.

   If an exam question gives you a recursive function, your very first step
   (often worth 1-2 marks on its own) is to write out the piecewise equation. 


THE VISUAL FRAMING: The Translation
   Imagine an exam question gives you this pseudocode:

```
function f(n):
  if n == 0 then return 1
  if n == 1 then return 2 
  else return f(n-1) + 2 * f(n-2)
```
   You immediately translate this into numbered "Given" equalities for your
   workspace:
      - (f.1) $f(0) = 1$
      - (f.2) $f(1) = 2$
      - (f.3) $f(n-1) + 2f(n-2) for all n >= 2$

## Subtopic 3: Recursion-to-Induction Translation (The Exam Engine)

### The Intuition
   This is the core loop of the exam... Examiners will give you a recursive
   function, and ask you to prove a closed-form mathematical property about it
   (e.g., "Prove that $f(n) = 2^n$")
   
   The structure is identical to Block 2, but instead of doing generic algebra,
   your Inductive Step is powered strictly by expanding your recursive 
   definitions.

### The Execution Strategy:
   1. TARGET $P(k+1)$: Write down the function call for $k + 1$
   2. Unroll: Use your numbered function equalities to exapnd $f(k + 1)$ one
      level deep. (e.g., $f(k + 1) = f(k) + 2f(k - 1)$ by f.3).
   3. INJECT IH: Now that you have smaller function calls ($f(k)$ and $f(k-1)$),
      use your Strong Inductive Hypothesis to instantly replace thoser function
      calls with their closed-form math equivalents.
   4. CLEAN-UP: Do the algebra until it matches the RHS of $P(k + 1)$.
