-- Proprioception in robotics is the "inner sense" that allows a robot to 
   monitor its own internal state, body awareness, and physical movements 
   without relying on cameras or external cameras. It is the mechanical 
   equivalent of the body awareness humans use to know where their limbs are 
   with their eyes closed.

KEY CONCEPTS
   - INTERNAL vs. EXTERNAL: Proprioception is strictly internal. It differs from
     extroception  (sesning the outside world via vision, lidar or touch).
   - REAL-TIME FEEDBACK: It provides a constant stream of data about the robot's
     mechanics, which... navigate uneven terrain, and adjust grips.

WHAT DO ROBOTS "SENSE" VIA PROPRIOCEPTION?
   Proprioceptive systems gather several core metrics about the robot's own 
   body:
   - JOINT ANGLES: 
   - KINEMATICS & SPEED:
   - TORQUE & FORCE: The amount of power a motor is exerting or the physical
     strain/load passing through structural joints.
   - ORIENTATION: Assessing the tilt or balance of the robot's chassis.

ESSENTIAL SENSORS
   To achieve this, robots use a variety of specialised internal sensors:
   - ENCODERS (Rotary/Linear): Mounted on motors or joints to track position
     and speed.
   - IMUs (Inertial Measurement Units): Gyroscopes and accelerometers that
     measure the robot's tilt, balance, and acceleration in 3D space.
   - CURRENT SENSORS: Measure the electrical current being drawn by a motor,
     which helps the system infer the amount of force or resistance it is 
     encountering.

WHY IT MATTERS
   - MOTOR CONTROL: It enables smooth, coordinated movement without needing
     to visually track every limb.
   - Dynamic Balancing: Legged robots (e.g. quadrupeds or humanoids) use 
     proprioception to instantly adapt to un expected drops or uneven terrain in
     real-time.
