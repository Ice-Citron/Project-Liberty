## Subtopic 1: Propositional Syntax (The Grammar)

### The Intuition
   Syntax is purely about the structure of the symbols, with absolutely no care
   for what they mean. Think of it like definining a strict recursive data type
   in Haskell or building an Abstract Syntax Tree (AST) in a compiler. A 
   sequence of symbols is only a "Well-Formed Formula" (WFF) if it perfectly 
   follows the recursive building rules. We saw a glimpse of this with binding
   strength earlier: syntax dictates the exact shape of the tree.

## Subtopic 2: Semantics (The Meaning)

### The Intuition      
   If syntax is building the AST, semantics is writing the evaluator function.
   We assign a "valuation" (a state where every variable is strictly True or 
   False) and evaluate the tree from the leaves up to the root.

   In exams, examiners don't jusst want you to calculate truth tables; they 
   want you to categorise the nature of the formula across all possible states. 
   Imagine testing the logic governing a rogue AI agent:

   1. VALID (Tautology): The formula evaluates to True in every possible 
      valuation. It is a universal truth of the system (e.g., $p \vee \neg p$).
      The agent's logic is ironclad.
   2. SATISFIABLE: The formula evaluates to True in at least one valuation. It
      is possible under certain conditions (e.g., $p \wedge q$). The agent can
      execute this path. (Note: Valid formulas are automatically satisfiable).
   3. CONTRADICTION (Unsatisfiable): The formula evaluates to False in every
      possible valuation (e.g., p ^ neg p). The logic is fundamentally broken. 


### The Visual Framing: The State Space Map
   A truth table is just a brute-force map of the entire state space. If an exam
   asks you to show a formula is a contradiction using "equivalence law" or 
   "direct argument", they are asking you to algebraically collapse that entire
   state space down to a single `False` without drawing the table.

## SUBTOPIC 3: Transation (English to Formulas)   

### The Intuition
   Natual language is messy and ambiguous; propositional logic is ruthless and 
   exact. Translation questions on the exam test your ability to spot English
   idoioms and map them to strict logical connectives. 


### The Translation Traps (Exam Focus)
   Examiners love to use specific English phrasing to trip you up. ... cheat
   sheet for the most common traps:

   - "A if B": B -> A
      - The trap: The order flips. The thing following "if" is always the premise
        (the left side of the arrow)
