4th method Proof Systems: Natural Deduction



We should be able to establish validity of argument by breaking it into smaller
arguments and showing the validity of these intermediate ones, i.e., construct a
proof.

A PROOF SYSTEM is a way of showing formulas to be valid by using purely 
syntactic rules--not using semantics at all. A computer algorithm should be able
to apply the rules.

There are many proof systems. We will focus on NATURAL DEDUCTION (ND), a system
invented by F. B. Fitch (and relate work of Gentzen); it is a major topic and
will take some time.

WHAT IS NATURAL DEDUCTION?

A formalisation of 'direct argument'.

Starting perhaps from some given facts--some formulas $\phi_1, \dots, phi_n$--
we use the rules of the system to reason towards a formula $\psi$.

(In practice, we reason forwards, backwards, and especially inwards).#


If we succeed, we can write:
$$\phi_1, \dots, \phi_n \vdash \psi$$


- $\phi_1, \dots, \phi_n$ are called PREMISES.
- $\psi$ is called a CONCLUSION.
- $\phi_1, \dots, \phi_n \vdash \psi$ is called a SEQUENT.



<<<<<<<<<<>>>>>>>>>>
ASSUMPTIONS IN NATURAL DEDUCTION

Proofs in Fitch's system are not based on axioms expressed in logical form, as 
in Hilbert's axiomatic proof theory.

They sometimes involve making "temporary" assumptions to prove a conclusion.

"Assume for the sake of argument..."


An ASSUMPTION is just a formula, but it's used in a special way.

We imagine a situation in which a formula is true. Then we derive some 
additional formulas that help us make progress towards the conclusion.

We need to be careful about how we use these assumptions. We will see examples 
of its use soon.

HOW NATURAL DEDUCTION WORKS

Deduction works by writing down intermediate formulas using INFERENCE RULES.
- State the set of premises $\phi_1, \dots, \phi_n$, and the conclusion.
- Intermediate formulas form the PROOF of $\psi$ from the givens
  $\phi_1, \dots, \phi_n$.
- Once established, they may be usable later.
- Each step of the proof is a VALID ARGUMENT.  





NATURAL DEDUCTION INFERENCE RULES

Mostly, there are two rules for each connective:
   - one for INTRODUCING it (bringing in a formula whose principal connective is
     the connective it introduces).
   - one for USING it (using a formula already present in a step of the proof,
     whose principal connective is the connective it wants to use)

1. THE CORE CONCEPT: THE PRINCIPLE OF EXPLOSION

The slide introduces a concept formally known as Ex Falso Quodilbet ("From
falsehood, anything follows").

   - THE RULE: If you encounter a contradiction (a statement that is logically
     impossible, like "It is raining AND it is NOT raining"), you can 
     immediately conclude ANYTHING YOU WANT.
   - THE SYNTAX: 
     $$p \wedge \neg p \vdash q$$
        - PREMISE: $p \wedge \neg p$ (A contradicton).
        - CONCLUSION: $q$ (Literally any random sentence).    


THE ANALOGY: THE "BROKEN GAME" Imagine a game of soccer. The rulebook says: "If
the ball goes into the net, you get 1 point." Now imagine a rule that says: "If
the ball turns into a dragon, you win the game instantly."#
   - Will the ball ever turn into a dragon? NO.
   - Is the rule valid? YES. It's in the rulebook. It just never triggers. But
     if it somehow did (in a broken reality), you would win.

THE EXAMPLE FROM THE SLIDE:
   - $p$: "Cows can fly."
   - Premise: $p \wedge \neg p$ ("Cows can fly AND Cows cannot fly.")
   - Conclusion: "I like going to the theatre."
   - LOGIC: The derivation is valid. The logic system does not care that the 
     conclusion has nothing to do with cows. It only cares that the premise is
     impossible.

2. THE SYMBOL FOR CONTRADICTION: BOTTOM ($\bot$)

In Natural Deduction, we don't usually write "Contradiction." We use a special
symbol called BOTTOM ($\bot$), which represents the value `False`.


THE RULE: BOTTOM ELIMINATION ($\bot E$)
This is the syntactic rule you will use in your proofs.

- SYNTAX:
   1. $\dots$
   2. $\bot$        (You derived a contradiction somehow)
   3. $\phi$        ($\bot E$ line 2)
- WHAT THIS MEANS: If you manage to write $\bot$ on a line, you can write ANY
  FORMULA YOU WANT on the next line.


TRACE EXAMPLE: Let's prove: $A, \neg A \vdash B$ (where $B$ is totally 
unrelated).
   1. $A$                       Premise
   2. $\neg A$                  Premise
   3. $\bot$                    $\neg E(1, 2)$      [Contradiction found!]
   4. $B                        $\bot E(3)$         [Explosion: Since we found Bottom, we can claim B]

3. THE "PHILOSOPHICAL" JUSTIFICATION (S2)
The lecturer tries to explain why we allow this "crazy" rule.

The "VACUOUS TRUTH" Argument: To prove an argument is invalid, you must find a
COUNTER-EXAMPLE: A situation where the Premises are TRUE and the Conclusion is
FALSE.
   - PREMISE: $\bot$ (Absolute Falsity).
   - QUESTION: is there a situation where $\bot$ is True?
   - ANSWER: No. By definition, no such situation exists.
   - RESULT: Since you can NEVER find a situation where the premise is True, you
     can NEVER find a Counter-Example. Therefore, the argument is technically 
     Valid.

LECTURER'S PARAPHRASE: "Suppose we are in a situaton in which $\bot$ is true...
BUT there is no such situation. So you can't dispute the crazy things that 
follow."

4. EXAM TIPS FOR CONTRADICTION
Based on typical Natural Deducation exams and this slide:

   1. DON'T FIGHT IT: Do not write on your exam paper "This doesn't make sense"
      or "Cows can't fly." If the syntax allows you to derive $\bot$, you can
      derive the moon is made of cheese. Just apply the rule $\bot E$.
   2. WHEN TO USE IT: Use this rule when you are stuck in a sub-proof (a box)
      and the thing you need to prove seems impossible to get.
         - Strategy: If you can find a contradiction anywhere in your current
           scope, you can turn that contradiction into the specific formula you
           need to close the box.
   3. NOTATION: The rule is often written as $\bot \vdash $\phi$ or `False E`.
      Check if your specific year.... we use $\bot E$

RULES FOR BOTTOM

The practical point is that there is no situation in which $\bot$ is true, so 
you can only prove $\bot$ under (contradictory) assumptions in a box in the 
middle of a bigger proof.

You often get such assumptions--e.g., in an argument by cases ($\vee E$).

From $\bot$ you can then deduce anything you want, so the box is no obstacle to
the overall goal. So $\bot$ is a useful formula to aim to prove.


- Note that $\bot I$ is the same rule as $\neg E$. (There are two names for this
  rule!)
- The rules for $\bot$ can actually be obtained from the other rules already
  explained.

4. HANDLING "NO PREMISES"
   - STRATEGY: If you are given a sequent with no premises (e.g., 
     $\vdash P \vee \neg P$), you MUST START BY MAKING AN ASSUMPTION.
   - TIP: You can assume whatever you want (e.g., assume the negation of what
     you're trying to prove to set up a contradiction).


5. SCOPE & BOXES (Crucial Warning)
   - SCOPE RULE: You CANNOT use a formula derived inside a box after that box
     has closed. It is "dead" outside the box.
   - EXCEPTION (NESTED BOXES): You CAN use formulas from an outer box inside an
     inner box (because the outer assumption is still active).
   - EXAM TIP: Be very careful with scope. Using a line from a closed box is a
     common error that invalidates the proof.


5. CONTRADICTION STRATEGY:
   - "FROM CONTRADICTION, ANYTHING FOLLOWS": The lecturer reiterated that 
     syntactically, if you derive $\bot$ (bottom), you can concluse any formula
     $\phi$. This is a powerful tool to escape dead ends in a proof.
   - TIP: If you are stuck, try to find a contradiction. Once you have $\bot$,
     you can immediately write down the formula you are trying to prove (using
     $\bot E$)


7. DOUBLE NEGATION

1. PRIMITIVE vs. DERIVED RULES (The "Shortcut" Concept)
THE CONCEPT:
   - PRIMITIVE RULES: These are the basic building blocks we defined earlier
     ($\wedge I$, $\wedge E$, $\rightarrow I$, $\rightarrow E$). They are the
     "atoms" of the proof system. They are independent--you cannot remove one
     without breaking the system.
   - DERIVED RULES: These are "shortcuts." They are complex formulas that we
     have already proven to be valid using primitive rules. Because we known 
     they work, we can just use them as a single step instead of writing out the 
     10-line proof every time.

   ANALOGY:
      - PRIMITIVE RULES: Writing code in Assembly Language (moving individual
        bits in memory).
      - DERIVED RULES: Calling a Python function. The function is built out of
        Assembly eventually, but you just type `sort()` to save time.


   LECTURER'S EXAM TIP:
      "Derived rules are NOT neccessary, but they do help to speed proofs up...
      Some `purists` only allow for "primitive rules"... We don't here!"

      TAKEAWAY: You are ALLOWED and ENCOURAGED to use these shortcuts in the
         exam to save time.

This is the most famous derived rule in Classical logic.

SYNTAX: $\phi \vee \neg \phi$
MEANING: "For any statement $\phi$, either it is True, or it is NOT True. There 
   is no third option."

HOW IT WORKS AS A LEMMA: Usually, proving $P \vee \neg P$ from scratch takes
about 10-15 lines of nested boxes (as seen in the "no premises" example in the 
previous lecture section). Instead of writing that out, you can just write.


$$\begin{array}{l|ll}
\vdots & \vdots \\
n & p \vee \neg p & \text{LEM} \\
\vdots & \vdots
\end{array}$$


   WHEN TO USE IT: Use LEM when you are stuck and don't know where to start. It
      allows you to split the universe into two cases:
      1. CASE 1: Assume $p$ is True.
      2. CASE 2: Assume $p$ is Fale ($\neg p$). If you can reach your conclusion
         in both cases (using $\vee E$), then the conclusion is valid.




<<<<<<<<<>>>>>>>>>
3. MODUS TOLLENS (MT)
This is the reverse of Modus Ponens. It is a "destruction" rule for implication.

THE LOGIC: "If it rains, the grass is wet" ($P \rightarrow Q$) is true, and 
"The grass is dry" ($\neg Q$) is true, then "It is NOT raining." ($\neg P$).


THE TRACE:
$$\begin{array}{l|ll}
1 & \phi \rightarrow \psi & \text{premise} \\
2 & \neg \psi & \text{premise} \\
3 & \neg \phi & \text{MT(1, 2)}
\end{array}$$

WHY IT'S A DERIVED RULE: If you didn't have the MT shortcut, you would have to 
prove it via contradiction like this:
   1. Assume $\phi$
   2. Apply $\rightarrow E$ to get $\phi$
   3. Crash $\psi$ against $\neg \psi$ to get $\bot$.
   4. Conclude $\neg \phi$. MT DOES ALL 4 STEPS IN 1 LINE.




4. DOUBLE NEGATION (DN)
This allows you to toggle between "positive" and "double negative" forms.


THE LOGIC: $\neg \neg \phi \Longleftrightarrow \phi$

THE TRACE:
$$\begin{array}{l|ll}
1 & \neg \neg p & \text{premise} \\
2 & p & \text{DN(1)}
\end{array}$$

OR
$$\begin{array}{l|ll}
1 & p & \text{premise} \\
2 & \neg \neg p & \text{DN(1)}
\end{array}$$


LECTURER'S TIP: This is crucial for PROOF BY CONTRADICTION. Often, a 
contradiction proofs ends with $\neg \neg \phi$. You need this rule to clean
it up to just $\phi$.






<<<<<<<<<>>>>>>>>>
5. PROOF BY CONTRADICTION (PC) / Reductio ad Absurdium
This is a strategy (often treated as a rule) where you assume the OPPOSITE of
what you want, and show it destroys the universe ($\bot$).


THE TRACE (Classical Logic): 
To prove $\phi$:
$$\begin{array}{l|ll}
\vdots & \vdots \\
n & \neg \phi & \text{asm (Assume the opposite)} \\
\vdots & \vdots & \text{(Work until you find a crash)} \\
m & \bot & \neg E(\dots) \\
\hline
m+1 & \phi & \text{PC (or RAA) n-m}
\end{array}$$

NOTE: This is different from $\bot E$:
   - $\bot E$ (EXPLOSION): "I found a contradiction, so I can claim random 
     nonsense." ($A, \neg A \vdash Unicorns$)
   - PC (CONTRADICTION): "Assuming $\neg \phi$ led to a crash, so $\phi$
     must be true."  










6. DEDUCTION WITH LEMMAS#
The lecturer introduces the concept of a LEMMA. In Math, a lemma is a "helper 
theorem".

HOW TO USE IT IN NATURAL DEDUCTION (S7): You don't need to prove the helper 
theorem inside your box. You can just invoke it.


EXAMPLE TRACE using LEM as a LEMMA: Seq: $\vdash (p \rightarrow q) \vee (q \rightarrow r)$
$$\begin{array}{l|ll}
1 & q \vee \neg q & \text{Lemma (LEM)} \\
2 & \quad q & \text{asm (Case 1 of LEM)} \\
\vdots & \quad \dots & \text{...derive conclusion...} \\
k & \quad (p \to q) \vee (q \to r) & \text{...} \\
\cline{2-2}
k+1 & \quad \neg q & \text{asm (Case 2 of LEM)} \\
\vdots & \quad \dots & \text{...derive conclusion...} \\
m & \quad (p \to q) \vee (q \to r) & \text{...} \\
\cline{2-2}
m+1 & (p \to q) \vee (q \to r) & \vee E(1, 2-k, k+1-m)
\end{array}$$

DERIVED RULES vs. PRIMITIVE RULES

The deduction we saw in the last slide (for $p \vee \neg p$) is very handy in 
(other) proofs.

It has a special name: ... law of excluded middle (LEM).

If you haven't got $\phi$, you've got $\neg \phi$. This law is true of classical
logic, but not of some other logics.

We call it a DERIVED RULE (DL) because we can infer it from the `primitive` 
rules $\vee I, \neg I,$ and $\neg E$.



Derived rules are NOT necessary, but they do help to speed proofs up.

Some `purists` only allow for `primitive rules` (one that are independent of 
each other) in proofs. We don't here!

There are other common derived rules.

MORE on NATURAL DEDUCTION


With rules for $\wedge, \rightarrow, \vee$ we can already do quite a lot. (We
still need rules for $\neg, \bot, \top$).

The rules are natural... they are simple and self-evident inference rules based

DEFINITION 0.1 (Natural Deduction Proof)
   ...

You can read $\phi_1, \dots, \phi_n \vdash \psi$ as '$\psi$ is provable from
$\phi_1 \dots \phi_n$'.

$\vdash \psi$ means we can prove $\psi$ with no premises at all. We then say 
that $\psi$ is a THEOREM (of natural deduction).

Do not confuse '$\vdash$' with $\models$.
   $\vdash$ is syntactic and involves proofs.
   $\models$ is semantic and involves situations.




ADVICE ON NATURAL DEDUCTION
   1. Learn the rules. Practice!
   2. Think of a direct argument to prove what you want. Then translate it into
      natural deduction.
   3. If really stuck in proving $\phi$, it can help to:
      - Assume $\neg \phi$ and prove $\bot$. ('Reductio de absurdum', or 'proof
        by contradiction', PC).
      - Use the lemma $\psi \vee \neg \psi$, for suitable $\psi$. It amounts to
        arguing by cases.

1. THE STRICT RULE of "THE BOX" (S18)
   The lecturer emphasizes that the BOX is not just a drawing; it is a strict
   logical boundary.
   - THE "VEGAS" RULE: What happens in the box, stays in the box.
      - ASSUMPTION: Every box must start with an assumption (`asm`) on the very
        first line.
      - EXPORT RESTRICTIONS: You cannot pull a formula out of a box and use it
        later. It is "dead" once the box closes.
   - HOW TO EXIT: The only valid ways to leave a box are by using rules that
     specifically close scopes:
        1. $\rightarrow I$: "I assumed A and got B, so $A \rightarrow B$."
        2. $\neg I$: "I assumed $A$ and got $\bot$, so $\neg A$."
        3. $\vee E$: "I checked both cases ($A$ and $B$) and got the same 
           result."
        4. $PC$: "I assumed $\neg A$ and got $\bot$, so $A$."

   ANALOGY: Think of a Box like a FUNCTION in programming.
      - The ASSUMPTION is the specific input argument.
      - The formulas inside are LOCAL VARIABLES.
      - You cannot access a local variable from outside the function. You can
        only access the RETURN VALUE (the conclusion).




...
   The ONLY ways of exporting information from a box are by the rules 
   $\rightarrow I, \vee E, \neg I, and PC$ (and also $\exists E, \forall I$ in
   predicate logic). The first line after a box MUST be justified by one of 
   these.

   No formula inside a box can be used outside, except via the above rule.

   You can check these conditions for yourself. If your box doesn't meet them,
   your proof is wrong.

- Hilbert's axiomatic proof theory, developed by David Hilbert in the 1920s, is
  a foundational, formal approach to mathematics aiming to establish the
  consistency of mathematical theories using finite methods. It structures logic
  using few axioms and a single inference rule (modus ponens) to derive theorems
  , aiming to prove the consistency and completeness of arithmetic and geometry.

- Supposition = a belief held without proof or certain knowledge; an assumption
  or hypothesis.  


- In Mathematics, a SUPPOSITION is an assumption, hypothesis, or "what if" 
  scenario used as a problem-solving tool, particularly for solving complex word
  problems. It involves temporarily assuming an extreme situation to simplify a
  problem, calculate a difference, and then work backward to the actual 
  solution.


- The law of excluded middle (LEM) is a fundamental axiom in classical 
  mathematics stating that for any proposition $P$, either $P$ is true or its
  negation ($\neg P$) is true, formulated as $P \vee \neg P$. It asserts that
  there is no intermediate truth value between true and false, making it central
  to PROOF BY CONTRADICTION (reductio ad absurdum).
