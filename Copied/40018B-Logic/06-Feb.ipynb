1. THE PROBLEM: "IT DEPENDS" (S44-45)

   - ANSWER: It depends on what $x$ is!
      - If $x$ is the object `heron` ($o_10$), it might be True.
      - If $x$ is the object `clyde` ($o_7$), it might be False.


   THE LOGIC ISSUE: We cannot assign a Truth Value (True/False) to a formula 
   with free variables unless we fix the meaning of those variables first.




   <<<<>>>>
   2. FREE vs. BOUND VARIABLES (S46)

   To fix this, we first need to identify which variables are "Free" 
   (Uncontrolled) and which are "Bound" (controlled).
      - BOUND VARIABLE: An occurrence of a variable that is inside the scope of
        a quantifier ($\forall x$ or $\exists x$).
      - FREE VARIABLE: An occurrence that is NOT inside any quantifier scope.

   EXAMPLE TRACE:
      $$\forall x (R(x, y) \wedge R(y, z) \rightarrow \exists z (S(x, z) \wedge z = y))$$     

   Let's check every variable:
   1. $x$:
      - Left side ($R(x, y)$): Inside $\forall x$. BOUND.
      - Right side ($S(x, z)$): Still inside $\forall x$. BOUND.

   2. $y$:
      - Everywhere: There is no $\forall y$ or $\exists y$. FREE.

   3. $z$:
      - Left side ($R(y, z)$): Outside the $\exists z$ on the right. FREE.
      - Right side ($S(x, z)$): Inside $\exists z$. BOUND.


EXAM TIP: A variable can be both free and bnnound in the same formula (like $z$
above). You must check each specific occurence individually.

3. THE SOLUTION: ASSIGNMENTS (S49)
   Since the Structure ($M$) doesn't tell us what free variables mean, we need
   an extra component called an ASSIGNMENT ($h$).

   DEFINITION: An assignment $h$ is a function that maps every variable 
   ($x, y, z \dots$) to a specific obnject in the Domain ($D$).
      - $h(x) = o_5$ (The assignment "pretends" $x$ is Object 5)
      - $h(y) = o_3$ (The assignment "pretends" $y$ is Object 3$)
      
   THE COMPLETE SITUATION: To evaluate a formula, we now need a pair: $(M, h)$.
      - $M$ (Structure): Gives meaning to Constants (`frank`), Predicates
        (`Human`), and Functions (`father_of`).
      - $h$ (Assignment): Gives meaning to Variables $(x, y, z)$.

4. EVALUATING TERM WITH ASSIGNMENTS (S50-52)

   We define the "Value of term $t$" in model $M$ under assignment $h$, written
   as $|t|^{h}_M$.

   
   THE RULES:
      1. If $t$ is a Constant ($c$): Look at the Structure $M$. 
         ($|frank|^h_M = o_3$).
      2. If $t$ is a Variable ($x$): Look at the Assignment $h$.
         ($|x|^h_M = h(x)$).
      3. If $t$ is a Function ($f(t_1)$): Evaluate the input $t_1$, then apply
         the function rule from $M$.


   EXAMPLE TRACE (S51): 
      Term: `fathher_of(x)`

      - SCENARIO: $h(x) = o_3$ (`frank`)               
      - STEP 1: Evaluate $x$
         - $|x|^h_M = o_3$
      - STEP 2: Apply function `father_of` to $o_3$
         - Looking at $M$'s definition, $o_3 \mapsto o_1$
      - RESULT: $|father_of(x)|^h_M = o_1$

5. SEMANTICS OF QUANTIFIER-FREE FORMULAS (S53)
   Now we can finally evaluate logical truth formally.

   SYMBOL: $M, h \models \phi$ ("In model $M$, with assignment $h$, formula 
           $\phi$ is True").

   THE FORMAL RULES:
      1. ATOMIC FORMULA ($P(t_1, t_2)$):
         - Calculate the values of terms $t_1$ and $t_2$ (using $h$ if needed).
         - Check if that pair of objects is in the set $I(P)$.
      2. EQUALITY ($t_1 = t_2$):
         - Calculate $|t_1|$ and $|t_2|$.
         - Check if they are the EXACT SAME OBJECT in the domain.
      3. CONNECTIVES ($\wedge, \vee, \neg$):
         - Standard boolean logic. $M, h \models A \wedge B$ iff both $A$ and 
           $B$ are true under $h$.

   EXAMPLE: Formula: `Bought(x, y)`
      - ASSIGNMENT $h$: Let $h(x) = o_4$ (`susan`), $h(y) = o_10$ (`heron`)
      - CHECK: Is ($o_4, o_10$) in the `Bought` list?
         - YES.
      - RESULT: $M, h \models Bought(x, y)$

   NOTE: If we changed $h$ so that $h(y) = o_1$, the exact same formula would
      become FALSE.

1. THE PROBLEM: "IT DEPENDS" (S55-57)
   We solved the "Free Variable" problem using assignments ($h$). But how do we
   check $\exists x \phi$ or $\forall x \phi$? The variable $x$ inside these
   formulas is controlled by the quantifier, not by our initial assignments $h$.


   THE SOLUTION: We have to "tweak" our assignment.
   - TO CHECK $\exists x \phi$: We need to find AT LEAST ONE modified assignment
     $g$ (where we change only the value of $x$) that makes $\phi$ True.
   - TO CHECK $\forall x \phi$: We need to ensure that EVERY POSSIBLE modified
     assignment $g$ (where we change only $x$) makes $\phi$ True.


   FORMAL SYNTAX ($g =_x h$): This notation means "Assignment $g$ is exactly the
   same as assignment $h$, EXCEPT possibly for the value of variable $x$."
      - Example: If $h(x) = o_1, h(y) = o_2$
      - And $g(x) = o_5, g(y) = o_2$.
      - Then $g =_x h$ is TRUE (only $x$ changed).

1. Why CAN'T we write $\phi(x)$? (The "PRIVATE VARIABLE" Rule)

   In the formula $\forall x (R(x, y) \rightarrow \exists y S(y, z))$, the 
   variable $x$ is BOUND.
      - LOGIC VIEW: $x$ is "trapped" inside the quantifier $\forall x$. It is an
        internal iterator, not an external input.
      - CODING VIEW: Think of $\phi$ as a function definition.  
         - FREE VARIABLES ($y, z$): These are PARAMETERS passed in from the
           outside.
         - BOUND VARIABLES ($x$): This is a LOCAL VARIABLE declared inside the
           function.

   Writing $\phi(x)$ would be like trying to pass a local variable as an
   argument:

   ```cpp
   // The Logic Formula: forall x...
   void Phi(int y, int z) {         // We define inputs y and z
       for (int x : Universe) {     // <-- x is defined internally here
           if (R(x, y)) ...
       }
   }

   // Writing Phi(x) is like trying to call:
   Phi(x_value, ...);
       // Error: The function expects `y` and `z`. `x` is just an implementation
       // detail inside.
   ```          
   If you wrote $\phi(x)$, you are claiming "I need an external value for $x$."#
   But the formula says, "No thanks, i create my own $x$ inside the loop."



2. HOW DOES THE ORDERING WORK? ($\phi(y, z)$ vs. $\phi(z, y)$)

   You asked: "how would you normally write it then?? like cpp style?"

   It works exactly like a FUNCTION PROTOTYPE (positional arguments), not named
   arguments.

   When the book writes "Let $\phi(y, z)$ be...", it is DECLARING THE SIGNATURE.
      - DECLARATION: $\phi(y, z)$
         - This sets the contract: "Argument 1 maps to $y$. Argument 2 maps to 
           $z$."
      - USAGE: $\phi(Alice, Bob)$
         - Logic engine executes: replace free $y$ with $Alice$, replace free 
           $z$ with Bob.

           
   CAN YOU REORDER THEM? Yes!  You declare the signature however you want, just
   like in code.

      - OPTION A: Define $\phi (y, z)$ ...
      - OPTION B: Define $\phi (z, y)$ ...

FACT 1
   Given a formula $\phi$, whether or not $M, h \models \phi$ only depends on
   $\phi(x)$ for those variables $x$ that occur free in $\phi$.



So, for a formula $\phi(x_1, \dots, x_n)$, if $h(x_1) = a_1, \dots, h(x_n) = a_n$
, it's OK to write $M \models \phi(a_1, \dots, a_n)$ instead of 
$M, h \models \phi$.

   - Suppose we are explicitly given a formula $\phi(y, z)$ such as
      $\forall x (R(x, y) \rightarrow \exists y S(y, z))$

     If $h(y) = a, h(z) = b$, say, we can write:
       $M \models \phi(a, b)$ or $M \models \forall x (R(x, a) \rightarrow ...)$

     instead of $M, h \models \phi$, Note: only the FREE occurrences of $y$ in

- In FOL, bound variables are constrained by quantifiers ($\forall x$ or
  $\exists x$) within a formula, acting as local placeholders whose specific
  values do not affect the formula's overall truth value.

  Conversely, free variables are not within a quantifier's scope, representing
  unspecified, open-ended values that determine the truth of the formula.



- In FOL, VARIABLES ARE NOT PART OF THE MODEL ITSELF, but are placeholders
  within formulas that take values from the model's domain. The model 
  ($M =< D, I>$) defined the domain of objects ($D$) and interpretations ($I$),
  while variables in formulas are bound by quantifiers ($\forall, \exists$),
  or assigned values from $D$ to evaluate truth.  




- Variable Shadowing = Variable Shadowing occurs when a variable declared within
  a specific scope (e.g., a function, `if` block, or inner class) has the same
  name as a variable in an outer scope, causing the inner variable to "shadow"
  or hide the outer one. References inside the inner scope will use the inner
  variable, temporarily making the outer variable inaccessible.
