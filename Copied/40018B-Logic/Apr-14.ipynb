## Subtopic 1: Natural Deduction Rules (The Core Toolkit)

### The Intuition
   Natural Deducition is a game of scaffolding. You are no longer mapping out
   "True/False" states. Instead, you are given a set of starting materials
   (premises) and a target blueprint (the conclusion). You must bridge the gap
   using microscopic, undeniable steps.

   Every logical operator in ND has exactly two types of rules:
      1. INTRODUCTION (I) Rules: How to build the operator.
      2. ELIMINTATION (E) Rules: How to tear it apart to extract its contents.

## THE VISUAL FRAMING: The Assumption Box
   The most critical visual concept in ND is the "Assumption Box" (or scope).
   Whenever you make a temporary assumption, imagine drawing a strict box around
   the subsequent lines. Once you reach your sub-goal, you close the box (discharge
   the assumption). YOU CAN NEVER REACH BACK 9INSIDE A CLOSED BOX TO GRAB A 
   VARIABLE. 

   The Essential Rules:
      - `^I` (And Intro): If you have $A$ on line 2, and $B$ on line 5, you can
        write $A \wedge B$.
      - `^E` (And Elim): If you have A ^ B, you can extract $A$ (or $B$) on its
        own.
      - `->I` (Implies Intro): Requires a box. Assume $A$. Work until you prove 
        $B$. Close the box. You now have $A -> B$.
      - `->E` (Implies Elim / Modus Ponens): If you have $A -> B$, and you also
        have $A$, you can extract $B$.
      - `vI` (Or Intro): If you have $A$, you can glue an OR onto it out of 
        nowhere: $A \vee anything$.
      - `vE` (Or Elim / Case Split): Requires two boxes. If you have $A \vee B$,
        you must prove your target $C$ twice. Box 1: Assume $A$, prove $C$. 
        Box 2: Assume $B$, prove $C$. Therefore, $C$ holds.
      - `¬I` (Not Intro): Requires a box. Assume $A$. If this causes a 
        contradiction ($\bot$), close the box and conclude $\neg A$.
      - `¬E` (Not Elim): If you have $A$ and $\neg A$ on different lines, they
        clash. Write $\bot$.  

## Subtopic 2: Derived Rules (The Shortcuts)

### The Intuition
   Derived rules are combinations of core rules that happen so frequently that
   the examiners let you use them as a single-step shortcuts to save time. You
   could prove them from scratch, but you don't have to.

### The Exam MVPs:
   1. Modus Tollens (MT): The reverse of $\to E$. if you have $A \to B$, and
      you know $\neg B$ is true, you can immediately conclude $\neg A$.
   2. DOUBLE NEGATION (DN): If you have ¬¬A, you can instantly drop both to get
      A.
   3. REDUCTIO AD ABSURDUM (RAA): The classical logic powerhouse. If you want to
      prove $A$ (a positive statement), you assume $\neg A$. If that leads to
      $\bot$, you  close the box and conclude $A$. (Note: This is different from
      $\neg I$, which is used to prove a negative statement).
   4. LAW OF EXCLUDED MIDDLE (LEM): You can inject $A \vee \neg A$ into our proof
      out of thin air at any time.

## SUBTOPIC 3: Proof Skeletons and Common Pitfalls

### The Intuition
   The biggest mistake students make in ND exams is starting at line 1 and just
   blindly applying rules downwards. You must build the skeleton from the 
   outside in. Look at the main connective of your target conclusion to know
   exactly which rule you need at the very end.

### The Exam Skeleton:
   - Target is $A \to B$:
      Immediately write "Assume A" at the top of your workspace. Write "B" at
      the bottom. Your final line will be $A \to B$ by `->I`.
   - Target is $\neg A$:
      Immediately write "Assume A". Your sub-target at the bottom is $\bot$. 
      Final line is $\neg A$ by `¬I`.
   - Target is $A \vee B$ (The tricky one):
      You usually cannot prove this directly. Youmust use RAA. Write "Assume
      $\neg(A \vee B)$ at the top. Target $\bot$ at the bottom."
   - Premise contains $A \vee B$:
      Immediately set up a `vE` skeleton. You will need two sub-proofs: "Assume
      $A$" and "Assume $B$".

## The Exam Pitfalls (What Loses Marks):
   1. VAGUE JUSTIFICATIONS: Our new strict formatting rules prevent this. Every
      line must have the exact rule name and the exact line numbers it relies on
      . e.g., `10. B (->E 4, 9)`
   2. LEAKING SCOPE: Citing a line number trapped inside an assumption box
      you have already closed.
   3. CONFUSING $\bot E$ witth RAA: If you have derived $\bot$ (Principle of
      Explosion) to instantly deduce any arbitrary formula $X$ you want. RAA
      specifically allows you to discharge an assumption of ¬X to deduce X.

## Subtopic 1: First-Order Signatures (The Vocabulary)

### The Intuition
   Before you can evaluate if a statement is true, you need to know what words 
   yoyu are legally allowed to use. A SIGNATURE ($L$) is exactly like declaring
   an API or importing a library. It is strictly defined list of symbols, but
   it does not contain any actual data or meaning yet. 

   THE THREE TYPES OF SYMBOLS:
      1. CONSTANTS: Names for specific, fixed individual objects.
         - Example: `saka`, `arsenal`, `zero`.
      2. FUNCTION SYMBOLS: Operations that take one or more objects as input
         and return a new object as output.
         - Example: `father_of(x)`, `add(x,y)`.
         - Exam Note: Every symbol has an "arity" (number of inputs). `father_of`
           is unary (1-ary)' `add` is binary (2-ary).
      3. PREDICATE (Relation) SYMBOLS: Tests that take one or more objects as 
         input and return a Boolean (True/False).
         - Example: `is_manager(x)`, `plays_for(x, y)`
         - Exam Note: Unary predicate act like "types" or "properties" (e.g., 
           $x$ is a manager. Binary predicates act like "relationships") (e.g.,
           $x$ plays for $y$).

## Subtopic 2: Structures, Interpretations, and Domains (The Database)

### The Intuition
   If the Signature is the API declaration, the STRUCTURE ($M$) is the actual
   database backend. It gives concrete meaning to the empty vocabulary.

   A structure is always defined as a pair: $M = (\mathbf{D, I})$.


1. The Domain ($\mathbf{D}$):
   This is your universe of discourse. It is the raw set of "things" that
   actually exist in this specific world. In exams, Imperial usually abstracts 
   these as raw objects like $o_1, o_2, o_3$.

   - Example: $\mathbf{D} = {o_1, o_2, o_3, o_4}$


2. THE INTERPRETATION FUNCTION ($\mathbf{I}$):
   This is the mapping engine. It takes the abstract symbols from your Signature
   and binds them to the concrete objects in your Domain.

   - INTERPRETING CONSTANTS: Maps a name to exactly one object.
     $\mathbf{I}(saka) = o_1$
   - Interpreting Unary Predicates: Maps a porperty to a subset of the domain (a 
     list of all objects that have that property).
     $\mathbf{I}(player) = {o_1, o_2}$
   - INTERPRETING BINARY PREDICATES: Maps a relationship to a set of tuples 
     (pairs).
     $\mathbf{I}(plays_for) = {(o_1, o_4), (o_2, o_5)}$
     (Read as: o1 plays for o4, and o2 plays for o5).

## The Intuition: What is "Soundness"?
   In logic, a proof system is SOUND if it never lies to you.

   Formally, it means: IF YOU CAN PROVE IT USING THE RULES ($\vdash$), IT IS 
   ACTUALLY LOGICALLY TRUE ($\vDash$). For a system to be sound, every single 
   rule i it must be TRUTH-PRESERVING. If you feed the rule True premises, it 
   must be mathematically impossible for it to spit out a False conclusion. If 
   even one rule can take True premises and generate a False conclusion, the 
   entire system is "broken" (not sound).

## The Visual Framing: Model Diagrams

### The Intuition
   Imperial examiners rarely make you parse large sets of tuples in your head.
   Instead, they draw the Structure $M$ as a graph. You must be able to 
   translate between the visual graph and the mathematical sets instantly.

### How to Read the Exam Diagrams:

   - LABELS: If a node has a word next to it (like `car`), this is a CONSTANT
     mapping to that node.

## Subtopic 1: Quantifier Scope && Formation Trees (The Grammar)

### The Intuition
   Quantifiers (\forall, \exists) act exactly like variable declarations in
   programming (like a `for` loop or a `let` binding). The "scope" of a 
   quantifier is the block of code it controls. If a variable fails inside the
   scope of a matching quantifier, it is BOUND. If it is floating around without
   a controlling quantifier, it is FREE.

### THE VISUAL FRAMING: The FOL Formation Tree
   Draw a formation tree in FOL is just like Propositional Logic, but quantifiers
   act as single-branch nodes (like the `¬` operator).

   1. Read the formula left to right to find the main connective or outermost
      quantifier.
   2. If a quantifier is attached to a large bracket block, that entire block is
      its child node.
   3. To determine if a variable leaf is bound or free, trace your finger straight
      up the tree. If you hit a matching quantifier (e.g., you trace from an `x`
      leaf up to a `\forall x` node), it is BOUND. If you reach the root without
      hitting one, it is FREE.


### THE EXAM TRAP: Order Matters!
   Examiners lvoe testing your intuition on nested quantifiers.
   - $\forall x \exists y R(x,y)$: "For every x, there is some y." (Every person
     has a mother. The $y$ an be different for each $x$).
   - $\exists y \forall x R(x, y)$: "There is a specific $y$ s.t. for all $x$..."
     (There is one single mother of every person on Earth. The $y$ is fixed).

## Subtopic 2: FOL Translation (The Language Map)

### The Intuition
   Translating from English to FOL is an exercise in pattern matching. You must
   pair the right quantifier with the right logical connective. Mixing them up
    is the #1 wa to lose marrks. 

### The Golden Rules of Translation:
   1. "All Xs are Ys" --> ALWAYS use \forall with --> (implies)
      - Formula: $\forall x (X(x) \to Y(x))$
      - Why: "If you are an X, then you must be a Y." If you accidentally use 
        $\wedge$ here ($\forall x (X(x) \wedge Y(x))$), you are climing that
        everything in the universe is both an X and a Y, which is absurd.

   2. "Some Xs are Ys" --> ALWAYS use $\exists$ with $\wedge$ (AND).
      - Formula: $\exists x (X(x) \wedge Y(x))$
      - Why: "There is at least one thing that is both an X and a Y." If you
        accidentally use $\to$ here ($\exists x (X(x) \to Y(x))$) , it is 
        trivially (and uselessly) true as long as there is one thing in the
        universe that is not an X.

   3. "No Xs are Ys" --? Two valid ways to write it.
      - $\forall x (X(x) \to \neg Y(x))$ ("If you are an X, you are not a Y.")
      - $\neg \exists x (X(x) \wedge Y(x))$ ("There does not exist a thing that
        is both X and Y.")

## Subtopic 3: Model Checking (The Database Query)

### The Intuition
   In Block 2, we looked at Structures ($M$) and Model diagrams (nodes and arrows)
   . Model checking asks: Is this FOL formula True or False inside this specific
   Structure? Treat the formula like a SQL query, and the structure like the 
   database. You must execute the query in your head.

### The Execution Strategy
   - EVALUATING $\forall x$ (The AND-Loop): You must test every single object
     $(o1, o2, ...)$ in the domain. If even one object fails the condition, the
     whole statement is False.
   - EVALUATING $\exists x$ (The OR-Loop): You test objects one by one. The
     moment you find one object that satisfies the condition, you stop. The
     whole statement is TRUE. If you check them all and none work, it is False.  
   - EVALUATING NESTED ($\forall x \exists y$) oN A DIAGRAM:
      1. Pick an object to be $x$.
      2. Look for an arrow (the relation) going outward to any $y$ that fits the
         rule.
      3. If you find one, that $x$ passes.
      4. Move to the next $x$ and repeat. Every node must pass.

## SUBTOPIC 1: The Four Quantifier Rules (The Upgrades)

### The Intuition
   In Propositional Logic, you dealt with static blocks. In FOL Natural
   Deduction, you have to "unpack" quantified statements, do your logical work
   on specific objects, and then "repack" them. 

   You only have four new rules to learn, but two of them are simple "one-liners"
   , while the other two require strict scoping rules.

### The "Easy" Rules (No Box Required):
   1. $\forall E$ (Universal Elimination): If you know a rule applies to 
      absolutely everything ($\forall x P(x)$), you can instantly apply it to 
      any specific constant or variable you want. 

      - Visual: $\forall x mortal(x) \to mortal(Socrates)$

   2. $\exists I$ (Existential Introduction): If you have proven a property for
      a specific object (e.g., $P(c)$), ou can instantly generalise it to say
      "Well, there exsits something with this property" ($\exists x P(x)$).

### The "Heavy" Rules (Require Scoping/Boxes):
   3. $\forall I$ (Universal Introduction): How do you prove something is true
      for everyone? You pick a completely random, anonymous person. If you can
      prove it for this arbitrary person--without relying on any special facts 
      about them--then it must be true for everyone. 
       * Exam rule: You must state "Let $x_0$ be arbitrary." You work with $x_0$
         . At the end, you attach the $\forall$ and swap $x_0$ back to $x$.

   4. $\exists E$ (Existential Elimination): If you know "Someone stole the 
      cookie" ($\exists x S(x)$), how do you use this fact? You must open an
      Assumption Box and assign a placeholder name to the culprit, like "Let $c$
      be the thief." You do your work inside the box.
       * Exam rule: The conclusion you pull out of this box cannot contain the
         placeholder name $c$. You must leave the fake name behind in the box.

## Subtopic 2: The "Freshness" Constraint (The Ultimate Exam Trap)

### The Intuition
   When you use $\forall I$ (arbirary variable) or $\exists E$ (placeholder 
   witness), the name you choose must be FRESH. This means the symbol cannot 
   appear anywhere in your premises, your target conclusion, or any currently
   open assumption boxes.

### Why it matters
   Imagine you are trying to prove a theorem about numbers.
   - Premise: $Even(2)$
   - You also know $\exists x Odd(x)$ (Someone is odd).
   - If you use $\exists E$ and say "Let 2 be the odd number", your logic 
     immediately explodes. You assigned the placeholder to an object that 
     already has existing baggage.
   - The Fix: Always pick a completely new letter like $a, b, c,$ or $x_0, y_0$
     when opening these specific boxes.

## Subtopic 3: The "FOL Sandwich" (Exam Skeletons)

### The Intuition
   When looking at recent papaers ... the sequents look terrifying because they 
   mix nested quantifiers with propositional arrows.

   The secret is that almost every FOL ND proof is structured like a sandwich:

   1. THE TOP BUN (Unpacking): Use $\forall E$ and $\exists E$ to strip away all
      the quantifiers from your premises. Turn them into statements about 
      specific dummy variables (like x_0 and y_0).

   2. THE MEAT (Propositional Logic): Once the quantifiers are gone, you are just
      doing standard Block 1 logic. You use $\to E$, $\neg I$, and case splits
      on the raw predicates.

   3. THE BOTTOM BUN (Repacking): Once you have reached your raw target conclusion
      , use $\exists I$ and to wrap the quantifiers back around the variables in
      the exact order the final conclusion requires.
