## Subtopic 1: Natural Deduction Rules (The Core Toolkit)

### The Intuition
   Natural Deducition is a game of scaffolding. You are no longer mapping out
   "True/False" states. Instead, you are given a set of starting materials
   (premises) and a target blueprint (the conclusion). You must bridge the gap
   using microscopic, undeniable steps.

   Every logical operator in ND has exactly two types of rules:
      1. INTRODUCTION (I) Rules: How to build the operator.
      2. ELIMINTATION (E) Rules: How to tear it apart to extract its contents.

## THE VISUAL FRAMING: The Assumption Box
   The most critical visual concept in ND is the "Assumption Box" (or scope).
   Whenever you make a temporary assumption, imagine drawing a strict box around
   the subsequent lines. Once you reach your sub-goal, you close the box (discharge
   the assumption). YOU CAN NEVER REACH BACK 9INSIDE A CLOSED BOX TO GRAB A 
   VARIABLE. 

   The Essential Rules:
      - `^I` (And Intro): If you have $A$ on line 2, and $B$ on line 5, you can
        write $A \wedge B$.
      - `^E` (And Elim): If you have A ^ B, you can extract $A$ (or $B$) on its
        own.
      - `->I` (Implies Intro): Requires a box. Assume $A$. Work until you prove 
        $B$. Close the box. You now have $A -> B$.
      - `->E` (Implies Elim / Modus Ponens): If you have $A -> B$, and you also
        have $A$, you can extract $B$.
      - `vI` (Or Intro): If you have $A$, you can glue an OR onto it out of 
        nowhere: $A \vee anything$.
      - `vE` (Or Elim / Case Split): Requires two boxes. If you have $A \vee B$,
        you must prove your target $C$ twice. Box 1: Assume $A$, prove $C$. 
        Box 2: Assume $B$, prove $C$. Therefore, $C$ holds.
      - `¬I` (Not Intro): Requires a box. Assume $A$. If this causes a 
        contradiction ($\bot$), close the box and conclude $\neg A$.
      - `¬E` (Not Elim): If you have $A$ and $\neg A$ on different lines, they
        clash. Write $\bot$.  

## Subtopic 2: Derived Rules (The Shortcuts)

### The Intuition
   Derived rules are combinations of core rules that happen so frequently that
   the examiners let you use them as a single-step shortcuts to save time. You
   could prove them from scratch, but you don't have to.

### The Exam MVPs:
   1. Modus Tollens (MT): The reverse of $\to E$. if you have $A \to B$, and
      you know $\neg B$ is true, you can immediately conclude $\neg A$.
   2. DOUBLE NEGATION (DN): If you have ¬¬A, you can instantly drop both to get
      A.
   3. REDUCTIO AD ABSURDUM (RAA): The classical logic powerhouse. If you want to
      prove $A$ (a positive statement), you assume $\neg A$. If that leads to
      $\bot$, you  close the box and conclude $A$. (Note: This is different from
      $\neg I$, which is used to prove a negative statement).
   4. LAW OF EXCLUDED MIDDLE (LEM): You can inject $A \vee \neg A$ into our proof
      out of thin air at any time.

## SUBTOPIC 3: Proof Skeletons and Common Pitfalls

### The Intuition
   The biggest mistake students make in ND exams is starting at line 1 and just
   blindly applying rules downwards. You must build the skeleton from the 
   outside in. Look at the main connective of your target conclusion to know
   exactly which rule you need at the very end.

### The Exam Skeleton:
   - Target is $A \to B$:
      Immediately write "Assume A" at the top of your workspace. Write "B" at
      the bottom. Your final line will be $A \to B$ by `->I`.
   - Target is $\neg A$:
      Immediately write "Assume A". Your sub-target at the bottom is $\bot$. 
      Final line is $\neg A$ by `¬I`.
   - Target is $A \vee B$ (The tricky one):
      You usually cannot prove this directly. Youmust use RAA. Write "Assume
      $\neg(A \vee B)$ at the top. Target $\bot$ at the bottom."
   - Premise contains $A \vee B$:
      Immediately set up a `vE` skeleton. You will need two sub-proofs: "Assume
      $A$" and "Assume $B$".

## The Exam Pitfalls (What Loses Marks):
   1. VAGUE JUSTIFICATIONS: Our new strict formatting rules prevent this. Every
      line must have the exact rule name and the exact line numbers it relies on
      . e.g., `10. B (->E 4, 9)`
   2. LEAKING SCOPE: Citing a line number trapped inside an assumption box
      you have already closed.
   3. CONFUSING $\bot E$ witth RAA: If you have derived $\bot$ (Principle of
      Explosion) to instantly deduce any arbitrary formula $X$ you want. RAA
      specifically allows you to discharge an assumption of ¬X to deduce X.

## Subtopic 1: First-Order Signatures (The Vocabulary)

### The Intuition
   Before you can evaluate if a statement is true, you need to know what words 
   yoyu are legally allowed to use. A SIGNATURE ($L$) is exactly like declaring
   an API or importing a library. It is strictly defined list of symbols, but
   it does not contain any actual data or meaning yet. 

   THE THREE TYPES OF SYMBOLS:
      1. CONSTANTS: Names for specific, fixed individual objects.
         - Example: `saka`, `arsenal`, `zero`.
      2. FUNCTION SYMBOLS: Operations that take one or more objects as input
         and return a new object as output.
         - Example: `father_of(x)`, `add(x,y)`.
         - Exam Note: Every symbol has an "arity" (number of inputs). `father_of`
           is unary (1-ary)' `add` is binary (2-ary).
      3. PREDICATE (Relation) SYMBOLS: Tests that take one or more objects as 
         input and return a Boolean (True/False).
         - Example: `is_manager(x)`, `plays_for(x, y)`
         - Exam Note: Unary predicate act like "types" or "properties" (e.g., 
           $x$ is a manager. Binary predicates act like "relationships") (e.g.,
           $x$ plays for $y$).

## Subtopic 2: Structures, Interpretations, and Domains (The Database)

### The Intuition
   If the Signature is the API declaration, the STRUCTURE ($M$) is the actual
   database backend. It gives concrete meaning to the empty vocabulary.

   A structure is always defined as a pair: $M = (\mathbf{D, I})$.


1. The Domain ($\mathbf{D}$):
   This is your universe of discourse. It is the raw set of "things" that
   actually exist in this specific world. In exams, Imperial usually abstracts 
   these as raw objects like $o_1, o_2, o_3$.

   - Example: $\mathbf{D} = {o_1, o_2, o_3, o_4}$


2. THE INTERPRETATION FUNCTION ($\mathbf{I}$):
   This is the mapping engine. It takes the abstract symbols from your Signature
   and binds them to the concrete objects in your Domain.

   - INTERPRETING CONSTANTS: Maps a name to exactly one object.
     $\mathbf{I}(saka) = o_1$
   - Interpreting Unary Predicates: Maps a porperty to a subset of the domain (a 
     list of all objects that have that property).
     $\mathbf{I}(player) = {o_1, o_2}$
   - INTERPRETING BINARY PREDICATES: Maps a relationship to a set of tuples 
     (pairs).
     $\mathbf{I}(plays_for) = {(o_1, o_4), (o_2, o_5)}$
     (Read as: o1 plays for o4, and o2 plays for o5).

## The Intuition: What is "Soundness"?
   In logic, a proof system is SOUND if it never lies to you.

   Formally, it means: IF YOU CAN PROVE IT USING THE RULES ($\vdash$), IT IS 
   ACTUALLY LOGICALLY TRUE ($\vDash$). For a system to be sound, every single 
   rule i it must be TRUTH-PRESERVING. If you feed the rule True premises, it 
   must be mathematically impossible for it to spit out a False conclusion. If 
   even one rule can take True premises and generate a False conclusion, the 
   entire system is "broken" (not sound).
