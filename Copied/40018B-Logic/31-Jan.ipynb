1. THE BRIDGE: ARGUMENTS vs. FORMULAS.
The lecture introduces a way to convert a "validity check" into a simple formula
calculation.


THE CONCEPT: CORRESPONDING IMPLICATION
- THE ARGUMENT ($\models$): You are given a list of premises 
  ($\phi_1, \dots, \phi_2$) and asked if they force a conclusion ($\psi$) to be
  True.
   - Notation: $\phi_1, \dots, \phi_n \models \psi$
- THE FORMULA ($\rightarrow$): We can glue all the premises together with 
  AND ($\wedge$) and point them at the conclusion with an IMPLICATION ($\rightarrow$)   
   - Notation: $(\phi_1 \wedge \dots \wedge \phi_n) \rightarrow \psi$

1. THE LINK BETWEEN ARGUMENTS AND FORMULAS (S10-11)

The lecture establishes a mathematical bridge between checking an ARGUMENT
(Premises $\therefore$ Conclusion) and checking a FORMULA. This is crucial 
because it allows us to use standard formula tools (like truth tables) to solve
arguments.

   - THE CONCEPT (DEFINITION 0.5): If you have an argument with premises 
     $\phi_1, \dots, \phi_n$ and conclusion $\psi$, you can convert it into a
     a single logical sentence called the CORRESPONDING IMPLICATION FORMULA.
      - SYNTAX: You connect all premises with AND ($\wedge$) and connect that
        whole block to the conclusion with IMPLIES ($\rightarrow$)
      - MATH FORM: $(\phi_1 \wedge \dots \wedge \phi_n) \rightarrow \psi$
   
   - THE RULE (THEOREM 0.6): An argument is VALID if and only if its 
     corresponding implication formula is a TAUTOLOGY (Valid Formula).    <-- true in all scenario.
      - Translation: If the big formula is TRUE IN EVERY SINGLE ROW of a truth
        table, then the argument is valid.

4. Pros and Cons of Truth Tables (S19)
The lecturer summarises why we use (and dislike) this method.

- PROS:
   - THEY ALWAY WORK: For propositional logic, there are finite possibilities, 
     so this method is foolproof.
   - VERSATILE: Can check validity, satisfiability, and equivalence all at once.
      - They illustrate how hard deciding argument validity, formula validity
        , satisfiability and equivalence, etc. can be--even for 'trivial'
        propositional logic. (Adding an atom doubles the length of the table.)

- CONS:
   - EXPONENTIAL EXPLOSION: Adding just one more atom doubles the table size
     ($2^n$ rows).
      - 2 atoms  = 4 rows
      - 3 atoms  = 8 rows
      - 10 atoms = 1024 rows  
   - "DUMB" and "TEDIOUS": It requires no clever reasoning, just brute force
     calculation, which makes it error-prone for humans.
   - NP-COMPLETE: The slide notes that determining satisfiability is

VALID FORMULA

Definition 0.2 (Valid Formula)
   - A propositional formula is (logically) valid if it is true in every 
     situation.

Valid propositional formulas are often called TAUTOLOGIES. A tautology is just
another name for a valid propositional formula.     




SATISFIABLE FORMULA

Definition 0.3 (Satisfiable Formula)
   - A propositional formula is SATISFIABLE if it is true in at least one
     situation.             (situation being $v$ the dictionary...)... situation is evalauted by atomic evaluation
                            function $v$...

We typically say a formula is a CONTRADICTION if it is not satisfiable.
   - $\top$     satisfiable
   - $\bot$     contradiction
   - $p$        satisfiable
   - $p \wedge \neg p$      contradiction
   - $p \rightarrow p$      satisfiable




EQUIVALENT FORMULAS

Definition 0.4 (Equivalent formulas)
   Two propositional formulas $\phi, \psi$ are LOGICALLY EQUIVALENT if they are
   true in exactly the same situations. Roughly speaking: they mean the same.

Some people write $\phi \equiv \psi$ for this. But $\equiv$ is also used in 
other ways, so watch out.

WAYS TO CHECK AN ARGUMENT IS PROPOSITIONALLY VALID
To check whether an argument is propositionally valid, we can:

1. Translate all English sentences into propositional logic;
2. Check whether the resulting argument is valid using:
      - TRUTH TABLES--the obvious way. You go through all the possible relevant
        situations and see if the conclusion is true wheneber the premises are
        true in each.
      - DIRECT "MATHEMATICAL" ARGUMENT--The fastest way, once you get used to it.
      - EQUIVALENCES--You make a stock of useful pairs of equivalent formulas. 
            You use them to reduce its corresponding implication formula step by
            step to $\top$. This takes some art but is quite useful.
      - VARIOUS PROOF SYSTEMS, including tableaux, Hilbert systems, natural 
        deduction. We will look at NATURAL DEDUCTION: it is similar to direct
        argument. We will cnver this intensively.

For valid formulas, equivalence and satisfiability, we proceed in a similar way.

1. SATISFIABILITY
   
   Your Question: "is it because since satisfiability only require AT LEAST ONE scenario... yet we have ALL scenarios true. hence both true???"

YES. 
   - THE BAR: Satisfiability is a low bar. It just asks: "Is it possible for 
     this to be true?" (Is there $\geq 1$ `tt`?).
   - THE RESULT: You have 4 `tt`s.
   - THE VERDICT: If you are perfect (Valid), you are definitely "good enough"
     (Satisfiable).



2. EQUIVALENCE (THE "MATCHING" NUANCE)

   Your Question: "is equivalence true because in all scenarios they are all true?"

CAREFUL HERE. Equivalence does NOT mean the two formulas are "always true." It
means they are "ALWAYS IN SYNC."

Look closely at the table in the slide:
   - Look at the column for $(p \rightarrow q)$: It has values `tt, ff, tt, tt`.
   - Look at the column for $(\neg p \vee q)$: It has values `tt, ff, tt, tt`.


WHY ARE THEY EQUIVALENT?
   - They are equivalent because those two columns are IDENTICAL FINGERPRINTS.
   - Notice the second row (where $p = tt, q = ff$). BOTH sides are `ff`.
      - They are FALSE TOGETHER.
      - Equivalence just requires them to agree. It doesn't care if they agree
        on being True or agree on being False, as long as they match.

THE "RED COLUMN" CONNECTION:
Because the two sides match perfectly, the "Equivalence Checker" (the 
$\leftrightarrow$ in the final red column) outputs `tt` every single time.


SUMMARY
   - SATISFIABILITY: "I found at least one `tt`." (Pass)

This section introduces DIRECT ARGUMENT, a method to check the validity of a
formula without constructing massive truth tables. Instead of brute-forcing 
every row, we use logical reasoning to "break" the formula. If we can't break it
(make it False), it must be Valid (always True).



1. THE STRATEGY: TRY TO BREAK IT
The core idea of a direct argument is to start with a skeptical mindset. We ask:
"Is there any situation where this formula fails (is False)?"

To do this, we usually:
   1. ASSUME THE ANTECEDENT (If-part) is True.
   2. Check if this forces the CONSEQUENT (Then-part) to be True.
   3. If the answer is "YES, it forces it," then the implication can never be
      False (True --> False is impossible). Therefore, it is Valid.





2nd method: Checking validity using direct argument      

It involves showing the truth or falsity of a propositional formula by 
constructing direct logical (or mathematical) arguments from zero or more 
premises to one or more conclusions.

    Is $p \rightarrow (p \vee q)$ A VALID FORMULA?



Take an arbitrary situation (truth value assignment to $p$ and $q$).    

To be a valid formula, then any situation must be such that, if $p$ is true in
that situation then $p \vee q$ is also true in that situation by semantics of 
$\rightarrow$.

Well, if $p$ evaluates to true, then so does $p \vee q$ (by semantics of $\vee$).

And there we are!


Recall that a formula of the form $\phi \rightarrow \psi$ evaluates to false in
one situation only: one in which $\phi$ is true and $\psi$ is false.

2. EXAMPLE 1: MODUS PONENS

   Formula: $p \wedge (p \rightarrow q) \rightarrow q$

   Goal: Proving this is valid.

   TRACE (Step-by-Step):
      1. ASSUMPTION: Take an arbitrary situation where the LEFT SIDE is true.
         - Assumption: $p \wedge (p \rightarrow q)$ is True.
      2. BREAK IT DOWN
         - For an AND ($\wedge$) to be true, both parts must be true.
         - Fact 1: $p$ is True.
         - Fact 2: $(p \rightarrow q)$ is True.
      3. CONNECT THE DOTS
         - Look at Fact 2 $(p \rightarrow q)$. We know from Fact 1 that $p$ is
           True.
         - By the SEMANTICS OF IMPLICATION, if $p$ is True and the implication
           is True, then $q$ MUST be True.
      4. CONCLUSION:
         - We have proven $q$ is True.
         - Since assuming the left side is True forced the right side ($q$) to
           be True, the formula is valid.






3. EXAMPLE 2: EQUIVALENCE (ASSOSCIATIVITY)
Formula: $(\phi \wedge \psi) \wedge \rho \equiv \phi \wedge (\psi \wedge \rho)$

GOAL: Show they are true in exactly the same situations.


TRACE:
   1. START LEFT: Assume $(\phi \wedge \psi) \wedge \rho$ is True.
   2. UNPACK OUTER LAYER:
      - This means $(\phi \wedge \psi)$ is True AND $\rho$ is True.
   3. UNPACK INNER LAYER:
      - Since $(\phi \wedge \psi)$ is True, then $\phi$ is True AND $\psi$ is
        True.
   4. INTERMEDIATE STATE:
      - We now know, $\phi$, $\psi$, and $\rho$ are ALL True.
   5. REBUILD RIGHT:
      - Since $\psi$ and $\rho$ are True, $(\psi \wedge \rho)$ is True.
      - Since $\phi$ is True, then $\phi \wedge (\psi \wedge \rho)$ is True.
   6. CONCLUSION:
      - The left side is True IF AND ONLY IF the Right side is True. They are
        logically equivalent.






4. EXAMPLE 3: PEIRCE'S LAW (THE "HARD" ONE)
FORMULA: $((p \rightarrow q) \rightarrow q) \rightarrow p$

This example is famous because you can't just assume the left side is true
easily. The slide uses an ARGUMENT BY CASES (Law of Excluded Middle).

LOGIC: In any situation, $p$ i either True or False. Let's check both universes.


CASE 1: THE UNIVERSE WHERE $p$ IS TRUE ($v(p) = tt$)
   1. Look at the whole formula: $(\dots) \rightarrow p$.
   2. The Consequent (the final $p$) is True.
   3. BY THE SEMANTICS OF $\rightarrow$, if the Consequent is True, the whole
      implication is True (e.g., "If [anything], then True" is always True).
   4. RESULT: Valid in this universe.


CASE 2: THE UNIVERSE WHERE $p$ IS FALSE ($v(p) = ff$) We need to prove the 
formula holds even here.
   1. ANALYSE THE INNERMOST PART ($p \rightarrow q$):
      - Since $p$ is False, $p \rightarrow q$ is True (BY SEMANTICS OF 
        $\rightarrow$, False implies anything).
   2. ANALYSE THE MIDDLE PART $((p \rightarrow q) \rightarrow p)$:
      - We found $(p \rightarrow q)$ is TRUE (Step 1).
      - We know $p$ is FALSE (Case 2 Assumption).
      - So we have: $True -> False$.
      - This evaluates to FALSE (BY SEMANTICS OF $\rightarrow$).
   3. ANALYSE THE WHOLE FORMULA ($((p \rightarrow q) \rightarrow p) \rightarrow p$):
      - We found the Left Chunk is FALSE (Step 2):
      - So the formula is: $False \rightarrow p$.
      - BY THE SEMANTICS OF $\rightarrow$, if the Antecedent is False, the 
        implication is TRUE (Vacuously True).
   4. RESULT: Valid in this universe.


FINAL CONCLUSION: Sine the formula is True when $p$ is True, AND it is True when
$p$ is False, it is valid in EVERY POSSIBLE SITUATION.

SUMMARY OF DIRECT ARGUMENT METHOD:
   - For $A \rightarrow B$: Assume A is True, prove B must be True.

1. THE CORE CONCEPT: LOGICAL EQUIVALENCE ($\equiv$)

We've moved past just checking if something is "valid" or "satisfiable". Now, we
are massaging formulas. We want to transform a messy formula into a cleaner, 
standard one without changing its meaning.

DEFINITION: Two formulas $\phi$ and $\psi$ are logically equivalent 
($\phi \equiv \psi$) if they have the exact same truth value in every possible
situation.

THE "ALGEBRA" OF LOGIC: Just like in algebra where $x(y + z) = xy + xz$, logic
has rewrite rules.



<<<<<<<>>>>>>>
2. KEY EQUIVALENCE RULES (THE "TOOLKIT")
You need to memorize these or at least recognise them instantly.

A. COMMUTATIVITY & ASSOSCIATIVITY (Order doesn't matter for $\wedge, \vee$)
   - Commutativity: $p \wedge q \equiv q \wedge p$
   - Assosciativity: $(p \wedge q) \wedge r \equiv p \wedge (q \wedge r)$
      - Analogy: "Rain AND Wind" is the same as "Wind AND Rain".

B. IDENTITY & DOMINATION (Constants $\top, \bot$)
   - Identity: $p \wedge \top \equiv p$ (True adds nothing to an AND)
   - Domination: $p \vee \top \equiv \top$ (True overpowers an OR)
   - Domination: $p \wedge \bot \equiv \bot$ (False kills an AND)

C. DOUBLE NEGATION
   - $\neg \neg p \equiv p$
      - Analogy: "It is not the case that it is not raining" = "It is raining."




D. IMPLICATION REWRITING (Crucial for Normal Forms)
This is the most used rule for converting formulas.
   - RULE: $p \rightarrow q \equiv \neg p \vee \q$
   - Trace: "If it rains ($p$), I bring an umbrella ($q$)."
      - This is equivalent to: "Either it doesn't rain $(\neg p)$, OR I bring
        an umbrella ($q$)."

1. THE CORE DEFINITION
   - The Formula: $p \vee \neg p$
   - The Translation: "Either $p$ is True, or $\neg p$ is True."
   - The Rule: There is no third option. You cannot be "sort of" pregnant, and a
     boolean variable cannot be "kinda" true.

2. WHY IS IT CALLED "EXCLUDED MIDDLE"?
   Imagine a line with two sides:
   - Side A: True.
   - Side B: False
   - The Middle: Maybe/Unknown/Sort-of

   This law takes that "Middle" section and DELETES IT. It "excludes" the 
   possibility of sitting on the fence.
      - Example: "It is raining."
         - Logic says: Right now, it is either raining ($T$) or it is not 
           raining ($F$).
         - EXCLUDED MIDDLE says: You cannot say "Well, it is msiting, so it's
           technically neither." Logic forces you to pick a bucket: Is "mist"
           rain? Yes or No?
