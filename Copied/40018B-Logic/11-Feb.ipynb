1. THE STRATEGY: SUB-CONCEPTS (S69)
   The slide advises breaking complex sentences into small pieces 
   ("sub-concepts") first to avoid getting lost in the brackets.


   EXAMPLE: "No lecturer bought a PC."
     1. SUB-CONCEPT: "$x$ bought a PC."
        - Logic: $\exists y (Bought(x, y) \wedge PC(y))$
     2. FULL SENTENCE: "There does NOT exist a lecturer ($x$) s.t. sub-concept 
        is true."
        - Logic: $\neg \exists x (Lecturer(x) \wedge [\exists y (Bought(x, y) \wedge PC(y))])$   

     KEY WARNING:
        - Correct: $\forall x ([\exists y \text{Bought}(y, x)] \to \neg \text{Human}(x))$
        - Incorrect: $\forall x \exists y (\text{Bought}(y, x) \to \neg \text{Human}(x))$
           - Why: Moving the quantifier $\exists y$ to the front changes the
             meaning. Keep quantifiers as close to their variables as possible.


   ---
   2. COMMON PATTERNS & MISTAKES
   
   The slide reinforces the standard pairing you must memorise.

   - UNIVERSAL ($\forall \rightarrow$): Used for "All", "Every".
      - Correct: $\forall x (Lecturer(x) \rightarrow Human(x))$
      - Mistake: Using $\wedge$. ($\forall x (Lecturer(x) \wedge Human(x))$)
        means everything in the universe is a lecturer.
   
   - EXISTENTIAL ($\exists \wedge$): Used for "Some", "At least one".
      - Correct: $\exists x (Lecturer(x) \wedge Human(x))$
      - Mistake: Using $\rightarrow$. ($\exists x (Lecturer(x) \rightarrow Human(x))$
        is almost always vacuously true).
     EXAM TIP: The slide notes that $\exists x (A \rightarrow B)$ is "EXTREMELY
        RARE." If you write this in an exam, stop and check your work--you 
        almost certainly made a mistake.







   ---         
   3. HOW TO COUNT: "AT LEAST" (S71)

   Since we can't write "$\geq 2$", we have to say "There exist two things that
   are NOT EQUAL."
      - "AT LEAST ONE PC":
         - $\exists x PC(x)$

      - "AT LEAST TWO PCs"
         - $\exists x \exists y (PC(x) \wedge PC(y) \wedge x \neq y)$
         - Note: If you forget $x \neq y$, you haven't proved there are two.
           $x$ and $y$ could just point to the same single PC.

      - "AT LEAST THREE PCs"
         - You need 3 variables and msut ensure NONE of them equal each other.
         - $\dots \wedge x \neq y \wedge y \neq z \wedge x \neq z$.





---
4. ADVANCED COUNTING: "AT MOST" & "EXACTLY" (S72)

   This is the hardest syntax to master.
   

   A. "AT MOST ONE PC"      
      This implies: "It is NOT the case that there are two PCs."
      - METHOD 1 (NEGATION): $\neg [At least two PCs]$
         - $\neg \exists x \exists y (PC(x) \wedge PC(y) \wedge x \neq y)$
      - METHOD 2 (DIRECT - PREFERRED): "If you find any two PCs $(x, y)$, they
        must actually be the same one."
         - $\forall x \forall y (PC(x) \wedge PC(y) \rightarrow x = y)$.


   B. "EXACTLY ONE PC"
      This combines "At least one" AND "At most one."
         - FORMULA: $\exists x (PC(x) \wedge \forall y (PC(y) \rightarrow x = y))$         
                                            <-- ahh! okay, I get the difference now. It's actually because $\exists x$
                                                helps to enforce the constraint! where there MUST be at least ONE
            - Translation: "There exists a 'main' $PC(x)$. And if you find any 
              PC(y) in the world, it must actually by $x$."

   SUMMARY OF COUNTING PATTERNS FOR EXAM:
      1. AT LEAST 2: $\exists x \exists y (\dots \wedge x \neq y)$
      2. AT MOST 1 : $\forall x \forall y (\dots \rightarrow x = y)$
      3. EXACTLY 1 : $\exists x (\dots \wedge \forall y (\dots \rightarrow x = y))$

... run-through of MANY-SORTED LOGIC. 

This is essentially "Typed Logic." Just like in programming (e.g., Haskell or 
Java) where you can't add a String to an Integer, in Many-Sorted Logic, you 
can't apply a predicate to the wrong type of object.




---
1. THE CORE CONCEPT: SORTS = TYPES (S74)
   
   In standard FOL, everything is just an "object" in one big bucket (the 
   Domain). In MANY-SORTED LOGIC, we split the universe into different buckets
   called SORTS (e.g., `Integer`, `Person`, `PC`).

   - VARIABLES & CONSTANTS: Must have a specific sort.
      - Instead of just $x$, we write $x : s$ (variable $x$ of sort $s$).
      - Instead of just $c$, we write $c : s$.

   - FUNCTION TEMPLATES: Functions are now strict about inputs and outputs.
      - Template: $f : (s_1, \dots, s_n) \rightarrow s$.
      - Rule: You can only put terms of sort $s_1, \dots, s_n$ into $f$. If you
        try to put a `Person` into a function expecting a `Number`, it's not
        just false--it's RUBBISH (ill-formed syntax).




----
2. FORMULAS AND "POLITE" QUANTIFIERS        
   - PREDICATES: Also have templates. $R(s_1, \dots, s_n)$
      - You can only check if $R(t_1 \dots t_n)$ is true if the terms $t$ match
        the sorts $s$.
   - EQUALITTY ($=$): You can only compare things of the SAME SORT.
      - $x : Person = y : Person$ is valid.
      - $x : Person = y : PC$ is RUBBISH.
   - QUANTIFIERS: You must specify the sort being quantified.
      - $\forall x : s \phi$ ("For all $x$ of sort $s$, $\phi$ is true.")
      - $\exists x : s \phi$ ("There exists an $x$ of sort $s$ s.t. ...")



---
3. THE STRUCTURE DIAGRAM (S76)
   This slide visualises a Many-Sorted Model ($M$). Instead of one big box, the
   universe is partitioned into disjoint (non-overlapping) regions for each 
   sort.

   - Sort `lecturer`: Contains ${frank, susan, tony}$
   - Sort `pc`: ...
   - Sort `rest`: Contains the other circles.
   - Relationships: Arrows can go between sorts (e.g., `Bought` goes from 
     `lecturer` to `pc`).





---
4. INTERPRETATION RULES (S77)
   
   The rules for the Model ($M$) are stricter now:
      1. CONSTANTS: If $c$ has sort $s$, the object $I(c)$ must be in the sort
         bucket $s$.
      2. FUNCTIONS: If $f : (s_1) \rightarrow s$, then for every input from 
         bucket $s_1$, the output must land in bucket $s$.  
      3. RELATIONS: If $R(s_1, s_2)$, the relation is only defined for pairs 
         where the first is from $s_1$, and the second from $s_2$.







---
5. PRACTICAL NOTES & HASKELL CONNECTION (S78)

   - SORTS REPLACE UNARY PREDICTES:
      - OLD WAY: One big domain, check `Human(x)` to see if $x$ is human.
      - NEW WAY: `Human` is a SORT. If $x$ is a variable of sort `Human`, we 
        know it's human. We don't need a predicate `IsHuman(x)`.
   - DISJOINTNESS: In this logic (like Haskell types), an object usually has
     EXACTLY ONE sort. You can't be both a `lecturer` and a `pc`.



EXAM TAKEAWAY: If asked to write a formula in Many-Sorted Logic, ALWAYS specify
the type of your quantifiers (e.g., $\forall x : Person$) and ensure your 
predicates match the types defined in the signature.

2. THE "CRASH" PROBLEM: PARTIAL FUNCTIONS (S81)
   
   This is a CRITICAL TECHNICAL DETAIL for exams.

   In Haskell, if you type `head []` or `tail []`, the program crashes 
   (exception). In FOL, CRASHES ARE NOW ALLOWED. Every function MUST return a 
   value for every possible input (functions must be TOTAL).

   THE LOGIC FIX: The Model ($M$) must assign an ARBITRARY VALUE to these 
   "illegal" operations.
      - QUESTION: What is `head([])` in the logic?
      - ANSWER: The logic assigns it a random garbage value (e.g., `0` or `99`).
      - WHY? Because the function $f : [Nat] \rightarrow Nat$ must map `[]` to
        something in the `Nat` bucket.
      - WARNING: The slide explicitly says: "Don't assume they have the values
        you might expect. Do not write proofs assuming `head([]) = error`."

      $\# ([]) = 0$

      $\forall x \forall xs (\#(x : xs) = #(xs) + 1)$
      - Translation: The length of a list `x:xs` is the length of the tail `xs`
        plus 1.
   - LAW 3 (HEAD/CONS RELATION):

4. VALIDITY AND SATISFIABILITY (S83-84)
   Now we zoom back out to the general theory. These definitions are almost
   identical to Propositional Logic, but "Situtation" has a new meaning.

   A. VALID ARGUMENT ($\models$)
      $$A_1, \dots, A_n \models B$$

      - DEFINITION: An argument is valid if, for EVERY POSSIBLE STRUCTURE $M$
        and EVERY POSSIBLE ASSIGNMENT $h$:
         - IF the premises ($A_1, \dots, A_n$) are all True in $(M, h)$.
         - THEN the conclusion $(B)$ must also be True in $(M, h)$.


   B. VALID FORMULA / TAUTOLOGY ($\models A$)
      - DEFINITION: A formula $A$ is valid if it is True in EVERY POSSIBLE 
        UNIVERSE (every structure $M$ and assignment $h$).
      - Example: $\forall x (P(x) \rightarrow P(x))$. (True no matter what $P$
        means).
      - Example: $\forall x P(x) -> \exists x P(x)$. (If everyone is $P$, then
        someone is $P$).


   C. SATISFIABLE FORMULA
      - DEFINITION: A formula $A$ is Satisfiable if there is AT LEAST ONE 
        structure $M$ and assignment $h$ where it evaluates to True.
      - Example: $\forall x P(x)$
         - Satisfiable? Yes (Imagine a world where everyone is happy)?
         - Valid? No (Imagine a world where someone is Sad).

   3. VALIDITY ($\models$): Means "True in EVERY structure." To disprove 
      validity, you only need to invent ONE weird structure (Counter-model)

1. EQUIVALENCE & THE "IMPOSSIBLE" PROBLEM (S85-86)

   LOGICAL EQUIVALENCE (Definition 1.13)

   Two formulas $A$ and $B$ are LOGICALLY EQUIVALENT if they mean the exact
   same thing in EVERY POSSIBLE UNIVERSE.
      - SYMBOL: $A \Longleftrightarrow B$
      - CONDITION: For every structure $M$ and every assignment $h$:
         - $M, h \models A$ if and only if $M, h \models B$.


   THE PROBLEM: YOU CAN'T "CHECK ALL" (S86)
   In Propositional Logic, to check validity, you just draw a Truth Table. It
   might be big, but it was finite. In FOL, TRUTH TABLES ARE IMPOSSIBLE.

     - WHY? Because a "Structure" ($M$) can be anything. It could have 1 object,
       10 objects, or infintie objects (like the Integers $\mathbf{N}$).
     - CHURCH'S THEOREM (1936): It is mathematically proven that NO COMPUTER
       PROGRAM can ever exist that perfectly decides if a FOL argument is valid.
     - Implication: We can't just "calculate" the answer. We have to be clever
       and PROVE it.

3. WALKTHROUGH: DIRECT ARGUMENT EXAMPLE 1 (S89)

   THE GOAL: Prove that if everyone is $(Human \rightarrow Lecturer)$ AND
      everyone is $(PC \rightarrow Lecturer)$ AND everyone is $(Human \vee PC)$,
      then EVERYONE IS A LECTURER.


   FORMAL ARGUMENT:   
      $$\forall x (Human(x) \rightarrow Lecturer(x)), \forall x (PC(x) \rightarrow Lecturer(x)),$$
      $$\forall x (Human(x) \vee PC(x))$$

   THE PROOF TRACE (Direct Argument):
      1. SETUP: Take an ARBITRARY STRUCTURE $M$ that satisfies the 3 premises.
         We want to show the conclusion holds in $M$.
      2. PICK AN OBJECT: To prove $\forall x Lecturer(x)$, let's pick a random
         object $a$ from the domain of $M$.
      3. USE PREMISE 3: We know $\forall x (Human \vee PC)$. Therefore, for our
         object $a$, it must be true that $Human(a) \vee PC(a)$.
      4. CASE ANALYSIS:
         - Case 1: $a$ is Human.
            - By Premise 1 $(Human \rightarrow Lecturer)$, $a$ must be a 
              LECTURER.
         - Case 2: $a$ is a PC,
            - By Premise 2 $(PC \rightarrow Lecturer)$, $a$ must be a LECTURER.
      5. CONCLUSION: In either case, $a$ is a Lecturer. Since $a$ was random,
           EVERYONE is a Lecturer.

4. WALKTHROUGH: THE "HORSE HEAD" EXAMPLE (S90)
   
   This is a famous "Harder Example" (De Morgan's argument) that PL fails to 
   solve.

   THE ARGUMENT:
      - Premise: "All horses are animals."
         - $\forall x (Horse(x) \rightarrow Animal(x))$
      - Conclusion: "All horse-heads are animal-heads."
         - $\forall x([\exists y(HeadOf(x,y) \wedge Horse(y))] \rightarrow [\exists y (HeadOf(x, y) \wedge Animal(y))])$   
   
   THE PROOF TRACE:
      1. Setup: Assume the Premise is True in model $M$.
      2. Pick an Object: Let $b$ be any object in the domain. We want to prove
         the Conclusion for $b$.
      3. Assume the "If" part: Assume the Left Side of the arrow is true for $b$
         - Assumption: $b$ is a horse-head.
         - Math: $M \models \exists y (HeadOf(b, y) \wedge Horse(y))$   
      4. EXTRACT THE WITNESS: Since the existential ($\exists y$) is true, there
         must be a specific object (let's name if $c$) that is the horse 
         attached to that head.
            - Facts established:
               1. $HeadOf(b, c)$ is True.
               2. $Horse(c)$ is True.
      5. APPLY THE PREMISE: The premise says "All horses are animals."
         - Since $Horse(c)$ is True, then $Animal(c)$ must be True.
      6. BUILD THE CONCLUSION:
         - We now have: $HeadOf(b, c)$ AND $Animal(c)$
         - Therefore, there exists a $y$ (specifically $c$) s.t. 
           $HeadOf(b, y) \wedge Animal(y)$.
         - This matches the definition of "Animal-Head" on the right side.
      7. FINAL RESULT: The argument is Valid.



EXAM TIP FOR DIRECT ARGUMENTS:
   - Start with "Let $a$ be an arbitrary object..." This is the standard opening
     move to prove a $\forall x$ sentence.
   - Unpack Definitions: If you see $\exists y$, immediately say "Let $c$ be
     the object that makes this true" (give it a name).
   - Keep track of your facts: List what you know about your specific objects
     $(a, b, c)$ step-by-step.

1. DIRECT ARGUMENT WITH EQUALTIY (S91)
   
   This slide demonstrates how to prove a formula is Valid when it contains the
   equality symbol ($=$). The key rule of equality is SUBSTITUTION: If $a = b$,
   anything true about $a$ is also true about $b$.

   THE ARGUMENT TO PROVE: 
      $$\forall x \forall y (x = y \wedge \exists z R(x, z) \rightarrow \exists v R(y, v))$$

   THE TRACE (Proof):
      1. ASSUME THE "IF" PART: Take an arbitrary structure $M$. Let $a$ and $b$
         be objects s.t. $a = b$ AND $\exists z R(a, z)$.
      2. UNPACK THE EXISTENCE: Since $\exists z R(a, z)$ is true, there is some
         object (let's call it $c$) where $R(a, c)$ holds.
      3. USE EQUALITY: Since $a = b$, they are the SAME OBJECT. We can replace
         $a$ and $b$ in our fact from Step 2.
            - Fact: $R(a, c)$ becomes $R(b, c)$.
      4. PROVE THE CONCLUSION: We need to show $\exists v R(y, v)$ (where $y$ is
         $b$).
            - Do we have a witness for $v$? Yes, object $c$.
            - Since $R(b, c)$ is true, $\exists v R(b, v)$ is true.
      5. DONE.

2. SATISFIABILITY vs. VALIDITY (S92)

   - VALIDITY: True in EVERY possible world (Hard to prove).
   - SATISFIABILITY: True in AT LEAST ONE world (Easy to prove--just give one
     example).

      - Result: Since it works in one math world, the formula is SATISFIABLE.

3. BASIC EQUIVALENCES (S94)#

   These are the "Algebra Rules" of Logic. You can use these to simplify 
   formulas without proving them from scratch.

   - SWAPPING ORDER:
      - $\forall x \forall y \equiv \forall y \forall x$ (You can swap universals).
      - $\exists x \exists y \equiv \exists y \exists x$ (You can swap existentials).
      - Warning: You can NEVER swap $\forall x \exists y$ with $\exists y \forall x$
        (Everyone loves someone $\neq$ Someone is loved by everybody).
   
   - DE MORGAN'S FOR QUANTIFIERS (Memorise This):
      - $\neg \forall x \equiv \exists x \neg A$ 
            (Not everyone is happy $\equiv$ Someone is not happy)    
      - $\neg \exists x A \equiv \forall x \neg A$
            (No one is happy $\equiv$ Everyone is not happy)

   - DISTRIBUTION:
      - $\forall$ distributes over $\wedge$. 
            ("Everyone is tall and smart" $\equiv$ "Everyone is tall" AND "Everyone is smart").
      - $\exists$ distributes over $\vee$.

4. ADVANCED EQUIVALENCES: "NULL" VARIABLES (S95)
   
   These rules apply when a variable DOES NOT APPEAR in part of the formula.
   (e.g., in $\forall x (\phi \wedge \psi)$, if $x$ is not in $\phi$).

   - THE "PULL OUT" RULE: If a part of the formula ($\phi$) doesn't use $x$,
     we can move it OUTSIDE the quantifier scope.
      - $\exists x (\phi \wedge \psi) \equiv \phi \wedge (\exists x \psi)$   
      - Analogy: In programming, if you have 
          `for x in list: print (constant + x)`
        , you can calculate the `constant` outside the loop.

   - THE "FLIPPER" TRAP (Rule 10 - Crucial for Exam): If you pull a quantifier
     out from the RIGHT SIDE of an implication ($\rightarrow$), it stays the
     same.
        - $\forall x (\phi \rightarrow \psi) \equiv \phi \rightarrow \forall x \psi$
          . But if you pull it out from the LEFT SIDE (antecedent), the quantifier 
          FLIPS.
        - $\forall x (\phi \rightarrow \psi) \equiv (\exists x \phi) \rightarrow \psi$
        - $\exists x (\phi \rightarrpw \psi) \equiv (\forall x \phi) \rightarrow \psi$
        - Why? Because $A \rightarrow B$ is logically $\neg A \vee B$. The
          negation touches the $A$, so when the quantifier comes out, De Morgan's
          Law flips it.

... run-through of VARIABLE RENAMING, SUBSTITUTION, AND INTRODUCTION TO
NATURAL DEDUCTION.

This section covers the technical rules you need to manipulate formulas without
"breaking" them, which leads into the final topic: Natural Deduction proofs for
Predicate Logic.



1. RENAMING BOUND VARIABLES (S97)
   
   This is the "Alpha Conversion" rule (if you know Lambda Calculus).
   THE RULE: You can rename a BOUND variable to anything else, as long as it
      doesn't clash with an existing variable.
      
      - VALID: $\forall x \exists y Bought(x, y) \equiv \forall z \exists v Bought(z, v)$.
            (Changed $x \rightarrow z, y \rightarrow v$)
      - MEANING: "Everyone bought something" is the same sentence regardless of
        whether you call the person $x$ or $z$. The variable name is just a 
        placeholder.




2. EQUALITY EQUIVALENCES (S98)
   
   These rules are specific to the equality symbol ($=$).
   - REFLEXIVITY: $t = t$ is always valid ($\top$)
   - Leibniz Principle (Substitution of Equals):
      - If $x = y$, then anything true about $x (\phi)$ is also true about $y$.
      - FORMULA: $x = y \rightarrow (\phi \Longleftrightarrow \phi[y/x])$
      - Example: If $x = y$, then $(\forall z R(x, z)) \Longleftrightarrow (\forall z R(y, z))$.





3. SUBSTITUTION: THE "FIND AND REPLACE" TOOL (S 105-106)
   
   DEFINITION ($\phi[t/x]$): This notation means "Take formula $\phi$, and
      replace every FREE OCCURENCE of variable $x$ with the term $t$".

   THE TRAP (CAPTURE): You cannot just blindly replace variables. You must
   ensure the new term $t$ doesn't get "captured" by a quantifier inside $\phi$.

      - EXAMPLE OF BAD SUBSTITUTION:
         - Formula $\phi$: $\exists y Loves(x, y)$ ("$x$ loves someone").
         - Substitution: Replaces $x$ with $y$
         - Result: $\exists y Loves(y, y)$ ("Someone loves themselves")
         - PROBLEM: The meaning changed completely! The free variable $y$ (from
           outside) got captured by the $\exists y$ inside.
      
      - THE RULE: You are not allowed to substitute $t$ for $x$ if any variable
        in $t$ would become bound in the process.




4. NATURAL DEDUCTION FOR PREDICATE LOGIC (S 103-104)        
   
   We are now moving back to Proofs (like in PL), but adding rules for 
   $\forall, \exists, =$.

   KEY CONCEPT:
      - DIRECT ARGUMENT: Informal, English-style proof (what we did in the last
        lecture).
      - NATURAL DEDUCTION: Formal, rule-based proof (drawing boxes and lines)
      - Strategy: The lecturer advises: "Think of a direct argument first, then
        convert to NATURAL DEDUCTION."






5. EXISTENTIAL INTRODUCTION ($\exists I$) (S-107)

   This is the first new rule. It says: "If I have a specific example, I can
   claim `Something` exists."

   THE RULE: 
      1. PREMISE: $\phi[t/x]$ (The formula is true for a specific term $t$)
      2. CONCLUSION: $\exists x \phi$ (Therefore, there exists an $x$).

   EXAMPLE:
      1. `Human(Socrates)` (Fact)
      2. `Mortal(Socrates)` (Fact)
      3. `Human(Socrates) ^ Mortal(Socrates)` (^I)   
      4. $\exists x (Human(x) \wedge Mortal(x))$ ($\exists-I$, replacing `Socrates` with $x$)

   
   EXAM TIP: You can replace any term with an existential quantifier.
      - From `Love(frank, susan)`, you can derive:
         - $\exists x Loves(x, susan)$ ("Someone loves Susan").
         - $\exists y Loves(frank, y)$ ("Frank loves someone").
         - $\exists x \exists y Loves(x, y)$ ("Someone loves someone").




... run-through of the NATURAL DEDUCTION RULES FOR QUANTIFIERS.

This section introduces the two "hard" rules: EXISTENTIAL ELIMINATION ($\exists E$)
and UNIVERSAL INTRODUCTION ($\forall I$). Both require you to use BOXES to handle
temporary variables properly.



<<<<>>>>
1. EXISTENTIAL ELIMINATION ($\exists E$) - THE "UNPACKING" RULE (S 108-111)
   
   THE PROBLEM: You know $\exists x \phi$ ("Something exists"), but you don't
   know what it is.
      - Example: "Someone stole my lunch." (True)
      - Trap: You can't just say "Frank stole it." You don't know that.

   THE SOLUTION: You give the unknown object a TEMPORARY DUMMY NAME (usually 
   $c$),
      1. ASSUME: "Let's call the thief $c$." ($\phi[c/x]$)
      2. REASON: Prove your conclusion ($\psi$) using $c$.
      3. ESCAPE: If you can prove $\psi$ (e.g., "I am angry") without relying on
        who specifically $c$ is, you can close the box and keep the conclusion.

   CRITICAL RULES FOR THE BOX:
      1. FRESHNESS: The name $c$ must be NEW. You can't use a name already in
         use (like `frank` or `susan`).
      2. NO LEAKS: The final conclusion $\psi$ CANNOT CONTAIN $c$. The dummy 
         name must die inside the box.

5. UNIVERSAL INTRODUCTION ($\forall I$) (S115)
   
   This corresponds to the mathematical phrase "Let $x$ be an arbitrary number..."

   THE RULE: To prove $\forall x \phi$, you must:
      1. Open a box
      2. Introduce a FRESH constant $c$ (just like $\exists E$)
      3. Prove $\phi[c/x]$ inside the box.
      4. CONDITION: You must prove it WITHOUT ASSUMING ANYTHING SPECIAL about $c$
         - You cannot assume $P(c)$ at the top of the box.
         - $c$ must be truly "blank"


   VISUALISING THE BOX: 
      - HEADER: Just the name $c$. (No formula assumption)
      - BODY: Proving the property holds for $c$.
      - EXIT: $\forall x \phi$.





SUMMARY FOR EXAM:    
   - $\exists E$: Box starts with "Assume $\phi [c/x]$". Used to USE an 
     existential premise.
   - $\forall I$: Box starts with just $c$. Used to PROVE a universal conclusion.
   - NAMES: Always pick a name ($c, d, e$) that hasn't appeared yet in the proof.

... run-through of the UNIVERSAL QUANTIFIER RULES ($\forall$) in ND.

This section covers how to prove a universal statement (hard) and how to use
one (easy).



1. UNIVERSAL INTRODUCTION ($\forall I$) - THE "HARD" RULE (S 116-117)
   
   GOAL: You want to prove a sentence like $\forall x \phi$ ("Everything is $\phi$")

   THE STRATEGY: To prove something is true for eevryone, you pick a totally
   random, generic object ($c$), prove it's true for that object, and then say
   "Since $c$ was arbitrary, it must be true for everyone."


   THE BOX STRUCTURE:
      1. OPEN A BOX:
         - HEADER: Just write the name $c$. DO NOT write "Assume $P(c)$" or
           anything else. Just the name.
         - Meaning: "Let $c$ be an arbitrary object."

      2. INSIDE THE BOX:
         - Work your way down to prove $\phi[c/x]$ (The formula applied to $c$).

      3. CLOSE THE BOX:
         - CONCLUSION: $\forall x \phi$
         - JUSTIFICATION: $\forall I$ (on the subproof).

      CRITICAL RULE (The "Freshness" Condition): The constant $c$ must be FRESH.
         It cannot appear in any uncancelled assumptions or premises outside the
         box.
         - WHY? If you know $c$ is "Frank", and you prove "Frank is happy", you
           cannot conclude "Everyone is happy." You can only conclude that if 
           $c$ is a generic "nobody".





<<<<>>>>
2. UNIVERSAL ELIMINATION ($\forall E$) - THE "EASY" RULE (S118)

GOAL: You already know $\forall x \phi$ ("Everything is $\phi$"), and you want
to use that fact.

THE STRATEGY: Since it's true for everything, you can instantiate it for 
anything you want.

THE RULE: 
   1. Premise: $\forall x \phi$
   2. CONCLUSION: $\phi[t/x]$
      - Note: You can replace $x$ with ANY TERM $t$ you like (a constant like
        `frank`, a variable, or a compelx term like `father_of(tony)`)


EXAMPLE TRACE:
   1. $\forall x Human(x)$              (Premise)
   2. $Human(Socrates)$                 ($\forall E$, substituting `Socrates`)
   3. $Human(father_of(Socrates))$      ($\forall E$, substituting complex term.)

   TRAP 1: THE LEAKING CONSTANT
      - SCENARIO: You use $\forall I$ to prove $\forall x P(x, c)$ inside a box
        where $c$ was introduced.
      - THE ERROR: You cannot conclude $\forall x P(x, c)$ if $c$ is the
        arbitrary constant yoy just created to prove the outer quantifier. The
        name $c$ cannot appear in the conclusion of the $\forall I$ rule.




TRAP 2: REUSING CONSTANT (S 121)
   - ...
   - WHY? The conclusion on line 4 contains $c$. You are trying to say 
     "Everything is related to $c$, but $c$ was just a temporary tool. You

   - $\forall I$ (Proving All): Requires a Box. Start with a name ($c$). End
     with the formula using $c$. Result is $\forall x \dots$
   - $\forall E$ (Using All): No Box needed. Just plug in whatever name or term
     you need to make the next step of your proof work.  

... run-through of the DERIVED RULES and EQUALITY RULES. These are advanced
shortcuts to make your proofs faster and cleaner.


1. THE "COMBO" RULE: $\forall \rightarrow E$ (S127)
   This is a shortcut that combines two steps: Universal Elimination 
   ($\forall E$) and Modus Ponens ($\rightarrow E$).

   THE SCENARIO: You often encounter a universal rule like: "For all numbers $x$
   , if $x$ is even, $x + 1$ is odd."

   - $\forall x (Even(x) \rightarrow Odd(x + 1))$
   - You also have a specific fact: $Even(2)$


   THE OLD WAY (2 STEPS):
      1. Use $\forall E$ to instantiate for 2: $Even(2) \rightarrow Odd(3)$
      2. Use $\rightarrow E$ (Modus Ponens) with $Even(2)$ to get: $Odd(3)$

   THE NEW SHORTCUT ($\forall \rightarrow E$): You can jump straight from the
   universal rule + the specific fact to the conclusion in ONE LINE.
      - RULE: If you have $\forall x (\phi \rightarrow \psi)$ and you have
        $\phi[t]$, you can derive $\psi[t]$.







<<<<>>>>        
2. EQUALITY RULE 1: REFLEXIVITY (`refl`) (S129)
   
   This rule essentially says: "Things are equal to themselves."

   THE RULE: At any point in a proof, you can write down $t = t$ for any term
      $t$.
      - JUSTIFICATIONS: `refl`
      - Why? Because it's an undeniable axiom. $a = a$ is always true.





<<<<>>>>      
3. EQUALITY RULE 2: SUBSTITUTION (`=sub`) (S130)

   This is the "Search and Replace" rule for proofs.


   THE RULE: If you know $t = u$, and you know a formula $\phi$ is true for $t$,
   then $\phi$ MUST also be true for $u$.

      1. LINE 1: $\phi [t / x]$ (e.g., `Human(Clark_Kent)`).
      2. LINE 2: $t = u$ (e.g., `Clark_Kent = Superman`).
      3. RESULT: $\phi[u/x]$ (e.g., `Human(Superman)`).
         - JUSTIFICATION: `=sub(1, 2)`

   NOTE: You can replace some or all occurrences. It doesn't matter, because
   they are the same object.





<<<<>>>>
4. EQUALITY RULE 3: SYMMETRY (`=sym`) (S 131)
   
   This is a "Derived Rule" (a shortcut built from the first two).
      - CONCEPT: If $a = b$, then $b = a$.
      - THE PROOF (How it works under the hood):
         1. Assume $c = d$
         2. Write $c = c$ (`refl`)
         3. Use `=sub` to replace the first $c$ with $d$ (using line 1).
         4. Result: $d = c$
      - EXAM USAGE: You don't need to prove it every time. Just write `=sym` to
        flip an equality statement.

   2. FRESHNESS: Always use a new name ($c$) when opening these boxes.
   3. INSTANTIATION: Use $\forall E$ and $\exists I$ to swap between 
      "Everyone/Someone" and specific names.
   4. EQUALITY: Use `refl` to introduce logic ($a = a$) and `=sub` to swap equal
      terms.

- In FOL, a model is a mathematical structure consisting of a non-empty domain
  (set of objects) and an interpretation that assigns meaning to symbols
  (constants, functions, predicates), making a set of formulas true. It defines
  the semantic context, determing if a statement is true or false.
     - STRUCTURE: A model $M$ is typically defined as $(D, I)$, where $D$ is the
       domain (universe) of discourse, and $I$ is the interpretation function.
     - INTERPRETATION: $I$ assigns specific elements to $D$ to constant symbols,
       functions over $D$ to function symbols, and relations over $D$ to
       predicate symbols. 
     - SATISFACTION: A mdoel satisfies a formula (or a set of formulas/theory)
       if all formulas are true under that interpretation, denoted as 
       $M \models \phi$.
     - PURPOSE: Models are used to determine the validity of formulas and 
       provide semantic meaning to syntactic expressions.
     

- Semantics in FOL defines the meaning of logical expressions by determining 
  their truth value (true or false) within a specific interpretation, or model.
  It maps syntactic symbols--constants, predicates, and functions--to a 
  non-empty DoD to evaluate formulas.     




- In FOL, a formula is satisfiable if there exists at least one interpretation
  (model) that makes it true, while it is VALID (a tautology) if it is true
  under all possible interpretations. A formula is valid if and only if its
  negation is unsatisfiable.
