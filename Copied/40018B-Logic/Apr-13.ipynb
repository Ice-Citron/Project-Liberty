## Subtopic 1: Propositional Syntax (The Grammar)

### The Intuition
   Syntax is purely about the structure of the symbols, with absolutely no care
   for what they mean. Think of it like definining a strict recursive data type
   in Haskell or building an Abstract Syntax Tree (AST) in a compiler. A 
   sequence of symbols is only a "Well-Formed Formula" (WFF) if it perfectly 
   follows the recursive building rules. We saw a glimpse of this with binding
   strength earlier: syntax dictates the exact shape of the tree.

## Subtopic 2: Semantics (The Meaning)

### The Intuition      
   If syntax is building the AST, semantics is writing the evaluator function.
   We assign a "valuation" (a state where every variable is strictly True or 
   False) and evaluate the tree from the leaves up to the root.

   In exams, examiners don't jusst want you to calculate truth tables; they 
   want you to categorise the nature of the formula across all possible states. 
   Imagine testing the logic governing a rogue AI agent:

   1. VALID (Tautology): The formula evaluates to True in every possible 
      valuation. It is a universal truth of the system (e.g., $p \vee \neg p$).
      The agent's logic is ironclad.
   2. SATISFIABLE: The formula evaluates to True in at least one valuation. It
      is possible under certain conditions (e.g., $p \wedge q$). The agent can
      execute this path. (Note: Valid formulas are automatically satisfiable).
   3. CONTRADICTION (Unsatisfiable): The formula evaluates to False in every
      possible valuation (e.g., p ^ neg p). The logic is fundamentally broken. 


### The Visual Framing: The State Space Map
   A truth table is just a brute-force map of the entire state space. If an exam
   asks you to show a formula is a contradiction using "equivalence law" or 
   "direct argument", they are asking you to algebraically collapse that entire
   state space down to a single `False` without drawing the table.

## SUBTOPIC 3: Transation (English to Formulas)   

### The Intuition
   Natual language is messy and ambiguous; propositional logic is ruthless and 
   exact. Translation questions on the exam test your ability to spot English
   idoioms and map them to strict logical connectives. 


### The Translation Traps (Exam Focus)
   Examiners love to use specific English phrasing to trip you up. ... cheat
   sheet for the most common traps:

   - "A if B": B -> A
      - The trap: The order flips. The thing following "if" is always the premise
        (the left side of the arrow)

   - "A only if B": A -> B
      - The trap: it sounds like it should be the other way around. But if A can
        only happen if B happens, then knowing A happened guarantees B must have
        happened.

   - "A if and only if B": A <-> B
      - The Intuition: They are perfectly locked together.

   - "A unless (if not) B": A v B (or symetrically, `\neg B -> A`)
      - The intuition..."

   - "But" / "Even though": A ^ B
      - The intuition: "It is raining but I am happy" is logically just 
        "It is raining AND I am happy." Logic doesn't care about the emotional
        contrast of "but".

## Subtopic 1: Satisfiable vs. Valid vs. Contradiction

### The Intuition
   Every logical formula describes a set of possible states (valuations of its
   variables). You can classify any formula into one of these three buckets by
   looking at its truth table outputs:
   
   1. Valid (Tautology): The formula outputs `True` in every single row. It is
      bulletproof.
   2. Contradiction (Unsatisfiable): The formula outputs `False` in every single
      row. It is logically broken.
   3. SATISFIABLE: The formula outputs `True` in at least one row.
      (Note: by definition, all VALID formulas are also SATISFIABLE. If an exam
      asks you to prove something is "satisfiable but not valid", it means it
      outputs True sometimes, and False sometimes--it is contingent on the 
      inputs.)



## Subtopic 2: Arguments and Validity      

### The Intuition
   An "argument" is just a list of pemises that leads to a conclusion:
   $P_1, P_2 \dots \models C$.

   An argument is considered VALID if it is strictly impossible for all the 
   premises to be true while the conclusion is FALSE. If you can find even one
   state where the premises are met but the conclusion fails, the entire 
   argument is INVALID. 


## SUBTOPIC 3: Direct Argument (The Exam Method)

### The Intuition
   Examiners do not want to grade 8-row or ... truth tables unless they 
   explicitly ask for one. "Direct Argument" is a targeted strike. Instead of
   mapping the whole state space, you play logical detective: you assume a 
   specific outcome and work backwards to force variables into specific 
   True/False states.

### The Strategy for Direct Arguments:
   - TO PROVE A CONTRADICTION: Assume the formula evaluates to `True`. Work
     backwards. You will eventually be forced to assign a variable to be both
     True and False simultaneously. This clash proves your assumption was 
     impossible, so it must be a contradiction.
   - TO PROVE VALID: Assume the formula evaluates to `False`. Work backwards. 
     You will hit clash. Therefore, it can never be False.
   - TO PROVE "SATISFIABLE BUT NOT VALID": You don't need a clash. You just need
     to play matchmaker twice. Find one state that makes it `True` (proves
     satisfiable), and one state that makes it `False` (proves not valid).

## Subtopic 2: Equivalence Laws (The Refactoring Rules)

### The Intuition
   Think of equivalence laws exactly like refactoring conditional logic. If you
   are writing a complex guard clause in Haskell or a `while` loop condition in
   Kotlin, you don't test every possible runtime state in your head; you apply
   algebraic rules to simplify the expression down to its cleanest form.

   The formulas are logically equivalent ($A \equiv B$) if their formulas in a
   truth table are identical. Equivalence laws are just memorised shortcuts for
   these identical columns.


### The Core Arsenal (What to Memorise for the Exam):
   You don't need to memorise all of them, but these four are non-negotiable for
   Imperial exams:

   1. IMPLICATION ELIMINATION: $A \to B \equiv \neg A \vee B$
      - Why it matters: You cannot easily distribute or manipulate arrows. Your
        first step in almost any algebraic proof is to nuke the arrows and turn
        them into ANDs/ORs.

   2. DE MORGAN'S LAWS: $\neg (A \wedge B) \equiv \neg A \vee \neg B$ (and vice
      versa).
      - THE VISUAL: When you distribute a negation into a bracket, it flips the
        sign of the variables AND flips the gate (AND <-> OR).

   3. DISTRIBUTIVITY: $A \vee (B \wedge C) \equiv (A \vee B) \wedge (A \vee C)$
      - The visual: Exactly like multiplying out brackets in GCSE math: 
        ... x(y + z) = xy + xz.

   4. DOUBLE NEGATION: $\neg \neg A \equiv A$.

## SUBTOPIC 3: Validity && Contradiction by Equivalence

### The Intuition
   We know from Block 2 that a Valid formula is always True (a tautology), and a
   Contradiction is always False.

   Proving this "by equivalence" means you take a terrifying, bracket-heavy 
   formula and apply your equivalence laws step-by-step until the whole thing 
   physically collapses into a raw $\top$ (True) or $\bot$ (False).

### The Collapse Triggers:
   You are always hunting for these specific paterns to force the collapse:
   
   - THE LAW OF EXCLUDED MIDDLE: $A \vee \neg \equiv \top$. (Either it's raining
     or it's not raining. Always true.)
   - CONTRADICTION: $A \wedge \neg A \equiv \bot$ (It is not raining AND it is
     not raining. Always False.)
   - IDENTITY LAW: $A \wedge \top \equiv A$, and $A \vee \bot \equiv A$    
   - DOMINATION LAWS: $A \vee \top \equiv \top$ (Once one side of an OR is True,
     the whole thing is True), and $A \vee \top \equiv \tee$ (Once one side of
     an OR is True, the whole thing is True), and $A \wedge \bot \equiv \bot$.

### The Exam Strategy:
   1. ELIMINATE: Turn all -> and <-> into $\vee$ and $\wedge$.
   2. PUSH: Use De Morgan's to push all $\neg$ symbols as deep into the formula
      as possible, right next to the variables.
   3. DISTRIBUTE/FACTOR: Look for common terms to pull out, or distribute terms
      to create $A \vee \neg A$ or $A \wedge \neg A$ pairings.
   4. COLLAPSE: Turn those pairings into $\top$ or $\bot$, and watch the rest
     of the formula disolve.


   ... the exact exam shape for this is "Using equivalence laws, rewrite the
   following into CNF" or "Use equivalence laws to show X is a contradiction..."

## Subtopic 1: Functional Completeness (The Universal Toolbox)

### The Intuition
   Imagine you are building a computer. You know that with three fundamental 
   logic gates--AND (^), OR (v), and NOT (¬)--you can construct a circuit for
   literally any possible truth table. The trifecta `{¬, ^, v}` is guaranteed
   to be "functionally complete."

   But engineers are cheap. They ask: Do we really need to manufacture all three
   gates? Can we build everything using just `{¬, v}`?

   The answer is yes. If you want to prove a smaller set of tools (like `{¬, v}`)
   is functionally complete, you do not need to prove it can build every formula
   in the universe from scratch. You only need to prove that it can SIMULATE THE
   MISSING TOOLS from the master toolkit.

### The Visual Framing: The Missing Tool Proof
   If your goal is to prove `{¬, v}` is complete, your only missing master tool
   is `^` (AND). If you can write an expression that uses only `¬` and `v`, but
   its truth table perfectly matches the truth table for $p \wedge q$, you win.
   (Hint: De Morgan's Law is the blueprint for this simulation).

### The Exam Reality
   Examiners will give you a weird set of connectives--sometimes standard ones 
   like `{¬, v}`, sometimes strange ones like NAND or NOR, or sets like 
   `{\bot, ->}`--and ask ou to prove they are complete. They will explicitly 
   state whether to use equivalence laws or truth tables for the proof. If they
   ask for truth tables, you literally draw the table for $p \wedge q$, then
   draw the table for your simulated versions, and show the final columns match.

## Subtopic 2: Normal-Form Manipulations (CNF / DNF)

### The Intuition
   Logic formulas can be written in infinite, messy ways. Normal forms are 
   strict templates. They force the logic into a perfectly flat, predictable
   shape, which makes it incredibly easy for a computer (or an eaxminer) to
   process.

   - CNF (Conjunctive Normal Form): An AND of ORs.
      - The mental model: A series of mandatory checkpoints. You must pass
        checkpoint 1 AND checkpoint 2 AND checkpoint 3. But at each checkpoint,
        you have multiple options to get through: (Option A OR Option B).
      - Shape: $(p \vee \neg q) \wedge (r \vee s) \wedge (\neg p)$

   - DNF (Disjunctive Normal Form): An OR of ANDs.
      - The mental model: Parallel pathways to victory. You win if you take Path
        1 or Path 2. But to complete a path, you must meet all its strict 
        conditions (Requirement A AND Requirement B).
      - Shape: $(p \wedge \neg q) \vee (r \wedge s)$

      
---
### The Visual Framing: Flattening the Tree
   If you look at the Formation Tree of a CNF formula, it is perfectly pruned to
   three layers maximum:
   1. The Root Layer: Only ^ (AND) joints.
   2. The Middle Layer: Only v (OR) joints.
   3. THE LEAF LAYER: Only "literals" (a raw variable p, or a negated variable
      ¬p).
