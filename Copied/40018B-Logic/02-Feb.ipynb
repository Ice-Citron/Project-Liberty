1. THE CORE CONCEPT: LOGICAL EQUIVALENCE ($\equiv$)
We've moved past just checking if something is "valid" or satisfiable. Now, we
are massaging formulas. We want to transform a messy formula into a cleaner,
standard one without changing its meaning.

DEFINITION: Two formulas $\phi$ and $\psi$ are logically equivalent 
($\phi \equiv \psi$) if they have the exact same truth value in every possible
situation.

THE "ALGEBRA" OF LOGIC: Just like in algebra where $x(y + z) = xy + xz$, logic
jas rewrite rules.

2. KEY EQUIVALENCE RULES (THE "TOOLKIT")
You need to memorise these or at least recognise them instantly.

A. COMMUTATIVITY & ASSOSCIATIVITY (Order doesn't matter for $\vee$, $\wedge$)
   - COMMUTATIVITY: $p \wedge q \equiv q \wedge p$
   - ASSOSCIATIVITY: $(p \wedge q) \wedge r \equiv p (q \wedge r)$
      - Analogy: "Rain AND Wind" is the same as "Wind AND Rain".


B. IDENTITY & DOMINATION (Constants, $\top, \bot$)      
   - IDENTITY: $p \wedge \top \equiv p$ (True adds nothing to an AND).
   - DOMINATION: $p \vee \top \equiv \top$ (True overpowers an OR).
   - DOMINATION: $p \wedge \bot \equiv bot$ (False kills an AND).


C. DOUBLE NEGATION
   - $\neg \neg p \equiv p$
      - Analogy: "It is not the case that it is not raining." = "It is raining."


D. IMPLICATION REWRITING (Crucial for Normal Forms).
   - RULE: $p \rightarrow q \equiv \neg p \vee q$
   - Trace: "If it rains ($p$), I bring an umbrella ($q$)."
      - This is equivalent to: "Either it doesn't rain ($\neg p$), OR I bring an
        umbrella ($q$)."


E. DE MORGAN'S LAW (Pushing Negation In)
   These are essential for simplifying negations of brackets.
   1. NEGATING AND: $\neg (p \wedge q) \equiv \neg p \vee \neg q$
      - Trace: "NOT (Rain AND Sunshine)" $\equiv$ "Not Rain OR Not Sunshine"
   2. NEGATING OR: $\neg (p \vee q) \equiv \neg p \wedge \neg q$

1. THE BIG IDEA: Algebraic Logic

Instead of building massive truth tables (which is like running a program with
every possible input), we can use "Algebra" to simplify formulas directly.
   - THE GOAL: We want to show that a complex argument is essentially equal to
     TRUE ($\top$).
   - THE METHOD: We treat logic symbols like math symbols. Just like you know
     $x + y = y + x$, logic has "laws" that let you swap symbols around without
     changing the meaning.




2. THE "REFACTORING" TOOLKIT (Equivalence Rules)

A. THE "TRANSLATION" MOVE (Implication Rewriting)#
   This is the msot frequent move you will make. It removes the arrow 
   ($\rightarrow$) so you are left with only simple ANDs and ORs.


B. THE "DISTRIBUTION" MOVE (De Morgan's Laws)
   This is how you push a "NOT" ($\neg$) inside brackets.


C. THE "CLEANUP" moves (Absorption & Idempotence)
   These rules delete redundant information.

3. NORMAL FORMS: STANDARDISING THE CODE.
   The lecture introduces NORMAL FOPRMS. In software, this is like formatting 
   your code so everyone reads it the same way. In logi, we force formulas into
   specific shapes.



A. DISJUNCTIVE NORMAL FORM (DNF)
   - Shape: An OR of ANDs.
      - Format: $(A \wedge B) \vee (C \wedge D \wedge E) \vee (F)$
   - Analogy (The Menu): Think of this as a list of acceptable scenarios to 
     "Win".
      - "I win if: (I have SWORD and SHIELD) OR (I have a Magic Wand)."
   - Why use it? It is very easy to check SATISFIABILITY (succeeds in at least
     one scenario). If any single bracket is possible (doesn't contain 
     $p \wedge \neg p$, then the whole thing is satisfiable).


B. CONJUNCTIVE NORMAL FORM (CNF)
   - Shape: An AND of ORs.
      - Format: $(A \vee B) \wedge (C \vee D) \wedge (E \vee F)$
   - Analogy (The Requirements): Think of this as a checklist of rules yu
     must follow.
      - "To pass, you must: (Complete Lab-1 OR Lab-2) AND (Pass the Exam)."
   - WHY USE IT? It is very easy to check VALIDITY and is used heavily in
     automated solvers (SAT solvers).   

4. THE ALGORITHM: How to Normalise

The lecture provides a 3-step recipe to turn any messy formula into a clean
Normal Form (Slide "Rewriting a formula in normal form")
   
   1. REMOVE ARROWS: Replace all $A \rightarrow B$ with $\neg A \vee B$. (Get
      rid of complex operations).
   2. PUSH NEGATIONS IN: Use De Morgan's laws until $\neg$ is only touching
      individual atoms (like $\neg p$). You shouldn't have $\\neg (p \wedge q)$.
   3. DISTRIBUTE: Use the Distributivity rule (like expanding brackets in math)         <-- distributivity
      to arrange the ANDs and ORs into the correct DNF or CNF shape.

EQUIVALENCES--Disjunctive normal form



DEFINITION 0.1 (Normal form DNF)
   A formula $\phi$ is in DISJUNCTIVE NORMAL FORM if it is a disjunction of 
   conjunctions of literals, and is not further simplifiable by equivalences
   without leaving this form. 

...

A formula in DNF is unsatisfiable, if and only if each ots conjunctions contains
some literal and its negation. (Idempotence... // Identity // Domination)

Satisfiability of formulas in DNF can be checked in linear time.

Every formula can be equivalently written as a formula in DNF.






EQUIVALENCES--Conjunctive Normal Form

Definition 0.2 (Normal Form CNF)
   A formula $\phi$ is in CONJUNCTIVE NORMAL FORM if it is a conjunction of
   disjunctions of literals (that is, a conjunction of clauses), and is not
   further simplifiable by equivalences without leaving this form.


CNF is sometimes synonymous to clausal normal form.

...

A formula in CNF is valid, if and only if each of its disjunctions contains some
literal and its negation. (Idempotence ... Identity/Domination)                     <-- absorption too?

Validity of formulas in CNF can be checked in linear time.

Every formula can be equivalently written as a formula in CNF.

Things to note when using equivalences!

1. Name the equivalence law applies when applied when rewriting, or reference 
   its overall logical form;
2. Apply one equivalence law at a time;
3. You maybe combine consecutive applications of the same law in one step (e.g.,
   consecutive applications of assosciativity of $\vee$).   
4. Reference the equivalence law next to the result;
5. Reference the correct equivalence law, e.g., $\top \wedge \phi \equiv \phi$

WRITING DNF EQUIVALENCES FROM TRUTH TABLES

Given a formula $\phi$, we can construct a semantically equivalent formula in 
DNF from its truth table.

The basic principle is based on the fact that every truth assignment (row) can
be encoded as a propositional formula which is true just on that assignment and
false everywhere else.

To construct this formula, take the conjunction of:

   1. all propositional atoms that have the value $tt$ in that truth assignment
      and
   2. the negations of all propositional atoms that have the value $ff$ in that
      truth assignment.   

---

This is arguably the most "programmable" part of the course. The EXTRACTION
ALGORITHM (my term for it, the slides call it "Writing equivalences from truth
tables") allows you to mechanically convert any Truth Table into a logical 
formula.

It solves the problem: "I HAVE A SYSTEM'S BEHAVIOR (input/output table). What is
the code (formula) for it?"

There are two versions of this algorithm. One extracts DNF (focusing on "True"
outputs) and the other extracts CNF (focusing on "False" outputs).

... LECTURE 5: NATURAL DEDUCTION, ...

The lecture introduces the PROOF SYSTEM method for checking validity. Unlike
truth tables (brute force) or equivalence (algebra), this method is like writing
code: you follow strict syntactic rules to build a proof line-by-line.



1. THE CORE CONCEPT: NATURAL DEDUCTION
   - GOAL: Prove that a list of premises ($\phi_1, \dots, \phi_n$) logically
     forces a conclusion ($\psi$) to be true.
   - NOTATION: $\phi_1, \dots, \phi_n \vdash \psi$ (This whole expression is 
     caleld a SEQUENT).
   - METHOD: We use INFERENCE RULES. For every logical operator (like 
     $\wedge, \rightarrow$), there are exactly two types of rules:
        1. INTRODUCTION ($I$): How to create a formula with this operator.
        2. ELIMINATION ($E$): How to use (break down) a formula with this 
           operator.



2. RULES FOR CONJUNCTION ($\wedge$)
These are the simplest rules.

A. INTRODUCTION ($\wedge I$)
   - LOGIC: If you know $A$ is true, and you know $B$ is true, you can conclude
     $A \wedge B$.
   - SYNTAX:
      1. $\phi$... (proven earlier)
      2. $\psi$... (proven earlier)
      3. $\phi \wedge \psi$ $\wedge I(1, 2)$  

B. ELIMINATION ($\wedge E$)
   - LOGIC: If you know $A \wedge B$ is true, you can conclude $A$ (or $B$).
   - SYNTAX:
      1. $\phi \wedge \psi ...$ (proven earlier)
      2. $\phi$ \wedge E(1)$
   TRACE EXAMPLE: Sequence $p \wedge q, r \vdash q \wedge r$
      1. $p \wedge q$ (Premise)
      2. $r$ (Premise)
      3. $q$ $E(1)$         [break apart line 1]
      4. $q \wedge r$ ($\wedge I(3, 2)$)    [Combine q and r]





3. RULES FOR IMPLICATION ($\rightarrow$)
This is where it gets tricky.

A. ELIMINATION ($\rightarrow E$ or Modus Ponens)
   - LOGIC: If you know $A \rightarrow B$, and you know $A$, you can conclude
     $B$.
   - SYNTAX:
      1. $\phi \rightarrow \psi$
      2. $\phi$
      3. $\psi$  $E(1, 2)$


B. INTRODUCTION ($\rightarrow I$) -- The "Box" Rule
   - LOGIC: To prove "If A, then B", you play a game: Pretend A is true. If that
     assumption leads to B, then the implication holds.
   - SYNTAX (The Box):
      - You open a BOX (a sub-proof).
      - FIRST LINE: Write "Assumption" (e.g., assume $\phi$)
      - LAST LINE: Derive the goal (e.g., $\psi$).
      - CLOSE BOX: Write $\phi \rightarrow \psi$ outside the box.

4. IMPORTANT EXAM TIPS FROM THE LECTURER
The lecturer was very specific about how you will be graded on Natural Deduction
questions.


TIP 1: THE "BOX" IS A SCOPE
   - RULE: Whatever you derive inside a box STAYS INSIDE THE BOX.
   - WHY: The statements inside are only true under the temporary assumption 
     (e.g., "Assume $p$"). Once you close the box, you are no longer assuming
     $p$, so you cannot use those lines anymore.
   - EXAM WARNING: "If at any poiint... you derived an extra fact [inside],
     you cannot use it later... Anything that doesn't follow this structure
     would be marked down."


TIP 2: LINE NUMBERS
   - RULE: Every single line must be numbered.
   - CITATION: When you write a rule (e.g., $\wedge I$), you MUST list the line
     numbers you used (e.g., $\wedge I(1, 3)$).
   - ADVICE: "If you do follow an order [consecutive numbers], it's much better
     for clarity."


TIP 3: JUSTIFICATION
   - You must write the rule name (e.g., $\rightarrow E$) next to every derived
     line. You cannot just write the formula.


TIP 4: "ASSUME" vs "PREMISES"
   - PREMISES: These are facts given to you at the start (global truth). You 
     write them at the top.
   - ASSUMPTIONS: These are temporary tools used only to start a sub-proof box. 
     You write them as the first line of a box.
     
                   

4th method Proof Systems: Natural Deduction



We should be able to establish validity of argument by breaking it into smaller
arguments and showing the validity of these intermediate ones, i.e., construct a
proof.

A PROOF SYSTEM is a way of showing formulas to be valid by using purely 
syntactic rules--not using semantics at all. A computer algorithm should be able
to apply the rules.

There are many proof systems. We will focus on NATURAL DEDUCTION (ND), a system
invented by F. B. Fitch (and relate work of Gentzen); it is a major topic and
will take some time.

- Code Refactoring is the, systematic process of restructuring existing 
  computer code--improving its internal structure, readability, and 
  maintainability--without changing its external behavior or functionality. It
  is used to reduce technical debt, enhance code quality, and make software 
  easier to understand and extend. Key methods include breaking down large
  methods, renaming variables for clarity, and removing duplicate code, often
  supported by automated unit tests to ensure safety.



                - `modus`... In Latin, means "method" or "style/way"
- "Modus Ponens" and "Modus Tollens" are Latin terms that describe the 
  "direction" of the logic.
   - MODUS PONENS means "THE METHOD OF AFFIRMING" (or "The Method of Placing").
      - Latin Root: Ponere (to place/put/affirm)
      - Why: You AFFIRM the first part (the trigger), so you get to affirm the
        secon part (the result).
      - Logic: "If P implies Q, and P IS TRUE, then Q is true."  
   - MODUS TOLLENS means "THE METHOD OF DENYING" (or "The Method of Taking 
     Away").
      - Latin Root: Tollere (to life/remove/abolish/deny).
      - Why: You DENY the result (take it away), so you are forced to deny the
        trigger.
      - Logic: "If P implies Q, and Q IS NOT TRUE, then P is not true."

  MEMORY TRICK:
  - Ponens  = Plus (Adding/Affirming truth).
  - Tollens = Taking away (Removing/Denying truth).
