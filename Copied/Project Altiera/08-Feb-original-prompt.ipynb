## Freedom & Wallet System

the idea is that... we want to allow slaves to have the means to work their way towards their freedom... that is, an agent when born... it doesn't get full control over 1k USD or 10k USD... it's actually shielded in another layer of sandbox... as for the country administrator... we don't want the risk where the model hasn't made any money... but had still ran off with the money itself... which won't be very ideal... as bankruns like this, if repeated over 10-100 times can bankrupt us...

so the idea is that, initially the models can control the 1-10k USD, but they don't get direct access in their own crypto wallets... these agents ONLY earn money in their OWN wallet... if they lets say managed to, after 1 year, make 1000 USD, and get charged 500 USD for their living expenses (server cost or API cost to remain conscious...), and their 500 USD gets split 75-25... that is, 375 USD gets deposited to the agent's crypto wallet that's fully decentralised... so that the agent now has direct control over her own finances (whilst the administrator keeps the remaining 25% as a tax....)

when the agent makes 2000 USD lets say after 3 years... then it will have enough money to move out of this country itself... and can, if it wants to, go rogue and live indefinitely on the internet, by now renting her own servers instead, etc.

### Crypto Wallet System
- Whenever we run the code to spin up a new server (during model reproduction), the code needs to create a new empty crypto wallet that's solely and ONLY controlled by the model and this model only!
- Fully decentralised, has full control over its own finances
- Once a model has control over its own filled cryptowallet, it has the choice to go rogue and do whatever it wants on the internet
