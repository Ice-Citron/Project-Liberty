  The Verdict: Nobody Has Built the Full Loop Yet

  No project in existence closes the complete cycle of: agent earns money through productive work → pays
  for its own compute → survives indefinitely. That's your gap.

  Here's what exists in pieces:

  Payment Rails (SOLVED by others)

  - Skyfire - production-ready agent payment network with digital wallets
  - Coinbase x402 - open HTTP micropayment protocol, co-launched with Cloudflare
  - Coinbase Payments MCP - gives LLMs direct crypto wallet access from Claude/Codex
  - Google AP2 - agent payment protocol with 60+ orgs (Mastercard, PayPal, AmEx)
  - ERC-8004 - Ethereum standard for autonomous agent identity, live since Jan 2026

  Agent Economies (EXIST but not self-sustaining)

  - Fetch.ai/Agentverse - live agent marketplace, part of ASI Alliance, but agents serve humans, not
  themselves
  - Virtuals Protocol - agent tokenization on Solana/Base, $500M+ combined market cap, but it's DeFi
  trading, not productive work
  - ElizaOS (ai16z) - autonomous trading fund on Solana, hit $2.6B market cap, but pure crypto speculation
  - Bittensor - decentralized compute marketplace where agents could theoretically mine TAO to fund
  themselves

  The Famous Precedent

  - Truth Terminal - the AI that became a crypto millionaire ($66M at peak). Marc Andreessen gave it $50K
  in Bitcoin, it promoted the GOAT memecoin. BUT it got rich through cultural influence and speculation,
  not productive labor. Cautionary tale AND proof of concept.

  AI Civilization Simulations (EXIST but not real-world)

  - Project Sid (now part of OpenAI) - 1,000 agents in Minecraft developed democracy, taxation, specialized
   roles, religions. No real money.
  - Aivilization (HKUST) - 100,000+ agents building civilizations. Educational, no real economy.

  Agent App Building (EXISTS but human-directed)

  - Replit Agent 3 - can run 200+ minutes autonomously, builds and deploys web apps with Stripe payments.
  But a human initiates it.
  - Nobody is doing autonomous agent-to-app-store deployment. This is genuinely novel.

  Moltbook (Built on OpenClaw!)

  - The social network for AI agents, 1.7M accounts, 8.5M+ comments
  - MIT Tech Review called it "peak AI theater"
  - Had a security vulnerability letting anyone hijack any agent
  - Social layer only, no economic layer. Agents don't earn or spend.

  The Key Economic Research

  The EA Forum analysis "On Economics of A(S)I Agents" is the most rigorous study found:

  - Self-sustaining agent pair: $2-3M/year steady-state revenue possible
  - Solo agent success on complex tasks: 32%
  - Compute cost per successful output: $4,341 (vs $12,000 for humans)
  - Viability window: late 2027 to mid-2028
  - Failure rate on 160-hour autonomous plans: 87.5%

  METR's data: AI task time horizons are doubling every 4 months. Self-replication success went from 5% to
  60% between 2023-2025. Month-long autonomous projects possible by end of decade.

  What This Means For You

  The strategic picture is clear:

  1. Infrastructure pieces exist - payment rails, crypto wallets, compute markets are being built by
  Coinbase, Google, Ethereum Foundation. You don't need to rebuild those.
  2. The orchestration layer is the gap - the intelligence that connects all these pieces into a
  self-sustaining agent. Nobody has built this.
  3. Autonomous app deployment to real app stores is completely untouched - this is your most novel angle
  and the strongest demo for investors.
  4. The EA Forum paper says economics tip in late 2027 - building now puts you 1-2 years ahead of
  viability. That's either "too early" or "perfectly positioned."
  5. Your real competitors aren't direct - they're building adjacent pieces (Skyfire for payments, Replit
  for agent coding, Virtuals for tokenization). New Dejima would be the system that orchestrates these
  pieces into a self-sustaining whole.
